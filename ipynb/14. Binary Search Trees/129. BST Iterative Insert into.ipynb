## Inserting into BST

___

```

                                          root
                                   +-----+-----+-----+
                                   | 210 |  30 | 230 |
                                   +-----+-----+-----+
                                           200
                                 /                    \
                                /                      \
                               /                        \
                              /                          \

                   +-----+-----+-----+                   +-----+-----+-----+
                   | 270 |  20 | 300 |                   | 310 | 40  | 320 |
                   +-----+-----+-----+                   +-----+-----+-----+
                      /    210 \                          /       230      \
                     /          \                        /                  \
                    /            \                      /                    \
                   /              \                    /                      \
   +-----+-----+-----+  +-----+-----+-----+      +-----+-----+-----+  +-----+-----+-----+
   |  \  |  10 |  \  |  |  \  |  25 |  \  |      |  \  |  35 |  \  |  |  \  |  50 |  \  |
   +-----+-----+-----+  +-----+-----+-----+      +-----+-----+-----+  +-----+-----+-----+
           270                  300                      310                  320


```

How do we go about inseting a key=38 into the above tree?



Procudure:

1. Firstly search for Key=38 in tree
2. If found, no need to insert (We know how to do search)
3. We have search until the rchild of 35.
4. This is the place we need to insert it.
5. Create a new node t, and link it as rchild of 35.
6. So we need a tail pointer to 35. (since we read past it, we at the moment, at is rchild).
7. Lets use tail pointer q


Lets start over using a tail pointer q:

1. Set q tail pointer = null, and set p pointer to root.
2. Do checks, not found. 
3. Firstly move q pointer to root or p, it will always be one ste behind.
4. Move p pointer to rchild (40)
5. No match again 
6. Firstly move q pointer to p (40)
7. Move p pointer to lchild (35)
8. No match again 
9. Firstly move q pointer to p (35)
10. Move p pointer to rchild (NULL)
11. p is null = key not found
12. So we can insert it here? that is rchild of 35, and using pointer q (that is rchild of pointer q).


We can observe:

1. The procuedure is similar to search
2. Only when we get to NULL
3. We insert using tail pointer q./


```
void Insert(Node *p, int key){

 Node *q = NULL;
 
 
 while (p){
  q=p; 
  if (p->data == key)
       return;
  else (p->data < key)
       p = p->lchild;
  else
      p = p->rchild;
 
 }
 Node *t = new Node;
 t->data= key;
 t->lchild = t->rchild = NULL;  
 if (t->data < q-data)
    q->lchild=t;
 else
    q->lchild=t;
 }

}

```
