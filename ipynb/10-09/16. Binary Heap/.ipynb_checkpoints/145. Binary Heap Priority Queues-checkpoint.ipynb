## Priority Queues

___

This is normal operation of priority queue (without yet using HEAP)
It is related to insertion and deletion of elements.

We have the following elements:

$$elements:4,9,5,10,6,20,8,15,2,18$$

Elements are inserted along with their priority. We can determine that elements with largest value has the highest priority. And when we delete, we delete element with the highest priority.



Lets insert into array:

1.We insert into array as we get it.
 

$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|}
\hline 
4\T&9\T&5\T&10\T&6\T&20\T&8\T&15\T&2\T&18  \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8\T&_9\T&_{10}\\\hline 
_\T&_\T&_\T&_\T&_\T&_\T&_{} \\\hline 
\end{array}
$$

2. And the above takes : $O(1)$. Constant time, we simple compying the element.
3. And to delete, must  take highest element, that is 20.
4. So we will search for element and shifting elements (we cant have blank spaces):, which n for searching and n for shifting.
5. Delete will take $O(n)$.

But insertion takes constant time, deletion takes order of n times. Is there a way for deletion also to take constant time.

Yes, we can create a HEAP or the keys in the array. And how much time does it take to insert? 

$$O(log_2{n})$$ 

And the above is much smaller than n.

So HEAP allows us to implement Priority Queues in much efficient way.


So with Normal Priority Queue:

1. Insertion = $O(1)$
2. Deletion = $O(n)$

Implementing Priority Queue using HEAP

1. Insertion = $O(log_2{n})$
2. Deletion = $O(log_2{n})$

So ucase for HEAP is a Priority Queue, and best data structure for it.
