## Adjacency Matrix

____

Here we have the following directed graph:


```
                                 +---+        +---+
                                 | 1 | +----> | 2 |
                                 +---+<--+    +---+
                                   +     |      +
                                   |     |      |
                                   |     |      |
                                   v     +---+  v
                                 +---+       ++---+
                                 | 4 | +----> | 3 |
                                 +---+        +---+
                                 
```


$$G=(V,E)$$
$$\vert{V}\vert=n=4$$
$$\vert{E}\vert=e=5$$

And we take a $4x4 matrix as the number of vertices are 4. Sp the size of the matrix is dependent on the number of vertices.


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  \\ 
{_2}  \\
{_3}  \\
{_4}  \\
\end{array}
\begin{bmatrix}
{} & {} & {} & {} & {}\\
{} & {} & {} & {} & {}\\
{} & {} & {} & {} & {}\\
{} & {} & {} & {} & {}\\
\end{bmatrix}
$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  & {_2}  & {_3}   & {_4}
\end{array}
$$


Remember
1. Each row and column represent a vertex.
2. And combination of row and column (together) represents an edge
3. But here the direction plays role:
    + Outgping edge, starts from from row to column combination
    + Incoming edge, does not get recorded (as in undirected graph)
    
    
$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  \\ 
{_2}  \\
{_3}  \\
{_4}  \\
\end{array}
\begin{bmatrix}
{0} & {1} & {0} & {1} \\
{0} & {0} & {1} & {0} \\
{1} & {0} & {0} & {0} \\
{0} & {0} & {1} & {0} \\
\end{bmatrix}
$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  & {_2}  & {_3}   & {_4}
\end{array}
$$

But how do we check incoming flows? We read the column first, then the row.

To get Out Degree:
    + We read row first, then column, that is A[i,j]
    
To get In Degree:
    + We read column first, then row, that is A[j,i]

So indegree of vertex 1 is 2, since:
    + A[1,2]
    + A[1,4]
    
We take row 1, and see which columns has value greater than 1, for row 1

So outdegree of vertex 1 is 1, since:
    + A[2,1]
    
We take column 1, and see which rows has value greater than 1, for column 1

    

We have similar situation as for undirected grap. That is we need to process all the elements to checkif 0 or 1.

$$n \times {n} = n^2 =O(n^2)$$



## Adjacency List
___

We have the followiong graph:


```
                                 +---+        +---+
                                 | 1 | +----> | 2 |
                                 +---+<--+    +---+
                                   +     |      +
                                   |     |      |
                                   |     |      |
                                   v     +---+  v
                                 +---+       ++---+
                                 | 4 | +----> | 3 |
                                 +---+        +---+
                                 
```

Again as, above, we only interested in the vertexes going out.

```
                   ARRAY             LINKED LIST

                   +---+     +---+---+    +---+---+
                  1| +-----> | 2 | +----> | 4 | / |
                   +---+     +---+---+    +---+---+
                  2| +-----> | 3 | /-|
                   +---+     +---+---+
                  3| +-----> | 1 | /-|
                   +---+     +---+---+
                  4| +-----> | 3 | /-|
                   +---+     +---+---+

```
Note here in adjecency list (compared to adcency matrix), we only have outdegree.

That is to get outdegree of 1. We get value in array (1)(start), and we counf the number of links to it. We see its two.

But to get the indegree, we dont start with the array value, we start by counting how linkes there are of value 1, in the linked list. We see there are 1. 

However, if the indegree is very important, then  we can use one more type of Adjacency list, which is called Inverse Adjacency list. So instead of representing outgoing flows, we will represent incoming flows.
 

``` 
                   ARRAY             LINKED LIST
                   +---+     +---+---+
                  1| +-----> | 3 |   |
                   +---+     +---+---+
                  2| +-----> | 1 | /-|
                   +---+     +---+---+    +---+---+
                  3| +-----> | 2 | +----> | 4 | / |
                   +---+     +---+---+    +---+---+
                  4| +-----> | 1 | /-|
                   +---+     +---+---+
```

So this is a requirement, then we can create an inverse list.





We can also show Compact or Inverse Compact List.


For analysis, we can say (not twice like unidirected graph):

When we analyse this method, we have set of vertices and set of edges(which is taken twice):

$$\vert{V}\vert+\vert{E}\vert$$

$$n+e= n+n = O(n)$$
