## Inserting into LL


Here we want to insert a new node. We need to mention the position we want to insert the new node. That is at what index. In arrays we start at index 0. Here we going to start at index 1. lets give the nodes an index. 


What are the place we can insert a new node:

1. Before the HEAD or first node.
2. After a given (index) position.
3. After the last node.

So if we say we want to insert at index 4, then that means we cant insert AFTER 4th node. 

(Note we call out HEAD node, first)

STEPS for creating a node before first node:

1. Create the new node with help of pointer t (Node *t = new Node)
2. Initialise the node, i.e. whatever data we need to assign to it here (t->data = x)
3. New node t, should now point to first node (t-next = first)
4. Move first to node t, since t will now become first node. (first = t) 


STEPS for inserting a node at given position, e.g. index 4:

1. Create the new node with help of pointer t (Node *t = new Node)
2. Initialise the node, i.e. whatever data we need to assign to it here (t->data = x)
3. Need to know node at index 4, and it position need to be updated: So we need to get pointer to previous node.
4. Lets take a pointer p, to traverse the list until we get to position 4. 
5. Thats is if index to insert at is at n, then p should step n - 1 times (previous node). Set p = first. And traverse p in loop until n - 1 times.
6. And t should be pointing to where node at index 4 was pointing (t->next = p->next)
7. And p should be pointing to t (p->next = t)

STEPS for inserting a node after the last node:

1. Maintain a pointer of last node (then we insert at constant time)
2. Or traverse to last node.



See Functions below for further information.

```c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
```

```c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
```

```c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    void InsertLast(int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
   
}
```

```c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
```

```c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
```

```c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
```

```c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
```

```c++
void LinkedList::InsertLast(int x){
    Node *t=NULL, *p=first;
    Node *last=NULL;
    
    
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    //Check if first position
    //count number of elements
    for(int i =0; i<Length()-1;i++){
      cout<<"p:"<<p->data<<endl;  
      p=p->next;
     
        
    }
    
    last = p;
    cout<<"p:"<<p<<endl;  
    cout<<"p->data:"<<p->data<<endl;  
    cout<<"last: "<<last<<endl;
     
    
    
    //There are no nodes
    if(first==NULL)
        first=last=t;
    else{
        
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        
        
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        last->next=t;
        last=t;
    }
    
   
}   
```

```c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}   
```

```c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
```
