## Quadraric Probing

___


This is a collision resolution technique. This technique falls under Open Adrressing, but Closed hashing.

Idea here: We have two hash functions, one the basic hash function, and when we get a collision the other hash function will help us.

We have default one, which also defines the size of the hash table:

$$ h1(x) = x\%{10}$$


Then we get the second has function:


$$ h2(x) = R - (x\%{R})$$

Where R is prime number smaller than size of indexes in has table, so prime number closest to max number 9, it is 7.


So lets modify our function, to reflect this:

$$ h2(x) = 7 - (x\%{7})$$

And this second has function must have the following objective

+ Must never give an index 0
+ It must probe all the locations, that is whenever there is a collision, it should not give the indexes in same pattern. It should give different indexes, such that all the locations are utilized.

So complying to these two objectives, we can thus define our own 2nd hashing function as well.


Then we also have the following modified hash function, if the defualt hash function gives us a collision:


$$ h'(x) =(h1(x) + i\times h2(x))\%10 \text{ where } i = 0,1,2...$$




Lets us the following key space and hash table to demonstrate double hashing:




$$5,25,15,35,95$$. And using the double hashing formula we get the following:

```
              +---+---+        +---+
              | 5 | 5 |       0|   |
              +---+---+        +---+
              | 25| 5 |       1| 15|
              +---+---+        +---+
              | 15| 5 |       2| 35|
              +---+---+        +---+
              | 35| 5 |       3|   |
              +---+---+        +---+
              | 95| 5 |       4| 95|
              +---+---+        +---+
              |   |   |       5| 5 |
              +---+---+        +---+
              |   |   |       6|   |
              +---+---+        +---+
              |   |   |       7|   |
              +---+---+        +---+
              |   |   |       8| 25|
              +---+---+        +---+
              |   |   |       9|   |
              +---+---+        +---+

```

Lets take the first collusion (25) ONLY:

$$ h1(x) = x\%{10}$$
$$ h1(25) = 25\%{10}=5$$

$$$$

$$ h2(x) = 7 - (x\%{7})$$
$$ h2(25) = 7 - (25\%{7})$$
$$ = 7 - 4=3$$

$$ h'(x) =(h1(x) + i\times h2(x))\%10$$

where i = 1

$$ h'(25) =(5 + 1\times 3)\%10$$
$$ =(5 + 1\times 3)\%10$$
$$ =(5 + 3)\%10=8$$

So 25 will get index 8 position.
