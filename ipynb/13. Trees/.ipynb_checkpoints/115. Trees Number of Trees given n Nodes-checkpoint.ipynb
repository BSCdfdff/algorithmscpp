## Formula 1 :  Catalan Number

___


**Formula 1**

First formula for getting the number of binary trees, given the number of nodes ($n$):

$$
\text{Catalan Number}\;\;=\;\;
T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}
$$

If we are given $5$ nodes (vertices):

$$
T(5) = \frac{^{2\times5}C_{5}}{5+1}
$$


$$
\frac{^{10}C_{5}}{6}
$$

$$
\frac{\frac{10\times9\times8\times7\times6}{5\times4\times3\times2\times1}}{6}
$$

$$
\require{cancel}
\frac{\frac{10\times9\times8\times7\times\cancel{6}}{5\times4\times3\times2\times1}}{\cancel{6}}
$$


$$
\require{cancel}
\frac{\cancel{10}\times9\times8\times7}{\cancel{5}\times4\times3\times\cancel{2}\times1}
$$

$$
\require{cancel}
\frac{\times9\times8\times7}{\times4\times3\times1}
$$

$$
\require{cancel}
\frac{\times9\times\cancel{8}2\times7}{\times\cancel{4}\times3\times1}
$$

$$
\require{cancel}
\frac{\times\cancel{9}3\times2\times7}{\cancel{3}\times1}
$$

$$
\require{cancel}
\frac{3\times2\times7}{1} \;\;=\;\; 42
$$

$$
T(5) = 42
$$


We call this a **COMNINATION FORMULA**

## What is Maximum Height (given n nodes)

___


For 3 nodes, the maximum height will be 2 nodes.

Remember height start counting from 0.

But how many trees with maximum height of 2, given 3 nodes?

Is there a formula for it? (No of trees with maximum height)

$$n = 3,  = 4 = 2^2$$

$$n = 4,  = 8 = 2^3$$

$$n = 5 = 16 = 2^4$$


Formula for n nodes:

$$=2^{n-1}$$



## Formula 2 :  Catalan Number

___

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
 n\T&0\T&1\T&2\T&3\T&4\T&5\T&6  \\\hline
T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}\T&1\T&1\T&2\T&5\T&14\T&42\T& ?     \\\hline
\hline
\end{array}
$$

We have calculated the catalan number up till 5. What is 6? Instead of using formula 1, lets derive the answer another way....

Starting at beginning and end, then working inwards:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
 n\T&0\T&1\T&2\T&3\T&4\T&5\T&6  \\\hline
T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}\T&1\T&1\T&2\T&5\T&14\T&42\T& ?     \\\hline
\T&_1\T&_2\T&_3\T&_3\T&_2\T&_1     \\\hline
\hline
\end{array}
$$

$$ 1\times42+1\times14+2\times5$$


Then swap, then move back outwards again

$$ 5\times2+14\times1+42\times1$$

That is:

$$ T(6)=(1\times42+1\times14+2\times5) + (5\times2+14\times1+42\times1)=132$$

Which is,

$$ T(6)=T(0)\times{T(5)}+ T(9)\times{T(4)}  + T(2)\times{T(3)} + T(3)\times{T(2)} + T(4)\times{T(1)} + T(5)\times{T(1)} $$

So each term is a sum of two value, and can be written as:


$$\sum T(\text{increasing}) \times T(\text{decreasing}) $$

But what values?

Onside its increasing, and other side is decreasing.

$$\sum_{i=1}^n T(i-1) \times T(n-i) $$


This a another way to get the number of binary trees, given n nodes.


We call this a **RECURSIVE FORMULA**

 ## Labelled Nodes
 
 ___
 
 Now we have n nodes but that are labelled: A B C
 
 Now, how many different binary trees can be generated.
 
 Bur we know the answer for unlabelled nodes of 3, we get 5 trees and we used the following formula: 
 
$$ 
 T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}
$$

But labelled means, we we just take one tree, we can have the following tree combinations of labels:


```
               +-+    +-+    +-+   +-+     +-+   +-+
               +A+    +A+    +B+   +B+     +C+   +C+
               /      /      /     /       /     /
              /      /      /     /       /     /
             /      /      /     /       /     /
            /      /      /     /       /     /
          +-+    +-+    +-+   +-+     +-+   +-+
          +B+    +C+    +A+   +C+     +A+   +B+
         /      /      /     /       /     /
        /      /      /     /       /     /
       /      /      /     /       /     /
     +-+    +-+    +-+   +-+     +-+   +-+
     +C+    +B+    +C+   +A+     +B+   +A+

```

Thats is $n!=6$ So one shape, can be filled in $6$ different ways or $n!$

So we have formula for labelled nodes:

$$ 
 T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1} \times {n!}
$$

And we can say:

$$ 
 T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1} \text{ is for Shapes}
$$

And


$$ 
 T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}  \text{ is for Filling (labels)}
$$
