## Reverse a  DLL

___

$6\iff{9}\iff{3}\iff{7}\iff{2}$

The main idea here is basically swap the node and node next to it, we interchanging positions. and then we move to next node.

And by swapping pointer:

1. Make next previous
2. And make previous next
3. And in order to do this we need to have temp variable first to store next.
Lets swap 6 and 9.

1. Create a pointer p and set it on first node. 
2. Make p's prev p's next  
3. And since its first node, make p-s next equal to NULL (**difference from rest loop**)
4. And since p's next was 9, but now p's prev is 9, so to move to pointer p to next node we traverse using p's prev (```p=p->prev```)

Lets swap 9 and 3
1. Pointer p is now on 9. And we want to swap 9 and 3.
2. Make Make p's prev p's next (9 was pointing to 6, now will point to 3) (swap*)
3. Make p's next p's prev (9 was pointing to 3, now will point to 6) (swap*)
4. And since p's next was 9, but now p's prev is 9, so to move to pointer p to next node we traverse using p's prev (```p=p->prev```)

And so we continue.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{ 
public:
    Node *prev;
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display(); 
    LinkedList(int A[], int n);
    int Length(); 
    void Insert(int index, int x);
    int Delete(int index);
    void ReverseDLL();
    
 
} 
````

````c++
void LinkedList::ReverseDLL(){
   Node *p=first;
   Node *temp;
    
   while (p){
      temp=p->next;
      p->next=p->prev;
      p->prev=temp;
      p=p->prev;
      if(p!=NULL && p->next==NULL)
          first = p;
    }
}    
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    
    
    int x = -1;
    
    if(index <0 || index > Length())
        return -1;  
    
    
    if(index==0){
        p=first;
        if (first->next!=NULL){
            first = first->next;
            first->prev=NULL;
            x=p->data; 
            delete p;
        }
        else{
            x=p->data; 
            delete p;
        }
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
           p=p->next;
        }
        
        p->prev->next = p->next;
        
        if (p->next)
            p->next->prev = p->prev;
         
        x=p->data;
        delete p;
        
    }  
    return x;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p;
    
    if (first==NULL){
        first= new Node;
        first->prev=first->next=NULL;
    }        
    
    p=first;
    
    if(index <0 || index > Length())
        return;
    
    if(index==0){
        t = new Node;
        t->data = x;
        t->prev=NULL;
        t->next = first;
        first->prev=t;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        
        t = new Node;
        t->data = x;
        
        t->next= p->next;
        t->prev = p;
        
        if (p->next)
          p->next->prev = t;      
        
        p->next = t;
      
    }
}   
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->prev=first->next=NULL;
    last = first; 
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i]; 
        t->next=NULL;
        //dont agree
        //last next is null???
        //does no harm
        //t->next=last->next;
        
        t->prev = last;
        last->next=t;
        last=t; 
    }     
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    if (first->next==NULL){
         cout<<first->data<<" ";
    }
    else{
        while(p){
            cout<<p->data<<" ";
            p=p->next;
        }
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    
    if (first->next==NULL)
         len++;
    else{
        while(p){
            len++;
            p=p->next;
        }
    }
    return len;
}
````
