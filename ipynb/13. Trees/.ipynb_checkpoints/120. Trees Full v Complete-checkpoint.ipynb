 ##  Full BT
 
___
 
 ```
                         +------------+
                               A                      +<-----+ 0
                         +------------+               |
                         /             \              |
                        /               \             |
                       /                 \            |
                      /                   \           |
                    +-+                   +-+         |
                     B                     C          v<-----+ 1
                    +-+                   +-+
                    / \                   / \         +
                   /   \                 /   \        |
                  /     \               /     \       |
                +-+    +-+             +-+   +-+      |
                 D      E               F     G       v
                +-+    +-+             +-+   +-+       <-----+ 2
 
                            n=7
                            Max
                       
```
$$$$ 
$$ 
T=
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&D\T&E\T&F\T&G   \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7  \\\hline 
\end{array}
$$ 
$$$$
 
 Must have the following requirements:
 
 
 1. It has the max nodes(given h height) . That is if h=2, then it has the max number of nodes for that height. And we do have formula for it:
  $$$$
  
 $$_\text{MAX NODES} = 2^{h+1}-1=2^3-1=7$$
 
 $$$$
 2. Does not have gaps, if you complete array
 3. Nodes filled from left to right, without skipping nodes


$$$$

By filling the array, and there are **NO** gaps, its a FULL BT


 
 

##  Complete BT
 
___

```
                         +------------+
                               A                      +<-----+ 0
                         +------------+               |
                         /             \              |
                        /               \             |
                       /                 \            |
                      /                   \           |
                    +-+                   +-+         |
                     B                     C          v<-----+ 1
                    +-+                   +-+
                    / \                   /           +
                   /   \                 /            |
                  /     \               /             |
                +-+    +-+             +-+            |
                 D      E               F             v
                +-+    +-+             +-+             <-----+ 2

```
$$$$

$$ 
T=
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&D\T&E\T&F   \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6  \\\hline 
\end{array}
$$ 
$$$$

Must have the following requirements:

1. Does not need comply with first requirement of Full BT, provide requirement 2 is met.That is, it fulls all positions in an array, without having a gap.
2. Does not have gaps, if you complete array


Here is an example of a BT that is NOT complete. It leaves gaps'

```
                         +------------+
                               A                      +<-----+ 0
                         +------------+               |
                         /             \              |
                        /               \             |
                       /                 \            |
                      /                   \           |
                    +-+                   +-+         |
                     B                     C          v<-----+ 1
                    +-+                   +-+
                    /                       \         +
                   /                         \        |
                  /                           \       |
                +-+                          +-+      |
                 D                            G       v
                +-+                          +-+       <-----+ 2


```

By completing the arra, from top to bott, left to right we get the following gaps:


$$$$

$$ 
T=
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&D\T&-\T&-\T&G   \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7   \\\hline 
\end{array}
$$ 
$$$$


## Strict BT

___

Here whe can have $0$ or $2$ children or $n$ children. That is children=${0,2}$. CANNOT have degree $1 $(as per our example). Must have $0$ or $max$ degree, nothing in between.

What does that it create? It will create gaps in array.

Example 1:

```
                         +------------+
                               A                      +<-----+ 0
                         +------------+               |
                         /             \              |
                        /               \             |
                       /                 \            |
                      /                   \           |
                    +-+                   +-+         |
                     B                     C          v<-----+ 1
                    +-+                   +-+
                                          / \         +
                                         /   \        |
                                        /     \       |
                                       +-+   +-+      |
                                        D     E       v
                                       +-+   +-+       <-----+ 2

```

$$$$

$$ 
T=
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&-\T&-\T&D\T&E   \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7   \\\hline 
\end{array}
$$ 
$$$$


Above:

1. Strict $\require{enclose}
\enclose{box}[mathcolor="green"]{\unicode{x2714}}({0,2})$
2. Complete $\require{enclose}
\enclose{box}[mathcolor="red"]{\unicode{x2718}}(gaps)$

Remember Strict allows blank spaces, complete does not


Example 2:

```
                                                +------------+
                               A                      +<-----+ 0
                         +------------+               |
                         /             \              |
                        /               \             |
                       /                 \            |
                      /                   \           |
                    +-+                   +-+         |
                     B                     C          v<-----+ 1
                    +-+                   +-+
                    / \                               +
                   /   \                              |
                  /     \                             |
                +-+    +-+                            |
                 D      E                             v
                +-+    +-+                             <-----+ 2

```

$$$$

$$ 
T=
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&D\T&E   \\\hline 
_1\T&_2\T&_3\T&_4\T&_5  \\\hline 
\end{array}
$$ 
$$$$


Above:

1. Strict $\require{enclose}
\enclose{box}[mathcolor="green"]{\unicode{x2714}} ({0,2})$
2. Complete $\require{enclose}
\enclose{box}[mathcolor="green"]{\unicode{x2714}}\text{(no gaps)}$

Thus we can say, Strict is also complete. But complete is not Strict.

Example 3:

