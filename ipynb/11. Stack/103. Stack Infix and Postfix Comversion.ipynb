## Introduction

___


We have the following cases:

1. Infix: Operand Operator Operand

$$a+b$$

2. Prefix: Operator Operand Operand

$$+ab$$

3. Postfix: Operand Operand Operator

$$ab+$$



The only one we know is Infix. Thats how we been taught. Where we have prefix, the operator is before the operands, and that means, that operations has to be performed on the operand(s). The same with Postfix, the operator following the operands, needs to be performed on the operand(s).



But why do we need prefix and postfix?



## But why do we need prefix and postfix?

__

We are given:

$$8+3\times(9-6)/2^2+6/2$$

Do we:

$$8+\frac{3\times(9-6)}{2^2}+\frac{6}{2} \text(    CORRECT)$$

Or do we:


$$8+3\times \frac{(9-6)}{2^2}+\frac{6}{2}$$

Which is the correct one? What is BODMA's Rule? Higher precedence gets executed first.

B : Brackets first (parentheses)

O : Of (orders i.e. Powers and Square Roots, Cube Roots, etc.)

DM: Division and Multiplication (start from left to right)

AS : Addition and Subtraction (start from left to right)

Using this rule we:

1. First do brackets : $(9-6)$
2. Then do power : $2^2$
3. Then Division $(9-6)/2^2$
4. Other division: $6/2$
5. Then multiplication: $3\times(9-6)/2^2$
6. Then additions: +, + (left to right)


So from the above what are we doing. We have to go into the expression, and see which one to perform, and jumping around from left to right. This is very inefficient way of doing things. We scanning the expression random times, based on precedence, going back and forth.

Is there a way to start scanning the expression from left to right, reading it sequentially, and execute the expression? Thats is get the answer by doing only one scan?

Yes we can use postfix form.

This is the postfix form for the above:
$$8+\frac{3\times(9-6)}{2^2}+\frac{6}{2} \text(    CORRECT)$$

$$8\;3\;9\;6 - 2^2\;/\;\times +\; 6\;2$$



Procedure:
1. First Use bodmas rule, to get the above expression
2. Read until you get the first operator, which is minus.
3. What do we do with the minus:  we perform this operation on the previous two(2) operands

$$ 9-6$$

4. Then move forward, look for next: $2^2$, then next operator is division.
5. What do we do with the division:  we perform this operation on the previous two(2) operands

$$\frac{9-6}{2^2}$$

From the above we can see the how we write postfix, we apply Bodma's Rule. But then do only one scan.Lets continue

The above seem to be written incorrectly, here is the correct version:

$$8\;3\;9\;6 - \times \;2^2\;/\;\times +\; 6\;2$$

So the purpose of postfix is to write expression, and with one scan get an answer. (fast)

How do we firstly get it in this postfix form?




## Postfix form Conversion

___

Table of Precedence:
$$

$$

When you write an expression you MUST make it fully parenthesed. First RULE, every expression MUST have parenthesis. If we do not add parenthesis, the compiler will add it, by using the precedence. That is, precendence does not mean order of execution. It means parenthesis on the order of precedence,i.e, put those ones in brackets.

We know the expression, so we MUST add it correctly.


Lets add parenthises to the following:

$$a+b*c$$

Will be :

$$a+(b*c)$$

Next, what is the remainding operator, its plus, so put that in brackets:

$$(a+(b*c))$$

So $a$ will be added to the result of $(b*c)$

So brackets give you a clear indication what operation needs to be performed on which operands.

Lets briefly look at prefix:

$$a+b*c$$


## Convert to Prefix

___


$$a+b*c$$

0. Brackets everything first

$$(a+(b*c))$$

1. Rule for Prefix: First we write operator, then operand.
2. Which bracket must be evaluated first, the multiplication bracket.
3. Write the previous statements as is, until you get there(multiplication bracket): 

$$(a+$$
4. Write the prefix operator, then two operands: 

$$(a+[\times ab]$$

6. Complete the expression:

$$(a+[\times ab])$$

Scan again to complete:

7.  Write the prefix operator:

$$+a\times ab$$

## Convert to Postfix

___


$$a+b*c$$

Rule here: Apply the operator at the end, but first, Bodmas rule to apply brackets first.

$$(a+(b*c))$$

1. Next which bracket do we convert first, multiplication, so go there:

We add the sqaure brackets just to determine what was done already (converted).

$$(a+[bc*])$$

2. Which is next bracket: result of what was done and $a +$. so we do that

$$abc*+$$

Again, procedure:.

1. Fully parenthesize everything, as per precedence table (to put in brackets)
2. Now we evaluate each bracket, as per precedence, (to put in pre/postfix form)

## Another example : Prefix

___


$$ a+b+c*d$$

Lets ignore brackets.


1. Which symbol as per table of precedence, multiplication

$$ a+b+[*cd]$$

2. Next, the addition, but which one,MUST be left to right

$$ [+ab]+[*cd]$$

3. Then the remaining plus is left over. Prefix the plus.

$$ [+ab]+[*cd]$$

$$++ab*cd$$

Note: We mostly use postfix form. Prefix form is rarely used.


## Another example : Postfix

___


$$ a+b+c*d$$

Lets ignore brackets.


1. Which symbol has higher precedence, multiplication

$$ a+b+[cd*]$$

2. Next, the addition, but which one,MUST be left to right

$$ [ab+]+[cd*]$$

3. Then the remaining plus is left over. Postfix the plus.

$$ [ab+]+[cd*]$$

$$ab+cd*+$$
