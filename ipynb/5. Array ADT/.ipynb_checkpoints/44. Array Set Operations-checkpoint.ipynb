```c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
```

```c++
class Array{
private:
    int *A;
    int size;
    int length;
    void Swap (int *x, int *y);
public:
    Array(){
        size=10;
        A=new int[10];
        length = 0;
        
    }
    Array(int sz){
        size=sz;
        A=new int[sz];
        length = 0;
        
    }
    ~Array(){
        delete []A;
    }
    void Display();
    void Insert (int index, int x);
    void Merge(Array arr);
    void INTERSECTION_US(Array arr);
    void INTERSECTION_Sorted(Array arr);
    
}
```

```c++
void Array::Swap (int *x, int *y){
    int temp;
    temp =*x;
    *x=*y;
    *y = temp;
}
```

```c++
void Array::INTERSECTION_US(Array arrB){
    int *C= new int[10];
    
    int i=0;
    int j=0;
    int k=0;
    
    int m = length;
    int n = arrB.length;
    
    i=0;
    bool found =false;
    //Insert from A into C
    //But check if first exists in B
    while (i<m){
        while (j<10){
            if (A[i] == arrB.A[j])
            {
                C[k++] = A[i++];
                found=true;
                break;
            }
            j++;
       }
       if (found ==false) 
       {
         i++;
       }
       found =false;
       j=0;
    }    

    //Display
    cout<<"This is C"<<endl;
    for(int i=0; i<k;i++)
      cout<<C[i]<<" ";      
    cout<<endl;   
}
```

```c++
void Array::INTERSECTION_Sorted(Array arrB){
    int *C= new int[10];
    
    int i=0;
    int j=0;
    int k=0;
    
    int m = length;
    int n = arrB.length;
    
    while (i<m && j<n){
        if (A[i] < arrB.A[j])
            i++ ;      
        else if (arrB.A[j] < A[i] )
            j++;
        else   //equal
        {
            C[k++] = A[i++];
            j++ ;      
        }    
    }    
    
    //Display
    for(int i=0; i<k;i++)
      cout<<C[i]<<" ";      
    cout<<endl;   
}
```

```c++
void Array::Merge(Array arrB){
    int *C= new int[10];
    
    int i=0;
    int j=0;
    int k=0;
    
    int m = length;
    int n = arrB.length;
    
    while (i<m && j<n){
        if (A[i] < arrB.A[j])
            C[k++] = A[i++];
        else     
            C[k++] = arrB.A[j++];
    }    
    
    //Deal for items left over
    //i starts where left off above
    //global
    for(;i<m;i++){
        C[k++]=A[i];  
    }
    
    for(;j<n;j++){
        C[k++]=arrB.A[j];
    }


    //Display
    for(int i=0; i<10;i++)
      cout<<C[i]<<" ";      
    cout<<endl;   
}
```

```c++
void Array::Insert(int index, int x){
   if  (index>=0 && index<=length){
     for(int i=length-1;i>=index;i--)
         A[i+1]=A[i];
     A[index]=x;
     length++;  
   }
}
```

```c++
void Array::Display(){
    for(int i=0; i<length;i++)
      cout<<A[i]<<" ";      
    cout<<endl;      
}
```
