## Inserting into Circular Linked List

___

We have the following

2-> 4-> 6 -> 8


We have the following scenarios (same as linear):

1. Inserting before HEAD
2. Insert at any other given position 

Insert at any other given position 

1. We want to insert at position $4$. When we say this it is after index $4$, that is between index $4$ and index $5$. 
2. This is similar to linear LL. Exactly the same.

Not further explanation required.



## Inserting before HEAD (Circular)


___


$$4-> 6 -> 8$$

And want to insert $2$ before first


Inserting before HEAD

1. Create a new node with pointer ```t```. We will be inserting this node before first.
2. Get the last node (Remember you need to step past first node first), that is where it becomes equal for first again.
3. Now we have our last node. And instead of pointing to First, make ```p->next=t```.
4. And point ```t->next = first.```
5. Optional set ```first= t```
6. Note. Also need to check if you have only one node: That is if ```first=NULL```, then ```first=t```, and to make it circular, ```first->next = first```.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL,*s=NULL,*result_t=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display();
    void Display_s();
    void Display_t();
    void Insert(int index, int x);
    void Insert_s(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    int LengthCircular();
    int CheckLoopLL();
    void DisplayCircularLL();
    void DisplayCircularLLRecursion(Node *p);
    void InsertCircularLL( int index, int x);
    
}
````

````c++
void LinkedList::InsertCircularLL(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > LengthCircular())
        return;
    
    
    if(index==0){
        t = new Node;
        t->data = x; 
        if (first==NULL){
           first = t;     
           first->next = first;
        }
        else{
            while(p->next != first)
                p=p->next;
            p->next=t;
            t->next=first;
            first = t;  
        }   
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t = new Node;
        t->data = x; 
        t->next=p->next;
        p->next=t;
    }
   
}    
````

````c++
void LinkedList::DisplayCircularLLRecursion(Node *p){
    static int cnt = 0;
    if (p!=first || cnt==0){
        cnt=1;
        cout<<p->data<<" ";
        DisplayCircularLLRecursion(p->next);
    }
    cout<<endl;
}
````

````c++
void LinkedList::DisplayCircularLL(){
    Node *p=first;
    
    do{
        cout<<p->data<<" ";
        p=p->next;
    } while (p!=first);

    cout<<endl;
    
}    
  
````

````c++
int LinkedList::CheckLoopLL(){
    Node *p, *q;
    p=q=first;
    do{
        p=p->next;
        q=q->next;
        //if q is not null, then assign q's next
        q=q?q->next:q;
       
    }while (p && q && p!=q);
    
    if (p==q) 
        return 1;   
    else 
        return 0;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
} 
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
    
    Node *q=second;
    while(second)
    {
      second=second->next;
      delete q;
      q=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
void LinkedList::Display_s(){
    Node *p=second;
    s = second;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}

````

````c++
void LinkedList::Display_t(){
    Node *p=result_t;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::LengthCircular(){
    Node *p=first;
    int len=0;
    do{
        len++;
        p=p->next;
    } while(p!=first);
        
    return len;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert_s(int index, int x){
    Node *t, *p=second;
    s = second;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = second;
        second = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}  
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
