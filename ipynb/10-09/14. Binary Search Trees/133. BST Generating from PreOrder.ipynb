Summary
1. Initial insert
    + Create new node (Root)
    + Push into Stack
2. Left insert:
    + Create new node (lchild)
    + Push into Stack
3. Right Insert, Between range
    + Seek Stack to check if in range (or Stacktop, as you do in one step)
    + Create new node (rchild)
    + Dont Push into Stack
4. Right Insert, Not Between range
    + Seek Stack to check if in range (or use Stacktop)
    + Pop off Stack
    + Do nothing

## Let code

___


We have:

$$ Pre = 30,20,10,15,25,40,50,45\;\;\;\;$$
$$ i= 0\;\;,1\;,2\;,3\;,4\;,5\;,6\;,7\;$$

$$n = 8$$

```
void CreatePre(int pre[], int n){
    Stack s,
    Node *  *p;
    int i = 0;
    
    root = new Node;
    root->data= pre[i++];
    root->lchild=root->rchild=NULL;
    p = root;
    
    while (i< n){
        //1.LHS
        if (pre[i] < p-><data){
            t=new Node;
            t->data = pre[i++]; //and move to next element
            t->lchild=t->rchild=NULL;
            p->lchild = t;
            Push(p);
            p=t;    
        }
        else if (pre[i] > p-><data && pre[i] < StackTop(p)->data){  //Within range: True
            t=new Node;
            t->data = pre[i++]; //and move to next element
            t->lchild=t->rchild=NULL;
            p->rchild = t; 
            p=t;    
        }
        else
            p=Pop();
        
    }
}

```
Time complexity : $$O(n)$$
