```c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
```

```c++
class Array{
private:
    int *A;
    int size;
    int length;
public:
    Array(){
        size=10;
        A=new int[10];
        length = 0;
        
    }
    Array(int sz){
        size=sz;
        A=new int[sz];
        length = 0;
        
    }
    ~Array(){
        delete []A;
    }
    void Display();
    void Insert (int index, int x);
    int BinarySearch(int key);
    int BinarySearchRecursive(int l, int h, int key);
}
```

```c++
int Array::BinarySearch(int key){
  int mid;  
  int l = 0;
  int h = length-1;  
  while (l<=h){  
    mid = (l+h)/2;
    if(key==A[mid])
        return mid;
    else if (key<A[mid])
        h = mid-1;
    else
        l = mid+1;
  }
  return -1;  
}
```

## This is an example of Tail Recursion

As we know where we have tail recursion, its better to choose the loop.

```c++
int Array::BinarySearchRecursive(int l, int h, int key){
  int mid; 
    
  if (l<=h){  
    mid = (l+h)/2;
    if(key==A[mid])
        return mid;
    else if (key<A[mid])
        return BinarySearchRecursive(l,mid-1,key);
    else
        return BinarySearchRecursive(mid+1,h,key);
  }
  return -1;  
}
```

```c++
void Array::Insert(int index, int x){
   if  (index>=0 && index<=length){
     for(int i=length-1;i>=index;i--)
         A[i+1]=A[i];
     A[index]=x;
     length++;  
   }
}
```

```c++
void Array::Display(){
    for(int i=0; i<length;i++)
      cout<<A[i]<<" ";      
    cout<<endl;      
}
```

```c++
Array arr(14);
arr.Insert(0,4);
arr.Insert(1,8);
arr.Insert(2,10);
arr.Insert(3,15);
arr.Insert(4,18);
arr.Insert(5,21);
arr.Insert(6,24);
arr.Insert(7,27);
arr.Insert(8,29);
arr.Insert(9,33);
arr.Insert(10,34);
arr.Insert(11,37);
arr.Insert(12,39);
arr.Insert(13,41);
arr.Insert(14,43);
```
