## Insertion Sort

___

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
8\T&5\T&7\T&3\T&2 \\\hline 
\end{array}
$$


Procedure

1. We assume the first element is sorted. 
2. And the rest, four elements is not sorted. 
3. In short, how do we do this, take next and insert into "sorted list" (which is only 8) at the moment, and repeat the process.
4. Lets start:

$$$$

$$\text{First Pass - Insert 5}$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
8\T&-\T&7\T&3\T&2 \\\hline 
\end{array}
$$

1. Compare 5 with 8, 5 < 8Insert 5
2. Shift 8 to 5's position.
3. Place 5 in 8's position.
4. New sorted list: 5,8

$$\text{First Pass Result}$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
5\T&8\T&7\T&3\T&2 \\\hline 
\end{array}
$$


*Compare*
1. One Comparison
2. One swap (or shifting)

$$$$

$$\text{Second Pass - Insert 7}$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
5\T&8\T&-\T&3\T&2 \\\hline 
\end{array}
$$

1. Compare 7 with 8, 7 < 8
2. Shift 8 to 7's position.
3. Compare 7 with 5, 5 > 7
4. Do not shift
5. New sorted list: 5,7,8

*Compare*
1. Two Comparison.
2. One swap (or shifting), but maximum Two Swaps


$$\text{Second Pass Result}$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
5\T&7\T&8\T&3\T&2 \\\hline 
\end{array}
$$



$$$$

$$\text{Third Pass - Insert 3}$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
5\T&7\T&8\T&-\T&2 \\\hline 
\end{array}
$$

1. Compare 3 with 8, 3 < 8
2. Shift 8 to 3's position.
3. Compare 3 with 7, 3 < 7
4. Shift 7 to 8's position.
5. Compare 3 with 5, 3 < 5
6. Shift 5 to 7's position.
7. Place 3 in 5's position.
4. New sorted list: 3,5,7,8

*Compare*
1. Three Comparisons
2. Three swaps (or shifting)


$$\text{Third Pass Result}$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
3\T&5\T&7\T&8\T&2 \\\hline 
\end{array}
$$


$$$$

$$\text{Fourth Pass - Insert 2}$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
3\T&5\T&7\T&8\T&- \\\hline 
\end{array}
$$

1. Compare 2 with 8, 2 < 8
2. Shift 8 to 2's position.
3. Compare 2 with 7, 2 < 7
4. Shift 7 to 8's position.
5. Compare 2 with 5, 2 < 5
6. Shift 5 to 7's position.
7. Compare 2 with 3, 2 < 3
8. Shift 3 to 5's position.
9. Place 2 in 3's position.
4. New sorted list: 2,3,5,7,8

*Compare*
1. Four Comparisons
2. Four swaps (or shifting)


$$\text{Fourth Pass Result}$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
2\T&3\T&5\T&7\T&8 \\\hline 
\end{array}
$$

Lets Analyse:

1. Number of Passes: 4 for 5 elements. So n -1 passes.
2. Number of Comparisons:

$$1+2+3+4+..+n-1 = \frac{n(n-1)}{2}=O(n^2)$$

3. Number of Swaps:

$$1+2+3+4+..+n-1 = \frac{n(n-1)}{2}=O(n^2)$$
4. After the first pass, we got 5 and 5. Is 5 the smallest element. No so the first pass was not very useful. som intermediate results will not give you any useuful results.
5. Compared to Bubble sort, the first pass, we got the lasrgest element, whereas herem there are no useful intermediate information.
6. With array, we saw we need to shift elements. Whereas with LL, this was not the case.So we say that LL is more suitable for Insertion Sort. 
7. We can go further and say that Insertion Sort was designed to be used primarily with LL.


 

## Lets code:

___

1. In our example the number of elements was 5.
2. So n = 5, but the index starts at 0.
3. The first for loop  will be for the passes.
4. And how many passes? n - 1 passes.
5. But 8 (index 0) is already there, so we need to insert from index 1 (that is 5)
6. So the loop should start from 1 onwards, and should go till 4 (which is n).

```
for (int i =1; i < n; i++){
...
}
```
6. As this for loop is for the repeating passes.
7. But now for each pass, what do we do?
8. We have the following scenario, as we want to swap value 7 :

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
5\T&8\T&-\T&3\T&2 \\\hline 
\T&\T&_{i}\T&\T& \\\hline 
\end{array}
$$

9. That means j will be the position before i:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
5\T&8\T&-\T&3\T&2 \\\hline 
\T&_{j}\T&_{i}\T&\T& \\\hline 
\end{array}
$$

10. Also need to asign the element we want to swap to x:
11. So we have the following:

```
x = A[i];
j = i -1;

```
12. So we compare values, if less, then j need to go to 0, so ```j--```, and Shift elements
13. When j of A, becomes greater we stop. 
14. And whereever j is pointing, copy element to next (j+1) location

```
while ((j > -1 && x < A[j] ){

    A[j+1] = A[j];
    j--;
```
(j must also be greater than -1 in loop condition)

15. When x becomes greater than A of j, thats when we stop. But element we need to place in j+1, that empty space. Remember j is where we have greater value, that is still occupied. But space next to it, that is j+1, is blank.

```
A[j+1] = x;

```
```
void InsertionSort(int A[], int n){
  int x;
   

  for (int i=1; i < n; i++){
     x = A[i];
     j = i -1;
     
     while (j > -1 && x < A[j] ){
      A[j+1] = A[j];
      j--;
     }
     
     A[j+1] = x;
  } 
}
```

## Adaptive

___

We know its start inserting from index 1, and insert to LHS.

Lest look at the following (sorted) array:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
2\T&5\T&8\T&10\T&12 \\\hline 
\end{array}
$$

1. We compare 5 with 2. But we never enter while loop, as 5 greater than 2.
2. We compare 8 with 5. But we never enter while loop, as 8 greater than 5.
3. We compare 10 with 8. But we never enter while loop, as 10 greater than 8.
4. We compare 12 with 10. But we never enter while loop, as 12 greater than 10.


Number of comparisons

$$n-1=O(n)$$


Number of swap (zero)


$$0 = O(1)$$

That means if the list is already sorted it takes minimum time. That means insertion sort is adaptive.

And we did not use any flag like Bubble sort, so by default its adaptive.




## Stable

___

Thats means elements keep their orginal position:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
2\T&5\T&8\T&10\T&12\T&5 \\\hline 
\T&\T&_{i}\T&\T& \\\hline 
\end{array}
$$


1. Last 5 will be shifted to front.
2. When it gets to the 5 at index 2, That 5 will keep its place. So the element is stable.

Yes, Insertion Sort is Stable.




````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std; 
````

````c++
void InsertionSort(int A[], int n){
  int x, j;
  for (int i=1; i < n; i++){
     x = A[i];
     j = i-1;
     
     while (j > -1 && x < A[j] ){
       cout<<"Sorted at "<<i<<" Pass."<<endl;   
         
       A[j+1] = A[j];
       j--;
     }
     A[j+1] = x;
  } 
}
````

````c++
int A[] = {8,5,7,3,2};
int n= 5;
    
InsertionSort(A,n);

for (int i = 0; i<5;i++){
    cout<<A[i]<<" ";
}
cout<<endl;
````
