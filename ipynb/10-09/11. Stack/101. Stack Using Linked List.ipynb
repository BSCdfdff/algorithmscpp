````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class Stack{
private:
  Node *top;
    
public:
    Stack(){
        top = NULL;
    }
    ~Stack(); 
    void Display();
    int StackTop();
    int IsFull();
    int IsEmpty();
    void Push(int x);
    int Pop();
    int Peek(int pos);
   
}
````

````c++
Stack::~Stack(){
    Node *p=top;
    while(top)
    {
      top=top->next;
      delete p;
      p=top;
    }
}
````

````c++
int Stack::StackTop (){
   if (top)
      return top->data;
   return -1; 
}
````

````c++
int Stack::IsFull (){
    Node *t =  new Node;
    int x = t?0: 1;
    delete t;
    return x;
}
````

````c++
int Stack::IsEmpty (){
   return top?0:1;
}
````

````c++
void Stack::Push (int x){
  Node *t = new Node;
  if (t==NULL)
    cout<<"Stack overflow"<<endl;
  else
    t->data=x;
    t->next = top;
    top = t;
}
````

````c++
int Stack::Pop (){
  Node *p;
  int x= -1;

  if (top==NULL)
     cout<<"Stack empty"<<endl;
  else
  {
    p=top;
    top = top->next;
    x=p->data; 
    delete p;
  }
  
  return x;
}
````

````c++
int Stack::Peek (int pos){
  int x = -1;
  Node *p = top;
  for (int i = 0; p !=NULL && i < pos -1; i++)
    p=p->next;
 
  if (p)
   return p->data;
  else
   return -1;
}
````

````c++
void Stack::Display (){
  Node *p = top;
  for (int i = 0; p !=NULL; i++){
     cout<<p->data<<" ";
     p=p->next;
  }    
  cout<<endl;
}
````

````c++
Stack s;
s.Push(10);
s.Push(20);
s.Push(30);
s.Push(40);
s.Push(50);
s.Display();
````
