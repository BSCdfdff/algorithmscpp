The logic is very simple, and we will be doing only the first three (3) rows/cols combinatons:


${row A, i}$ = reads as = Row Array $A$, which is represented by $i$.

Confusing as we use $i$  and $j$  as the counter, but is is also used as to represent the row $(i)$ and col$(j)$ of the array.

Where $i$ and $j$ outside {}, it refers to counter, and will be used as the index to array $A$ and $B$

**Dealing with rows:**

$$
\text{If } i \text{[row }A, i\text{]} \lt j \text{ [row } B, i\text{]}, \text { then copy } i \text { [row } A, i \text {]} \text { to } C$$


$$
\text{If } i \text{[row }A, i\text{]} \gt j \text{ [row } B, i\text{]}, \text { then copy } j \text { [row } B, i \text {]} \text { to } C$$



**Dealing with columns:**

$$
\text{If } i \text{[col }A, j\text{]} \lt j \text{ [col } B, j\text{]}, \text { then copy } i \text { [col } A, j \text {]} \text { to } C$$


$$
\text{If } i \text{[col }A, j\text{]} \gt j \text{ [col } B, j\text{]}, \text { then copy } j \text { [col } B, j \text {]} \text { to } C$$


**where rows/cols are EQUAL:**

(this will be the ```else``` part)

We need to add the values, in value position:

$$
\text{If }
i \text{[row }A, i\text{]} = j \text{ [row } B, i\text{]} \text{ && }  
i \text{[col }A, j\text{]} = j \text{ [col } B, j\text{]}, \text { then add } i + j \text { val to } C$$

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Element{
public:
    int i;
    int j;
    int x;
}
````

````c++
class Sparse{
private:
    int m;
    int n;
    int num;
    Element *ele;
public:
    Sparse (int i, int j, int no){
        m = i;
        n = j;
        num=no;
        ele = new Element[num];
    }
    ~Sparse(){
        delete []ele;
    }
    //Addition
    Sparse ADDITION_OPERATOR(Sparse &ss);
    //Read
    friend istream & EXTRACTION_OPERATOR(istream &is, Sparse &s);
    //Display
    friend ostream & INSERTION_OPERATOR(ostream &os, Sparse &s);
 
}
````

````c++
Sparse Sparse::ADDITION_OPERATOR(Sparse &s){
    
    int i, j, k;
//Need to return null    
//     if (m!=s.m || n!=s.n)
//         return NULL;
    
    //create C = Store the result
    //and which we will return
    Sparse *sum = new Sparse(m,n,num + s.num);
    
    i=j=k=0;
    
    
    while(i<num && j < s.num){
        if (ele[i].i < s.ele[j].i)
            sum->ele[k++]=ele[i++];
        else if (ele[i].i > s.ele[j].i)
            sum->ele[k++]=s.ele[j++];
        else
        {
            if (ele[i].j < s.ele[j].j)
               sum->ele[k++]=ele[j++];
            else if (ele[i].j > s.ele[j].j)
               sum->ele[k++]=s.ele[j++];
            else{
              //this will copy i, i and x  
              sum->ele[k]=ele[i]; 
              //but then update x 
              sum->ele[k++].x=ele[i++].x + s.ele[j++].x;
            }    
        }
    }
    //Remaining elements.
    for (;i<num;i++ )
        sum->ele[k++]=ele[i];
    for (;j<s.num;j++ )
        sum->ele[k++]=s.ele[j];
    
    sum->num=k;
    return *sum;
}
````

````c++
istream & EXTRACTION_OPERATOR(istream &is, Sparse &s){
    cout<<"Enter non-zero elements"<<endl;
    for (int i = 0; i <s.num;i++)
         is>>s.ele[i].i>>s.ele[i].j>>s.ele[i].x;
    return is;
}
````

````c++
ostream & INSERTION_OPERATOR(ostream &os, Sparse &s){
    int k =0;
    for(int i = 0; i < s.m; i++){
        for (int j = 0; j < s.n; j++){
            if (s.ele[k].i==i && s.ele[k].j==j)
                os<<s.ele[k++].x<<" ";
            else
                os<<"0 ";
        }
        os<<endl;
    }
    return os;
}
````
