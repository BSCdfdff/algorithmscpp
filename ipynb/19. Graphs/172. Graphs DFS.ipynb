In Summary DFS: 

1. Can start from any vertex you like
2. Then explore that vertex
3. Once you get a new vertex, suspend the current once exploration (in pinr 2)
4. And start exploring that new vertex.


And the above tree is called DFS Spanning tree.


And the dotted edges is called back edges. That is connecting back to already connected edges


But we said DFS is just like PreOrder of a tree.

Thats is if we draw PreOrder (with your finger from Spanning tree). we get DFS : $\text{DFS:} 1,2,3,5,7,6,4$



## Analysis

___

We visiting all the vertices once. How many vertices are there? n, So time is order on n

$$O(n)$$

But this is analytical time. When we write a program it may differ, as it depends on the data structure we may be using


1. Adjacency Matrix will be $O(n^2)$
2. Adjacency List it will be $O(n)$
