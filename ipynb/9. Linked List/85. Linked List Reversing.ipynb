## Reversing a LL

___


We have list:

$$3->7->9->15->20$$

(For saving space we'll ignore 3 and 7)

```REFERENCE
We have the following LL :

2->4->6->8

200-> 210 -> 250 -> 300
```

There are two methods:

1. Reversing or interchanging elements (data or contents of nodes, addresses stays the same)(e.g. first, last and then move inwards)
2. Reversing links, thats is changing the addresses of nodes.

 



## Reversing or interchanging elements 

___


1. Take an array of the same size. We have pointer ```p```, set equal to $\text{first}$. And we have counter ```i``` as counter for array.
2. Copy elements in array. Using pointer ```p```, and counter ```i``` for array. Copy first element ```2```, then move ```p``` and ```i``` to next element. 
3. Reverse copy from array to linked list. Bring ```p``` to first node after copy. Keep ```i``` at last element in array. Copy last element of array ```(i)``` into ```p```, first node, $\text{decrement}$ ```i```, and move ```p``` to next node. And so we go along. 
4. Now ```8``` will be the $\text{first}$ node
5. Remember the links or addresses stay the same

This method needs an extra array equal to the size of the LL.

## Reversing Links

___

We will reverse the links of

$$2->4->6->8$$


What are sliding pointers? It is similar to tailing pointers, where the pointers slide or move in unison with each other.

Reversing Links method:

1. Here we will use $3$ pointers: ```r```, ```q``` and ```p```. ```p``` will be set to first, and ```r``` and ```q``` set to NULL. Sequence of movements is: We in loop with ```p```, moving ```p``` to next node, ```q``` is set to ```p's``` previous position, then ```r``` is set to ```q's``` previous position. (Initially p is first)
2. Movements can be achieved as setting: $r = q, q = p, p=p->next$).
3. The above are the steps for sliding pointers.
4. Now for kicker: After the above steps are done, we point ```q's``` next to ```r```, that is we pointing ```q``` to previous node ```r```
5. Then again we do  $r = q, q = p, p=p->next$, sliding then again, and we point ```q``` again to previous node ```r```
6. As so we continue until ```p``` is null.

Above needs clarity. Main idea is that the last node, links must point to previous nodes. And last node becomes 1. You read it the largest to smallest values.


With LL, the preferred method is to reverse links, rather than reversing elements. Why?

This is the method we prefer. We dont like to move data around, as per the first method. As we may be copying large  of data, whereas we just simply slide pointers which is of a smaller size.

Again, Now we want 300-> 250->201->200. So the data remains where it is in the list, only the addresses are reversed. Thats means after reversing, where 200 was the first node, now 300 will become first node.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void ReverseLL_Elements();
    void ReverseLL_Links();
}
````

````c++
void LinkedList::ReverseLL_Links(){
    Node *r,*q, *p=first;
    r=NULL;
    q=NULL;
    while (p){
        r = q;
        q = p;
        p=p->next;
        q->next=r;
    }
    first=q;   
}
````

````c++
void LinkedList::ReverseLL_Elements(){
    Node *p=first;
    int *A = new int[Length()];
    int i = 0;
    while (p){
        A[i]=p->data;
        p=p->next;
        i++;
        
    }
    p=first;
    i--;
    cout<<"***NOW WE REVERSE***"<<endl;
    for(; i>=0; i--)
    {
       p->data =  A[i];
       p=p->next; 
    }
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
