## What is meant by Insertion (not insertion sort)

___

What does the word Insertion refer to.

We have the following sorted array:



$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
2\T&6\T&10\T&15\T&20\T&25\T&30 \\\hline 
\end{array}
$$

And we want to insert the element 12. We search the array, find the position where to insert, and shift all element to right.

So Insertion sort, here refer to inserting the value 12:

1. Into a sorted array
2. In a sorted position

That is what insertion refer to in Insertion Sort.

But instead of finding 12's position, and shifting the elements afterward, we have to first find out where element must come or its position in array, we can immediately shift the elements. Let revisit...
