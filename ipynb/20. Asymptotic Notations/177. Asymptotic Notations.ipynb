## Big O, Theta and Omega

____


If we have the following time functions

$f(n) = n$, we will have Order:

$$O(n)$$

___

Have the following time finction

$f(n) = n^2$, we will have Order:

$$O(n^2)$$

___

Have the following time finction


$$f(n) = \displaystyle \sum_{i=1}^{n}i=1+2+3+ ...\frac{n(n+1)}{2}=O(n^2)$$

___


Have the following time finction


$$f(n) = \displaystyle \sum_{i=1}^{n}\times {2^i}= \text{ cannot be solved, we can get approximate value}$$

So now we cannot give a definite answer for the time complexity.

Is the approxation a lower value or higher value?

Is we take the lower value we can make it Omega:

$$\Omega(..)$$


Is we take the higher/Upper value we can make it Big O:

$$O(..)$$


Is we take  or have the exact value, we make it Theta :

$$\theta(..)$$



$$\text{Lower Bound} = \Omega \text{ (Omega) }$$

$$\text{Upper Bound} = O \text{ (Big O) }$$

$$\text{Tight Bound} = \theta \text{ (Theta) }$$


So we can say the upper bound of time function is


$$f(n) = \displaystyle \sum_{i=1}^{n}\times {2^i}=  \text{ Big O}= O (n2^n)$$



But most time people use Big O to mean at most, as most of the time we prefer upper bound.

As why did we ignore this notation before, as we were always getting exact formula, or tight bound. That we were always getting theta, so there was no need for notation.

So when we mentioned order of in all the examples we were refering to theta.

But here is the catch, even for exaxt outcomes people still use Big formula, as they mean from EQUAL to UPPER bound, so they cover exact and uppper bound situation by using BIG O
