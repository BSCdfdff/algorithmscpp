## Recursive Count Function (Head)

We can write it as follows:

```
int LLCount(){
    if (p==0)
        return 0;
    else
        return Count(p->next)+1;

```
This is similar to the previous HEAD recursion, and can further be written, as we did before, breaking it up into further steps:

1. Count (p->next)
2. +1

So the function $\text{(step 1)}$ ```Count```, will be called ```n+1```  times in our case, until the base condition is satisfied, only then will  $\text{(step 1)}$ that is ```+1```, execute. And ```+1``` will also only execute on during back propagation

# Recursive Count Function (Tail)

Here we have the TAIL version, and it works the same, just the order is reversed.

We can write it as follows:

```
int LLCount(){
    if (p==0)
        return 0;
    else
        return 1 + Count(p->next);
        
```

So will ```1``` be added like the other TAIL versions of recursion, at the beginning? No it will NOT, as we still dependent on the result of ```LLCount```.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void DisplayRecurTail(Node *p);
    void DisplayRecurHead(Node *p);
    int CountRecurHead(Node *p);
    int CountRecurTail(Node *p);
    
}
````

````c++
int LinkedList::CountRecurTail(Node *p){
    static int cnt = 0;
    if (p==NULL)
        return 0;
    else{
        cnt++;
        cout<<"cnt: " <<cnt<<endl;
        cout<<"p->data: " <<p->data<<endl;
        return 1+CountRecurTail(p->next);
    }
    //cout<<"cnt: " <<cnt<<endl;
    
}
````

````c++
int LinkedList::CountRecurHead(Node *p){
    static int cnt = 0;
    if (p==NULL)
        return 0;
    else{
        cnt++;
        cout<<"cnt: " <<cnt<<endl;
        cout<<"p->data: " <<p->data<<endl;
        return CountRecurHead(p->next)+1;
    }
    //cout<<"cnt: " <<cnt<<endl;
    
}
````

````c++
void LinkedList::DisplayRecurHead(Node *p){
    static int cnt = 1;
    if (p!=NULL){
        cnt++;
        DisplayRecurHead(p->next);
        cout<<p->data<<endl;
    }
    cout<<"cnt: " <<cnt<<endl;
}
````

````c++
void LinkedList::DisplayRecurTail(Node *p){
    static int cnt = 1;
    if (p!=NULL){
        cnt++;
        cout<<p->data<<endl;
        DisplayRecurTail(p->next);
    }
    cout<<"cnt: " <<cnt<<endl;
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}   
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==1){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
            p=q;
            p=p->next;
            
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
