___

First initialize H with 0. Then Scan A, and then for each letter, set the corresponding index (letter - 97) in H equal to 1. (increment). Scan B, check value, if has number greater than 0, decrement. If all 0's then anagram,if -1, then NOT.

This approach will work for when there are duplicates as well.

If there are duplicates, then  Bitwise will not work.

```c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
```

```c++
char  A[]="decimal";
char  B[]="medical";
int H[26]={0};;
int i;
int k = 0;
int j =0;

for(int i = 0; A[i]!='\0';i++){
    k = A[i]-97;
    H[k]++;
   
    
}

i =0;

for(int i = 0; B[i]!='\0';i++){
    k = B[i]-97;
    H[k]--;
    if (H[k]<0){
        cout<<"Not Anagram"<<endl;
        break;
    }
    j++;
}

if (B[j]=='\0')
       cout<<"This is an Anagram"<<endl;   
```
