## Circular Linked List

___

8-> 3-> 9 -> 6-> 2

Where two points back to 8 (the first node), that is a circular LL. Or where a collection of nodes are circularly connected. So is there still a head? or beginning?

But what is benefit of circular LL? We can traverse it circulary, e.g Contact list, move from beginning to end.

Two methods for representing circular LL

1. Normal linear node, with last node pointing back to first node.
1. We have a head node that does not have any data, but just point to other nodes, and where the last node point back to node after the head node
2. 

How do we display a circular LL? Normal while loop will not work, as we need to see if the head/first was reached. So we first need to perform one search, then check for next nodes coming up if head was reached. Here we use a Do while loop.

But we first need to setup the circular loop:


```
Node *nad1, *nad2;

LinkedList links;
links.Insert(0,10);
links.Insert(1,20);
links.Insert(2,30);
links.Insert(3,40); 
links.Insert(4,50);
links.Display();


nad1=f; 
nad2=f->next->next->next->next;
nad2->next = nad1;
f = nad2; //make su
```

We set point nad1 equal to f(first). Then we move pointer nad2 to last node (50). Then we set last not to point back to first node (nad2->next = nad1).

Now we ahve circular ll, and we can do the display:

```
void LinkedList::DisplayCircularLL(){
    Node *p=first;
    do{
        cout<<p->data<<" ";
        p=p->next;
    } while (p!=first);

    cout<<endl;
}   
```
We read until the first node is reached again.



## Using Recursion to Display Circular LL

We need an additional flag for the function to call itself. Lets trace:

```
    static int cnt = 0;
    if (p!=first || cnt==0){
        cnt=1;
        cout<<p->data<<" ";
        DisplayCircularLLRecursion(p->next);
    }
    cout<<endl;
```    

First Time:

p!=first is false
cnt==0 is true, so on this basis it will be able to call itself

Second Time

p!=first is true, on this basis wit will be able to call itself.
cnt==0 is false (see above)


And based on p!=first condition, it will exit once we get back to the first node.


````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL,*s=NULL,*result_t=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display();
    void Display_s();
    void Display_t();
    void Insert(int index, int x);
    void Insert_s(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    int CheckLoopLL();
    void DisplayCircularLL();
    void DisplayCircularLLRecursion(Node *p);
    
}
````

````c++
void LinkedList::DisplayCircularLLRecursion(Node *p){
    static int cnt = 0;
    
    //it check either or first time around
    //1. p!=first wont let ti enter
    //2. but cnt==0) will let it enter
    //3. next try  cnt==1 wont let it enter 
    //4. but p!=first will let it enter.
    if (p!=first || cnt==0){
        cnt=1;
        cout<<p->data<<" ";
        DisplayCircularLLRecursion(p->next);
    }
    cout<<endl;
}
````

````c++
void LinkedList::DisplayCircularLL(){
    Node *p=first;
    
    do{
        cout<<p->data<<" ";
        p=p->next;
    } while (p!=first);

    cout<<endl;
    
}    
  
````

````c++
int LinkedList::CheckLoopLL(){
    Node *p, *q;
    p=q=first;
    do{
        p=p->next;
        q=q->next;
        //if q is not null, then assign q's next
        q=q?q->next:q;
       
    }while (p && q && p!=q);
    
    if (p==q) 
        return 1;   
    else 
        return 0;
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first; 
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
    
    Node *q=second;
    while(second)
    {
      second=second->next;
      delete q;
      q=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
void LinkedList::Display_s(){
    Node *p=second;
    s = second;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}

````

````c++
void LinkedList::Display_t(){
    Node *p=result_t;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    f = first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
   
}    
````

````c++
void LinkedList::Insert_s(int index, int x){
    Node *t, *p=second;
    s = second;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = second;
        second = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}  
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
