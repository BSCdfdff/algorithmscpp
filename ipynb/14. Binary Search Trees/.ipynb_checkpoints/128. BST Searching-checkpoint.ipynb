## Searching BST

___

```

                                          root
                                   +-----+-----+-----+
                                   | 210 |  30 | 230 |
                                   +-----+-----+-----+
                                           200
                                 /                    \
                                /                      \
                               /                        \
                              /                          \

                   +-----+-----+-----+                   +-----+-----+-----+
                   | 270 |  15 | 300 |                   | 310 | 50  | 320 |
                   +-----+-----+-----+                   +-----+-----+-----+
                      /    210 \                          /       230      \
                     /          \                        /                  \
                    /            \                      /                    \
                   /              \                    /                      \
   +-----+-----+-----+  +-----+-----+-----+      +-----+-----+-----+  +-----+-----+-----+
   |  \  |  10 |  \  |  |  \  |  25 |  \  |      |  \  |  40 |  \  |  |  \  |  60 |  \  |
   +-----+-----+-----+  +-----+-----+-----+      +-----+-----+-----+  +-----+-----+-----+
           270                  300                      310                  320


```

We want to find key=25 [FOUND]

Procedure:

1. Take pointer p on root
2. Check p's data (30) against key of 25
3. Is p's data less or greater than 25
4. If less, move pointer to left child of p
5. If greater, move pointer to right child of p
6. It less, so move p's pointer to left child
7. Check p's data (15) against key of 25
8. If less, move pointer to left child of p
9. If greater, move pointer to right child of p
10. Its greater, so move p's pointer to right child
11. Check p's data (25) against key of 25.
12. Found. stop.

## Recursive Search

____

```
Node* RSearch(Node *p, int key){
 if (p==NULL)
   return NULL;
 if (p->data == key)
   return p;
 else (p->data < key)
   return RSearch(p->lchild, key);
 else
   return RSearch(p->rchild, key);
}

```


This is Tail Recursion.


## Iterative Search

____


```
Node* ISearch(Node *p, int key){

 while (p){
     if (p->data == key)
       return p;
     else (key < p->data)
       p = p->lchild
     else
       p = p->rchild
   
   }
 return NULL  
}
```
With Tail Recursion, can without effort be converted to Iterative way, so no need for using a STACK. Thats fact.
