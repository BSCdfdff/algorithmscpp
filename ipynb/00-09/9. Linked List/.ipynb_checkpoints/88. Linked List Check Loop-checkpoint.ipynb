What is a loop? Where the last node is pointing to some node in the link list. So 9 will point to 4. So that is forming a  loop. And the other side, where it is not coming back to some node inside the LL, it is Linear.



Check for Linear:

1. Loop until we get the last node
2. And the last node is null

## Check for Loop

___


1. Loop the list, using ```p```
2. We use two pointer ```q and ```p```. (start off at same first node)
3. Both ```p``` and ```q``` start at same position. Where ```p``` use ```1 step``` and ```q``` uses ```2 steps.``` 
4. But before we let ```q``` takes its second step, make sure it does not become null.
5. If ```q's next``` is null, do not move it.Just return its existing value. And wait for ```p``` to catch up.
6. Continue increasing, ```p```, that is move to next node.
5. And the loop will exit of p becomes null or p does not eqial q.
6. The loop will exit if $p==q$

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL,*s=NULL,*result_t=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display();
    void Display_s();
    void Display_t();
    void Insert(int index, int x);
    void Insert_s(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    int CheckLoopLL();
}
````

````c++
int LinkedList::CheckLoopLL(){
    Node *p, *q;
    p=q=first;
    do{
        p=p->next;
        q=q->next;
        q=q?q->next:q;
    }while (p && q && p!=q);
    
    if (p==q) 
        return 1;   
    else 
        return 0;
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
    
    Node *q=second;
    while(second)
    {
      second=second->next;
      delete q;
      q=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
void LinkedList::Display_s(){
    Node *p=second;
    s = second;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}

````

````c++
void LinkedList::Display_t(){
    Node *p=result_t;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    f = first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
   
}    
````

````c++
void LinkedList::Insert_s(int index, int x){
    Node *t, *p=second;
    s = second;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = second;
        second = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}  
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
