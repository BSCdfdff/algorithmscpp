## Finding a Duplicates in a unsorted array using Hash Table

___

Lets call this A

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
8\T&3\T&6\T&4\T&6\T&5\T&6\T&8\T&2\T&7  \\\hline 
_0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8\T&_9 \\\hline 
\end{array}
$$


Lets call this H. 


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|}
\hline \require{cancel}  
0\T&0\T&\cancel{0}1\T&\cancel{0}1\T&\cancel{0}1\T&\cancel{0}1\T&\cancel{0}3\T&\cancel{0}1\T&\cancel{0}2            \\\hline 
_0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8 \\\hline 
\end{array}
$$


Scan $A$ and increment the count (change from $0$ to number of instances of that element in $H$)

Time complexity:

$$\text {O}(n)$$


```c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
```

```c++
class Array{
private:
    int *A;
    int size;
    int length;
    void Swap (int *x, int *y);
    int SumNatNos(int y);
    bool FindDiffenceSuccess (int x,int y, int z);
public:
    Array(){
        size=10;
        A=new int[10];
        length = 0;
        
    }
    Array(int sz){
        size=sz;
        A=new int[sz];
        length = 0;
        
    }
    ~Array(){
        delete []A;
    }
    void Display();
    void Insert (int index, int x);
    void FindDplicatesinUnsorted();
    void FindDplicatesUnSortedHashing(int h);
}
```

```c++
int Array::SumNatNos (int y){
    return  (y*(1+y))/2;    
}

```

```c++
bool Array::FindDiffenceSuccess (int x, int y, int z){
    int a=0;
    a= x - y;
    if (a==z){
     return true;    
    }
    return false; 
}
```

```c++
void Array::Swap (int *x, int *y){
    int temp;
    temp =*x;
    *x=*y;
    *y = temp;
}
```

```c++
void Array::FindDplicatesinUnsorted (){
    int count = 1;
 
    for(int i=0; i < length; i++ ){
       
        for (int j = i+1; j<length;j++){
            if (A[j]!= -1 ){
                if (A[i] ==A[j])
                {
                    count++;
                    A[j]=-1;
                }
            }
        }
        if (count > 1){
            cout<<"Count:"<<count<<endl;
             count=1;
        }   
    }
    
 
     
     
}
```

```c++
void Array::FindDplicatesUnSortedHashing (int h){
    int *H = new int[8];
    
    for (int j = 0; j<h; j++) 
       H[j]=0;
    
    //same H[A[i]]=1
    for(int i=0; i < length; i++ ){
        for(int j=0; j < length; j++ ){
            if (A[i]==j)    
                H[A[i]]++;
        }
    }
    
    for (int j = 0; j<h; j++) 
       cout<<H[j]<<" ";      
    cout<<endl;      

    for (int j = 0; j<h; j++) {
        if (H[j]>=2)
         cout<<j<<" is appearing "<<H[j]<< " times."<<endl;     
    }
    
      
     
}
```

```c++
void Array::Insert(int index, int x){
   if  (index>=0 && index<=length){
     for(int i=length-1;i>=index;i--)
         A[i+1]=A[i];
     A[index]=x;
     length++;  
   }
}
```

```c++
void Array::Display(){
    for(int i=0; i<length;i++)
      cout<<A[i]<<" ";      
    cout<<endl;      
}
```

```c++
Array arr(10);
arr.Insert(0,8);
arr.Insert(1,3);
arr.Insert(2,6);
arr.Insert(3,4); 
arr.Insert(4,6);
arr.Insert(5,5);
arr.Insert(6,6);
arr.Insert(7,8);
arr.Insert(8,2);
arr.Insert(9,7);
arr.FindDplicatesinUnsorted();


Array arrA(10);
arrA.Insert(0,8);
arrA.Insert(1,3);
arrA.Insert(2,6);
arrA.Insert(3,4); 
arrA.Insert(4,6);
arrA.Insert(5,5);
arrA.Insert(6,6);
arrA.Insert(7,8);
arrA.Insert(8,2);
arrA.Insert(9,7);
arrA.FindDplicatesUnSortedHashing(9);
```
