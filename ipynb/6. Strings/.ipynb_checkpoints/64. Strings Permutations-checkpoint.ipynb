## Find all Permutations

___

What is permutations? That is all arrangement of a given string.


Lets call this $S$:


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&\text {\\}0  \\\hline 
_0\T&_1\T&_2\T&_3 \\\hline 
\end{array}
$$


Possible arrangements:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&\text {\\}0  \\\hline 
A\T&C\T&B\T&\text {\\}0  \\\hline 
B\T&A\T&C\T&\text {\\}0  \\\hline 
B\T&C\T&A\T&\text {\\}0  \\\hline 
C\T&A\T&B\T&\text {\\}0  \\\hline 
C\T&B\T&A\T&\text {\\}0  \\\hline 
_0\T&_1\T&_2\T&_3 \\\hline 
\end{array}
$$

For 3 chars, there are 3! possibilities. So n! arrangements.

So we can get state space tree, by using back tracking going up the tree




We use bruteforce to get all permutations.

Since we backtracking, we need to use recursion. And backtracking is normally implemented using recursion.

Lets write a recursive function.

````c++
void perm (char S[], int k){

static int  A[3]={0};
static char Res[3];
int i;

    if (S[k] =='\0'){
        cout<<"Permutation: "<<Res<<endl;
        Res[k]='\0';
    }
    else
    {
        for(int i = 0; S[i]!='\0';i++){
            if (A[i]==0){
                Res[k]=S[i];
                A[i]=1;
                perm(S,k+1);
                A[i]=0;
            }
        }    
    }
}
````

````c++
void swap(char *x, char *y){
    char temp;
    temp = *x;
    *x = *y;
    *y = temp;
    
}
````

````c++
void perms (char S[], int l, int h){
    
int i = 0;
    if (l == h)
      cout<<"Permutation: " <<S<<endl;
    else{
        for(i = l; i <= h;i++){
            swap(&S[l], &S[i]);
            perms(S,l+1, h);
            swap(&S[l], &S[i]); 
        }    
    }
}
````

````c++
char  SS[]="ABC";
perms(SS,0,2);
````
