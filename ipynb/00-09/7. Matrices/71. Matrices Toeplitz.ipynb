## Toeplitz Matrix

___


Lets call this $M$:


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  \\ 
{_2}  \\
{_3}  \\
{_4}  \\  
{_5}  \\
\end{array}
\begin{bmatrix}
2  & 3 & 4 & 5 & 6  \\
7  & 2 & 3 & 4 & 5  \\
8  & 7 & 2 & 3 & 4  \\
9  & 8 & 7 & 2 & 3  \\
10 & 9 & 8 & 7 & 2 \\
\end{bmatrix}
$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
& {_1} & {_2} & {_3} & {_4}  & {_5}  
\end{array}
$$
$$$$


Here all elements in the diagonals are the same. 


$$
i = \text { index of row} \\
j = \text { index of column}
$$

Thats means: 

```M[5,5] == M[4,4] == M[3,3] == M[2,2]== M[1,1]``` $=2$ 

```M[5,4] == M[4,3] == M[3,2] == M[2,1]``` $=7$ 

```M[5,3] == M[4,2] == M[3,1]``` $=8$ 

```M[5,1] == M[4,1]``` $=9$ 
$$$$

```M[2,5] == M[1,4]``` $=5$ 

```M[3,5] == M[2,4] == M[1,3]``` $=4$ 

```M[4,5] == M[3,4] == M[2,3] == M[1,2]``` $=3$ 

```M[5,5] == M[4,4] == M[3,3] == M[2,2]== M[1,1]``` $=2$  (already done)


We have the following pattern:

$$ M[i,j] = M[i-1, j - 1] $$



## How many elements are sufficient to be stored?

We can store:

1. Top row of elements
2. Left column of elements (except the top row element as its common)


So we only store the first row and first column.


So how many elements do we need to store? First row is $n$, first column (except first element of first row) is n -1

Size array: $$ n + n -1$$

In our case: $$ 5 + 5 -1 = 9 \text { elements}$$

$$

Lets get the formula for retrieving the elements. If the elements are in the upper triangle, then it will be in the row segment of the array. If the elements are in the lower triangle, then it will be the the column segment of the array.



Lets build up a formula for accessing a specific element:


Lets look any any n case:

To get $\text{index}$ for **Upper Triangle**:


$$$$

$$\text{if } i\le {j} = j-i$$
Then, 

$$A[\text {index }] = j -i$$

$$$$


To get $\text{index}$ for **Lower Triangle**:

$$$$

$$\text{if } i\gt {j} = n+ i - j - 1$$

Then, 

$$A[\text {index }] =  n+ i - j - 1$$

$$$$
