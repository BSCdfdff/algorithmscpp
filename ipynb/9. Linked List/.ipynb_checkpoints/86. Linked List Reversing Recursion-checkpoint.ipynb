## Reversing a LL using Recursion

____

E.g if we have :
Addresses:

200-> 210 -> 250 -> 300

Values:
2->4->6->8

We know recursion has two phases:

1. Calling Phase (Down the tree)
2. Returning Phase (Back up the tree)


That is,  should we REVERSE the links of each node while CALLING (down the tree) or while RETURNING (back up the tree). While returning seems to be the better option. Where work needs to be done on the returning phase we should use HEAD Recursion, as the statements doing the work will following the function, thus, the RETURNING phase will do the work.



How will this work:

1. We need a pointer ```p```, and a tail pointer ```q```, that will be one step behind ```p```.
2. So when ```p``` returns, it can point on ```q```, allowing us to reverse the links. (```p->next=q```).Or the previous node set to ```q```. This works similar to the looping implementation of reversal (i.e. where ```q's next``` is set to ```r``` (i.e. ```q's``` previous node).
3. When we reach end ```(p==NULL)```, then first must be set to q (```first = q```). Again similar to loop implementation, just less code and logic required.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void ReverseLL_Elements();
    void ReverseLL_Links();
    void ReverseLL_LinksRecurs(Node *q, Node *p);
}
````

````c++
void LinkedList::ReverseLL_LinksRecurs(Node *q, Node *p){
    if (p){
        ReverseLL_LinksRecurs(p, p->next);
        p->next = q;
    }
    else
        first = q;
}
````

````c++
void LinkedList::ReverseLL_Links(){
    Node *r,*q, *p=first;
    r=NULL;
    q=NULL;
    while (p){
        r = q;
        q = p;
        p=p->next;
        q->next=r;
    }
    first=q;   
}
````

````c++
void LinkedList::ReverseLL_Elements(){
    Node *p=first;
    int *A = new int[Length()]; 
    int i = 0;
    while (p){
        A[i]=p->data;
        p=p->next;
        i++;
        
    }
    p=first;
    i--;
    
    for(; i>=0; i--)
    {
       p->data =  A[i];
       p=p->next; 
    }
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    f = first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````

````c++
LinkedList linky;
linky.Insert(0,2);
linky.Insert(1,4);
linky.Insert(2,6);
linky.Insert(3,8);
linky.Display();
cout<<"***NOW WE REVERSE***"<<endl;
linky.ReverseLL_LinksRecurs(NULL, f);
linky.Display();
````
