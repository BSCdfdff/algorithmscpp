 ```
                +-+    +-+             +-+   +-+
                 D      E               F     G
                +-+    +-+             +-+   +-+
 
 ```

We need to store it in an array, and we use the index (which we start from 1, not 0) to represent the relationship beteen the verices.

How do we fill the array:

1. Top to bottom
2. Left to right
3. Moving down to each level


$$ 
T=
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
A\T&B\T&C\T&D\T&E\T&F\T&G   \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7  \\\hline 
\end{array}
$$ 

Have the relationships been preserved?

$$
 \newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|}
\hline 
Element & Index & LChild & RChild   \\\hline 
A & 1 & 2 & 3   \\\hline 
B & 2 & 4 & 5   \\\hline 
C & 3 & 6 & 7   \\\hline 
\end{array}
$$  

From the above we can find a formula for:

1. Finding LChild
2. Finding RChild
3. Finding Parent


$$_\text{LCHILD} = 2i$$

$$_\text{RCHILD} = 2i+1$$

$$_\text{PARENT}= \frac{i}{2} \;\;_\text{(take floor)}$$

But if we fill values level by level from left to right, automatically the formula will be followed.


 

 ##  Linked Representation
 
 ___
 
 Its called only linked, not linked list, as a tree is non linear.
 
  
 ```
                         +------------+
                               A 
                         +------------+
                         /             \
                        /               \
                       /                 \
                      /                   \
                    +-+                   +-+
                     B                     C
                    +-+                   +-+
                    / \                   / \
                   /   \                 /   \
                  /     \               /     \
                +-+    +-+             +-+   +-+
                 D      E               F     G
                +-+    +-+             +-+   +-+
 
 ```
 
Here we will be using Nodes of Linked List. The Nodes strucure be represented as follows:

1. Left Child
2. Data
3. Right Child

This structure is the same as a Doubly Linked List:
$$$$

$$
\begin{array}{r}
\\ 
\text{Node}   \\ 
\\
\end{array}
$$
$$
\newcommand\T{\Rule{0pt}{2em}{1em}}
\begin{array}{|r|r|r|}
\hline 
_{LChild} \T& \text{Data} \T& _{RChild}   \\\hline 
\end{array}
$$  

And defined as follows:

```
class Node{
public:
    Node *lchild;
    int data;
    Node *rchild;
}*f=NULL;

```
We will be creating the following structure in HEAP:

```
                                      +---+-----+---+
                                      |   |  A  |   |
                                      +---+-----+---+
                                      /             \
                                     /               \
                                    /                 \
                                   /                   \
                     +---+-----+---+                   +---+-----+---+
                     |   |  B  |   |                   |   |  C  |   |
                     +---+-----+---+                   +---+-----+---+
                        / \                                     / \
                       /   \                                   /   \
                      /     \                                 /     \
       +---+-----+---+       +---+-----+---+   +---+-----+---+       +---+-----+---+
       | \ |  D  | \ |       | \ |  E  | \ |   | \ |  F  | \ |       | \ |  G  | \ |
       +---+-----+---+       +---+-----+---+   +---+-----+---+       +---+-----+---+

```

We make the following observations:

We have $7$ nodes. How many NULL pointers? $8$ null pointers.

That is for any shape BT, there will be:

$$_\text{NULL POINTERS} = n+1$$

And this is similar to number of external nodes (for strict BT):

$$e = i+1$$

**We will be using this (Linked) Representation for representing BT.**
