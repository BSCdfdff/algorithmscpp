## Searching in LL


$\text{Linear}$ search will be performed here, as $\text{Binary}$ search one need to get to middle of list, and that means we need to get there, and with LL this means we need to traverse to get to middle of list. This will not be possible.


Simple as traversing again until we find the key element. We search for key, and if found, return the node. If not found return NULL


We have the following simple function

```

Node* Search(Node *p, int key)
{
    while(p){
        if(key==p->data)
            return(p); 
        p = p->next;    
    }
   return NULL; 
}
```






## Recursive Search Function (Head)

```
int RecurSearch(Node *p){
    if (p==NULL)

    if (key==p->data)
        return p;
    else
      return SearchLLRecurHead(p->next, key);
}
```

This is similar to the previous HEAD recursion, and can further be written, as we did before, breaking it up into further steps:

1. ```RecurSearch (p->next, key)```
2. Condition to check ```p->data```, is equal to the ```key```


Once the key is found, then the function ends. 

The base condition will be ```p==NULL```



## Improved Linear Search


____
We can use the following methods:

1. Transposition
2. Move to HEAD

The above was covered in array topic. $\text{Transposition}$ is where we change the position of with the previous value. And $\text{Move to Head}$, is where the key element is moved to the beginning, so when you search for key the next time, it will be completed in only one comparison.


We wont attempt $\text{Transposition}$ as it involves movement of data. We prefer movement of Nodes(i.e. links).

Better approach is to move the found(12) node to head,make it first node ($\text{Move to HEAD}$). That is bring ```12``` to front.


Lets start: the key we searching for is ```12```. Once we found the key we want to make it the HEAD. Again we need some form of a pointer ```p``` node. In searching we used the point ```p```. Where the pointer ```p``` stopped (after traversing the list) on the key node.

So once we have Node with ```data = 12```. We need to do the following:

1. Look at the previous node, ```q```
2. And point this previous in our case 7, to point to where Node 12 is pointing to now.
3. So since we modifying the previous node, we need pointer, lets call it q pointer on previous node as well.


So p starts at the beginning, and q (this can also be called TAIL pointer) will be following p (one step behind), as it traverse the list always one behind.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void DisplayRecurTail(Node *p);
    void DisplayRecurHead(Node *p);
    int CountRecurHead(Node *p);
    int CountRecurTail(Node *p);
    int MaxLLRecurHead(Node *p);
    int MaxLLRecurHead2(Node *p);
    int MaxLL(Node *p);
    Node* SearchLL(Node *p, int key);
    Node* SearchLLRecurHead(Node *p, int key);
    Node* SearchLLImprovedHead(Node *p, int key);
}
````

````c++
Node *LinkedList::SearchLLImprovedHead(Node *p, int key){
    Node *q=NULL;
    while(p){
        if(p->data==key){
            q->next = p->next;
            p->next = first;
            first=p;
            return p;
        }
        q = p;
        p = p->next;    
    }
   return NULL; 
}
````

````c++
Node *LinkedList::SearchLL(Node *p, int key){
    while(p){
        if(p->data==key)
            return p;
        p = p->next;    
    }
   return NULL; 
}
````

````c++
Node *LinkedList::SearchLLRecurHead(Node *p,int key){
    if (p==NULL)
        return NULL;
    cout<<"p->data: " <<p->data<<endl;
    
    if (key==p->data)
        return p;
    else
      return SearchLLRecurHead(p->next, key);
    
    
}
````

````c++
int LinkedList::MaxLLRecurHead(Node *p){
    static int x = 0,MIN_INT;
    cout<<"MIN_INT: " <<MIN_INT<<endl;
    if (p==NULL)
        return MIN_INT;
    x = MaxLLRecurHead(p->next);
    cout<<"x: " <<x<<endl;
    cout<<"p->data: " <<p->data<<endl;
    return x > p->data?x:p->data;
}
````

````c++
int LinkedList::MaxLLRecurHead2(Node *p){
    static int x = 0,MIN_INT;
    cout<<"MIN_INT: " <<MIN_INT<<endl;
    if (p==NULL)
        return MIN_INT;
    x = MaxLLRecurHead(p->next);
    
    if (x > p->data)
        return x;
    else
        return p->data;
    
    cout<<"x: " <<x<<endl;
    cout<<"p->data: " <<p->data<<endl;
    
}
````

````c++
int LinkedList::CountRecurTail(Node *p){
    static int cnt = 0;
    if (p==NULL)
        return 0;
    else{
        cnt++;
        cout<<"cnt: " <<cnt<<endl;
        cout<<"p->data: " <<p->data<<endl;
        return 1+CountRecurTail(p->next);
    }
    //cout<<"cnt: " <<cnt<<endl;
}
````

````c++
int LinkedList::MaxLL(Node *p){
    int m = -32768, MIN_INT;
    while(p){
        if(p->data> m)
            m = p->data;
        p = p->next;    
    }
   return(m); 
}
````

````c++
int LinkedList::CountRecurHead(Node *p){
    static int cnt = 0;
    if (p==NULL)
        return 0;
    else{
        cnt++;
        cout<<"cnt: " <<cnt<<endl;
        cout<<"p->data: " <<p->data<<endl;
        return CountRecurHead(p->next)+1;
    }
    //cout<<"cnt: " <<cnt<<endl;
}
````

````c++
void LinkedList::DisplayRecurHead(Node *p){
    static int cnt = 1;
    if (p!=NULL){
        cnt++;
        DisplayRecurHead(p->next);
        cout<<p->data<<endl;
    }
    cout<<"cnt: " <<cnt<<endl;
}
````

````c++
void LinkedList::DisplayRecurTail(Node *p){
    static int cnt = 1;
    if (p!=NULL){
        cnt++;
        cout<<p->data<<endl;
        DisplayRecurTail(p->next);
    }
    cout<<"cnt: " <<cnt<<endl;
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}   
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==1){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
            p=q;
            p=p->next;
            
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
