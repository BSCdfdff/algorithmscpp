For 8 elements how many passes have we performed. We have perfomed 3 passes, to get the result.

If it is recursive, it will not have passes.

The merging is done in $log_2{n}$ times.

And its done n times.

Time complexity is:

$nlog_2{n}$



## Lets Code

____

1. Take A as parameter, and number of elements of 8.
2. First for loop is for the passes, which is log n passes, that is 3 for 8 elements.

```
for (p=2; p<=n;p=p*2){
```

That is p is 2. Next time p is 4, next time p is 8. Provided p is less than or equal to n, which is 8. That is same behaviour we see above:

+ Frist we got, 4 groups of 2 elements       [p=2]
+ Next we got 2 groups of 4 elements (2x2)   [p=4]
+ Next we got 1 groups of 8 elements (2x2x2) [p=8]
 
    
3. And for the individual list to be merged we have the following for loop:


```
for (i=0; i+p-1<n; i=i+p)

```
4. And then we call the merge we did in the previous lession, where we have low, mid and high:



. 
```
void IMergeSort(int A[], int n){

    int p, i, l, mid, h; 
    
    for (p=2; p<=n;p=p*2){
      for (i=0; i+p-1<n; i=i+p)
      {
        l=i;
        h=i+p-1;
        mid=(l+h)/2;
        
        Merge(A,l, mid, h);
      }  
    }
    
    if(p/2 < n)
        Merge(A,0,p/2,n-1);
}

```

````c++
void MergeSingleArray(int A[],int l, int mid, int h){

    int i, j, k;
    i=l;
    j=mid+1;
    k=l;
    
    int B[100];
    
    while (i <=mid && j <= h){
        if (A[i] < A[j])
            B[k++] = A[i++];
        else
            B[k++] = A[j++];
    }
    
    
    for (; i <= mid; i++)
    {
      B[k++] = A[i];
    }

    for (; j <= h; j++)
    {
     B[k++] = A[j];
    }
    
    
    //Copy B To A
    for (int i = l; i<=h;i++){
       A[i]=B[i];
    }
    
    
}
````

````c++
void IMergeSort(int A[], int n){

    int p, i, l, mid, h; 
    
    for (p=2; p<=n;p=p*2){
      for (i=0; i+p-1<n; i=i+p)
      {
        l=i;
        h=i+p-1;
        mid=(l+h)/2;
        
        MergeSingleArray(A,l, mid, h);
      }  
    }
    
    if(p/2 < n)
        MergeSingleArray(A,0,p/2-1,n-1);
}
````
