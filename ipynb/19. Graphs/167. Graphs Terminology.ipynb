## Terminology

____

It is defined as a collection of

1. Verices(V) (nodes)
2. Edges(E) (lines or links connecting the nodes)

That is :

$$G=(V,E)$$



## Directed Graph

___

Where the edges have an arrows that goings into some direction (incident from/indicent to):

Where the following is called:

1. Pararallel Edges, arrow go from Vertex 4 to 3, and another arraow going from vertex 3 to 4
2. Self Loop, arrow forming a loop on itself
3. Indegree, is number of edges coming into vertex
4. Outdegree, is number of edges going out from vertex

```

                                +---> +---+        +---+
                      Selfloop  |     | 1 | +----> | 2 |
                                +---+ +---+        +---+
                                        +            +
                                        |            |
                                        |            |
                                        v            v
                         Indegree=2   +---+        +---+
                         Outdegree=1  | 4 | +----> | 3 |
                                      +---+ <----+ +---+
```





## Simple Graph


## Non - Connected Graph

___

More than one piece (or components,sub-graph) and they are not connected


```


                    +---+        +---+      +---+
                    | 1 | +----+ | 2 |      | 5 |+---------+
                    +---+        +---+      +---+          |
                      +            +          +            +
                      |            |          |          +---+
                      |            |          |          | 9 |
                      +            +          +          +---+
                    +---+        +---+      +---+          +
                    | 4 | +----+ | 3 |      | 6 | +--------+
                    +---+        +---+      +---+


```

And we we do connect it, then we say the graph has connected components:

```

                    +---+        +---+      +---+
                    | 1 | +----+ | 2 |      | 5 |+---------+
                    +---+        +---+      +---+          |
                      +            +          +            +
                      |            |          |          +---+
                      |            |          |          | 9 |
                      +            +          +          +---+
                    +---+        +---+      +---+          +
                    | 4 | +----+ | 3 | +--+ | 6 | +--------+
                    +---+        +---+      +---+

```

And where the connected is made between the two graphs (that is node 3 and 6), we call those articulation points.


But if we also connect node 2 and 5, then node 3 nand 6 will no longer be articulation points. (since it no longer becomes a critical point)



## Strongly Connected

___

From every vertex, you can reach other vertices.(So they are directed). That means I can to to one vertex, from another vertex, there is always a path available.

If we cant reach one node, from another node, then it only becomes connected.


So to get from node 1 to p, we take the path 1 to 2, then 2 to 9.

So a path is:

1. Set of edges to get from one node to the desired node.
2. Connecting a PAIR of vertices.


So we refer to 

1. Shortest Path
2. Longest Path

```


                    +---+        +---+
                    | 1 | +----> | 2 |+---------+
                    +---+        +---+          |
                      ^            ^            v
                      |            |          +---+
                      |            |          | 9 |
                      +            +          +---+
                    +---+        +---+          +
                    | 4 | <----+ | 3 |<---------+
                    +---+        +---+

```



And we call path going to node 1 to 2 to 3 to 4 as a cycle.

That is cycle is 

1. Circular path, connecting vertices.
2. Starting from one vertex, and ending at the same vertex.



## Directed Acyclic Graph  (DAG)

___

A directed graph with no cycles. If you start at one vertex, there is no path or way to get back to same vertex.


```


                    +---+        +---+        +---+
                    | 1 | +----> | 2 | +----> | 5 |
                    +---+        +---+    +-> +---+
                      +            ^      +     ^
                      |            |      |     |
                      |            |      |     |
                      v            +      |     +
                    +---+        +---+ +--+   +---+
                    | 3 | +----> | 4 | +----> | 6 |
                    +---+        +---+        +---+

```


Since it does not form a cycle, you can represent a DAG linearly, as follows

```


               ++-----------------------+ +----------------------+
               |                        | |                      |
               +                        v +                      v
             +---+        +---+        +---+      +---+        +---+        +---+
             | 1 | +----> | 3 |        | 2 |      | 4 | +----> | 5 | +----> | 6 |
             +---+        +---+        +---+      +---+        +---+        +---+
                            +            ^          +                         ^
                            |            |          |                         |
                            +------------+          +-------------------------+


```

And here the endges are only going into a forward direction (They not coming (circling) back)

And if we can arrange the edges only going forward, we call it topological ordering. And this is only possible through DAG graphs
