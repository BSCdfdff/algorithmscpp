## Quick Sort

___

Mostly discussed and most famous technique. Sometimes its seen as diffifult, but it is very interesting. 

It revolves around the following basic idea. Lets explore:

If we have the following list, what can we see in single glance:

$$\fbox{10}, 30, 20,70, 40, 90, 80$$

We can see that 10 is already sorted. Just by looking at the list. That is, it is the smallest element.

From the following list we can see in single glance, that 90 is already sorted.

$$30, 40, 20,70, 50, 10, \fbox{90}$$


In single glance, which element is already sorted.already sorted? it will be 50.

$$40, 30, 20,\fbox{50}, 90, 70, 80$$


But on what bases is 50 sorted? The 10 and 90 was easy to see? But how did we determine that 50 was sorted.

We see:

+ All elements before 50, is smaller than 50
+ All elements after 50, is greater than 50

From the above, we can find a definition of quick sort: It works on the idea that a element is sorted if, all the elements  before that element is less and all elements after that element is more. Then that element is in sorted position.

##  Analysis

___

There is an example for this. But we skipping over it.


For a sorted list (in ascending order) the time complexity is :

$$O(n^2)$$

This is not good.

And when it is in descending order is also:

$$O(n^2)$$
 
This is the worst case time of quick sort.


But for the best case, we have 1 to 15 elements, so best case split is partioning splits the list in middle. 1- 7, and 9 to 15 elements, but best case again split 1 to 3, and 5 to 7, and  9 to 11, and 13 to 15.

Total number of Comparison (for 15 elements):

$$\frac{16}{2}=\frac{8}{2}=\frac{4}{2}=\frac{2}{2}=1 $$

This is a successive division by 2 is our favourite $log_2{n}$, and there n items. Best case, that is if partitioing is done in middle.

Total number of Comparison (for 15 elements):

Best case
$$nlog_2{n}$$

Worst case, that is partitioning on LHS or RHS:

$$O(n^2)$$

Average Case:
$$nlog_2{n}$$


But in order to put us in better position, we can also select any element as pivot, and making it the first element (to allow us to get a "middle partition"--flashlights!---take half of total sum of elements....)

Then time will be $nlog_2{n}$, best case.



## Compare Selection and Quick

___

Rememeber with Selection sort, we find a position, normally starting at first, then we move down the list, to find an element for that position. We look for an element to fill the correct position.


Whereas with Quick sort we find and element, that is pivot, and then find a position for that element. That is where that element should be in the list. We look for correct position for an element.



````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
void  Swap (int *a, int *b){
    int temp;
    temp =*a;
    *a=*b;
    *b=temp;
};
````

````c++
void SelectionSort(int H[], int n){
  int x, y, i, k, j;
  int swaps=0;  
  int passes=0; 
  int swap=0;  
  
  for (i = 0; i < n-1; i++){
    x = H[i];
    passes++;  
    j = i;  
    k=j;  
    while (j<n){
      y = H[j];      
      if (y < x){
        k =j;
        x = H[j];  
        swap = 1;  
      }    
      j++;
    }
    if  (swap==1) { 
        Swap (&H[i], &H[k]); 
        swaps++;
        swap=0;
    }
      
  }
  cout<<"Passes: "<<passes<<" "<<endl;
  cout<<"Swaps: "<<swaps<<" "<<endl;
}
````

````c++
int Partition(int A[], int l, int h){
  int pivot=A[l];
  int i=l, j = h;
  int swaps=0;  
  int passes=0; 
  int swap=0;    
  
  do{
  
    do {
      i++;
    }while(A[i]<=pivot);
    
    do {
      j--;
    }while(A[j]>pivot);
    
    if(i < j)
        Swap(&A[i], &A[j]);
      
  }while(i<j);
  
  Swap(A[l], A[j]);
  
  return j;

}
````

````c++
int A[] = {8,5,7,3,2,INT_MAX};
int n= 5;
    
SelectionSort(A,n);

for (int i = 0; i<n;i++){
    cout<<A[i]<<" ";
}
cout<<endl;
````
