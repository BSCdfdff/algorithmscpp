## Introduction

____

Logical data structure, based in LIFO principle. That is Last in First out. Last item place on stack, will be removed First.


We know a recursive function is a function that calls itself, and they internally uses a STACK. But we also know that we can rewrite a recursive function to an iterative function. And vice versa. 

But sometime when converting from Recursion to Iterative, we will need to use a STACK. Ok, we do know recursion uses a stack, and internally the compiler deals with this. We don't have to worry about it. But when going from Rec' to It', then we need to provide a STACK ADT for it.



## ADT Stack
 
___

Abstract Data Type of Stack will contain

1. Data Representation
2. And Operations on the STACK

The data representation will need to address:

1. Space for storing elements
2. Top pointer

We need space for storing the elements, e.g can be int or char based. We also need the pointer to top most element, and that is the most important element.

Operations of the STACK.

1. Push (x)
2. Pop()
3. Peek (index)
4. StackTop()
5. isEmpty()
6. isFull()
