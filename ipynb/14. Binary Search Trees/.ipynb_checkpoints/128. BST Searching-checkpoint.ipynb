## What is BST

___

For every node, starting at root, 

1. All the elements in its left subtree is less than that node (root)
2. Al the elements in its right subtree is greater than that node (root).

That is,

1. Left child is less than parent node
2. Right child is greater than parent node

```
                         +------------+
                              30
                         +------------+
                         /             \
                        /               \
                       /                 \
                      /                   \
                    +-+                   +-+
                    15                    50
                    +-+                   +-+
                    / \                   / \
                   /   \                 /   \
                  /     \               /     \
                +-+    +-+             +-+   +-+
                10     20              40    60
                +-+    +-+             +-+   +-+

```

This tree is useful for searching. But how? Lets search for key = 40.

1. Start at top is 40 less/greater than 30
2. Greater, go right to 50
3. Is it less/greater 50
4. Less, go left.

We observed, that only few elements where compared, than reading each element in linear way. So much faster.

This is very similar to binary search (of array), but this is on tree.



## Properties of BST

___


1. No duplicates allowed.
2. InOrder traversal produces a sorted list

$$In\;\;Order=10,15,20, 30, 40, 50, 60$$

3. No of BST for n nodes:

E.g if we have nodes 10,20,30. We know that Inorder will give us this list as it is sorted

So for these three nodes, 5 trees are possible,as per the 

$$
\text{Catalan Number}\;\;=\;\;
T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}
$$

These are all valid number of trees for 10, 20, 30 nodes for this Inorders

```

               +-+   +-+            +-+             +-+    +-+
               30    30             20              10     10
               +-+   +-+            +-+             +-+    +-+
               /     /              /  \               \      \
              /     /              /    \               \      \
             /     /              /      \               \      \
            /     /              /        \               \      \
          +-+   +-+            +-+        +-+            +-+    +-+
          20    10             10         30              30    20
          +-+   +-+            +-+        +-+            +-+    +-+
         /        \                                       /        \
        /          \                                     /          \
       /            \                                   /            \
     +-+           +-+                                 +-+           +-+
     10            20                                  20            30
     +-+           +-+                                 +-+           +-+


```
