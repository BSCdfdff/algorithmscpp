## Procedure
____


1. Find the element to delete as per search (done already)
2. Decide Inorder Predecessor/Successor.
3. Base on height, take that one:
    + if Left side has greater height, delete Predecessor
    + if Right side has greater height, delete Successor
4. Check Height of p's left child greater than p's right child
5. If true, find Inorder Predecessor of p's left child
    + Remember this is rightmost child of left subtree
6. Assign this result to pointer q.
7. Once we have q, assign p's data to q-data (replace it)
8. Now we need to delete pointer q,
9. But we must make sure it does not have children
10. So we recursively delete q (that is steps 1 tp 10). 
11. That is modify more than one node.

And where we have reached end,p is leaf node
12. Delete p
13. But only delete if p is root node.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>> 
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    Node *lchild;
    int data;
    Node *rchild;    
}*root=NULL;
````

````c++
void Insert(int key){
    Node *p=root;
    Node *q,*t;
    if (root==NULL){
       t = new Node;
       t->data=key;
       t->lchild =t->rchild = NULL;
       root = t; 
       return;
    }     
    
    while (p){
      q=p;
      if (key > p->data)
          p = p->rchild;
      else if (key < p->data)
          p = p->lchild; 
      else 
          return;
        
    }
    
   t = new Node;
   t->data= key;
   t->lchild = t->rchild = NULL;  
   if (key< q->data)
      q->lchild=t;
   else
      q->rchild=t;
        
}
````

````c++
void Inorder(Node *p){
   if(p)
   {
       Inorder(p->lchild);
       cout<<p->data<< " ";
       Inorder(p->rchild);
   } 
}
````

````c++
Node* ISearch(int key){
 Node *p=root;
 while (p){
     if (p->data == key)
       return p;
     else if (key < p->data)
       p = p->lchild;
     else
       p = p->rchild;

   }
 return NULL ;
}
````

````c++
int Height(Node *p){
    int x, y;
    
    if (p==NULL)
        return 0;
    x=Height(p->lchild);
    y=Height(p->rchild);
    return x>y?x+1:y+1;
    
}
    
````

````c++
Node* InPre(Node *p){
    while (p && p->rchild!=NULL){
        p=p->rchild;
    } 
    
    return p;
}
````

````c++
Node* InSucc(Node *p){
    while (p && p->lchild!=NULL){
        p=p->lchild;
    } 
    
    return p;
}
````

````c++
Node* RDelete(Node *p,int key){
  Node *q = NULL;
  
  if (p==NULL)
     return p;
  
  
  //if p is leaf node
  if (p->lchild ==NULL && p->rchild ==NULL) {
    if (p==root)
        root=NULL;
    cout<<"Node to be deleted:"<<p->data<<endl;  
    delete p; 
    return NULL;  
  } 
    
  //First get it
  if (key< p->data)
      p->lchild= RDelete(p->lchild, key);
  else if (key > p->data)
      p->rchild= RDelete(p->rchild, key);
  else{
      //Key found, now delete according to above
      //Replace with Inorder Predecessor/Successor
      //Cannot just decide, need to select as per following criteria.
      if (Height(p->lchild) > Height(p->rchild)){
          q = InPre(p->lchild);
          cout<<"Right Most Child of p("<<p->data<<"):"<<q->data<<endl;
          p->data=q->data;
          p->lchild=RDelete(p->lchild, q->data);
      }
      else{
          q = InSucc(p->rchild);
          cout<<"Left Most Child of p("<<p->data<<"):"<<q->data<<endl;
          p->data=q->data;
          p->rchild=RDelete(p->rchild, q->data);
      }
  }  
  return p;
}
````

1. Try to delete $50$

2. Cannot delete 50, as lchild/rchild is not NULL.    
3. Get right most(Inorder predecessor) of $50$, which is $40$
