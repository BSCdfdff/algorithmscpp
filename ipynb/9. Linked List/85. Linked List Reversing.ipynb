## Reversing a LL


There are two methods:

1. Reversing or interchanging elements (data or contents of nodes, addresses stays the same)(e.g. first, last and then move inwards)
2. Reversing links, thats is changing the addresses of nodes.

E.g if we have :

200-> 210 -> 250 -> 300

Now we want 300-> 250->201->200. So the data remains where it is in the list, only the addresses are reversed. Thats means after reversing, where 200 was the first node, now 300 will become first node.

With LL, the preferred method is to reverse links, rather than reversing elements. Why?



```c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
```

```c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
```

```c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    void InsertLast(int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void InsertSorted(int x);
    int CheckifSortedLL();
    void RemoveDups();
   
}
```

```c++
void LinkedList::RemoveDups(){
    Node *p=first;
    Node *q=first->next;
    
    while(q){
        if (p->data == q->data){
            p->next= q->next;
            delete q;
            q=p->next;    
        }
        else{
            p=q;
            q=q->next;
        }
    }

}
```

```c++
int LinkedList::CheckifSortedLL(){
    Node *p=first;
    int x = -32768;
    while(p){
        if (p->data < x)
            return 0;
        x=p->data;    
        p=p->next;
    }
    return 1;    
}
```

```c++
void LinkedList::InsertSorted(int x){
    Node *t=NULL,*q=NULL,*p=first;
    
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    //if it is first node
    if (p==NULL)
        first=t;
    else{    
        while(p && p->data < x){
                cout<<"p data: "<<p->data<<endl;
                q=p;
                p=p->next;
        }
        t->next=p;
        q->next=t;
    }
}
```

```c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
```

```c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
```

```c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
```

```c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
```

```c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
```
