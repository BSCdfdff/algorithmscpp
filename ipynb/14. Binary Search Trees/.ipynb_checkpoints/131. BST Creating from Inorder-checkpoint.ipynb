## Creating a BST from given set of keys

Time taken so create BST:

We inserting $n$ elements, and then finding it position in tree, which take $log_2{n}$ time, that is 

$$O(nlog_2{n}$$









````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    Node *lchild;
    int data;
    Node *rchild;    
}*root=NULL;
````

````c++
void Insert(int key){
    Node *p=root;
    Node *q,*t;
    if (root==NULL){
       t = new Node;
       t->data=key;
       t->lchild =t->rchild = NULL;
       root = t; 
       return;
    }     
    
    while (p){
      q=p;
      if (key > p->data)
          p = p->rchild;
      else if (key < p->data)
          p = p->lchild; 
      else 
          return;
        
    }
    
   t = new Node;
   t->data= key;
   t->lchild = t->rchild = NULL;  
   if (key< q->data)
      q->lchild=t;
   else
      q->rchild=t;
        
}
````

````c++
void Inorder(Node *p){
   if(p)
   {
       Inorder(p->lchild);
       cout<<p->data<< " ";
       Inorder(p->rchild);
   } 
}
````

````c++
Node* ISearch(int key){
 Node *p=root;
 while (p){
     if (p->data == key)
       return p;
     else if (key < p->data)
       p = p->lchild;
     else
       p = p->rchild;

   }
 return NULL ;
}
````

````c++
//9,15,5,20,16,8,12,3,6
Insert (9);
Insert (15);
Insert (5);
Insert (20);
Insert (16);
Insert (8);
Insert (12);
Insert (3);
Insert (6);
Inorder(root);

````

````c++
Node *temp;
temp = ISearch(20);
if(temp)
    cout<<"Element found: "<<temp->data<<endl;
else
    cout<<"Element NOT found: "<<endl;
````

````c++
Node *tempp;
tempp = ISearch(1);
if(tempp)
    cout<<"Element found: "<<tempp->data<<endl;
else
    cout<<"Element NOT found"<<endl;
````
