```c++
class Array{
private:
    int *A;
    int size;
    int length;
public:
    Array(){
        size=10;
        A=new int[10];
        length = 0;
        
    }
    Array(int sz){
        size=sz;
        A=new int[sz];
        length = 0;
        
    }
    ~Array(){
        delete []A;
    }
    void Display();
    void Insert (int index, int x);
    int LinearSearch(int key);
}
```

```c++
int Array::LinearSearch(int key){
    for(int i=0;i<length;i++){
        if (key==A[i]){
            return i; 
        }
    }    
    return -1;
}
```

```c++
void Array::Insert(int index, int x){
   if  (index>=0 && index<=length){
     for(int i=length-1;i>=index;i--)
         A[i+1]=A[i];
     A[index]=x;
     length++;  
   }
}
```

```c++
void Array::Display(){
    for(int i=0; i<length;i++)
      cout<<A[i]<<" ";      
    cout<<endl;      
}
```

```c++
Array arr(10);
arr.Insert(0,8);
arr.Insert(1,9);
arr.Insert(2,4);
arr.Insert(3,7);
arr.Insert(4,6);
arr.Insert(5,3);
arr.Insert(6,10);
arr.Insert(7,5);
arr.Insert(8,14);
arr.Insert(9,2);
```

```c++
cout<<arr.LinearSearch(5)<<endl;
```
