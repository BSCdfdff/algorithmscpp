The condition is normally for:

1. Continuation
2. Termination

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

## Global pointer *f

This allows us to access and pass the node as a parameter for Recursive function.

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void DisplayRecurTail(Node *p);
    void DisplayRecurHead(Node *p);
    
}
````

````c++
void LinkedList::DisplayRecurHead(Node *p){
    static int cnt = 1;
    if (p!=NULL){
        cnt++;
        DisplayRecurHead(p->next);
        cout<<p->data<<endl;
    }
    cout<<"cnt: " <<cnt<<endl;
}
````

````c++
void LinkedList::DisplayRecurTail(Node *p){
    static int cnt = 1;
    if (p!=NULL){
        cnt++;
        cout<<p->data<<endl;
        DisplayRecurTail(p->next);
    }
    cout<<"cnt: " <<cnt<<endl;
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}   
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==1){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
            p=q;
            p=p->next;
            
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
