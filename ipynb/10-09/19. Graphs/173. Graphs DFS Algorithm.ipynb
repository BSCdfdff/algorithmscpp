````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<< 
#define EXTRACTION_OPERATOR operator>> 
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
//Takes a matrix 

void DFS(int G[][7], int start, int n){
  int j;
  static int visited[7] = {0}; 
  
  if(visited[start]==0){
      cout<<start<< " ";
      visited[start]=1;
       for(j=1; j<=n; j++){
         if(G[start][j]==1 && visited[j]==0)
             DFS(G, j, n);
       } 
  } 
}    
      
````
