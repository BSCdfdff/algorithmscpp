## Column Major Mapping

___

This is for indexes that starts at 0
 
**Picture**
 
int A[3][4]

Previously with row major mapping we filled the rows linearly.

Now we fill the columns linearly.

E.g.


$$\text{Address of A[0][0]} = A_{00}$$
$$\text{Address of A[1][0]} = A_{10}$$
$$\text{Address of A[2][0]} = A_{20}$$

This will be identified as $col_{0}$


$$\text{Address of A[0][1]} = A_{00}$$
$$\text{Address of A[1][1]} = A_{11}$$
$$\text{Address of A[2][1]} = A_{22}$$


This will be identified as $col_{1}$


$$\text{Address of A[0][2]} = A_{20}$$
$$\text{Address of A[1][2]} = A_{21}$$
$$\text{Address of A[2][2]} = A_{22}$$


This will be identified as $col_{2}$


To get the address for A[1][2]:
$$ $$
$$\text{Address of A[1][2]} = 200 + [2\times3 + 1]\times 2$$
$$ $$

$$\text{Address of A[1][2]} = 214$$


To get the address for A[2][3]:
$$ $$

$$\text{Address of A[2][3]} = 200 + [2\times4 + 3]\times 2$$
$$ $$

$$\text{Address of A[2][3]} = 222$$



To get the address for A[1][3]:
$$ $$

$$\text{Address of A[1][3]} = 200 + [3\times3 + 1]\times 2$$
$$ $$

$$\text{Address of A[1][3]} = 220$$



$$ $$

$$ $$

Address of any location - COLUMN MAJOR FORMULA USED BY COMPILER:

$$\text{Address of A[i][j]} =  L_0 + [j\times m + i]\times w$$

$$\text{where } m  = \text{row index of array}$$
$$\text{where } n  = \text{column index of array}$$
$$(\text{int } A[m][n])$$

## Comparison Row Major and Column Mapping

___

$$ $$

$$ $$


$$(\text{int } A[3][4])$$

$$\text{where } m  = \text{row index of array}$$
$$\text{where } n  = \text{column index of array}$$
$$(\text{int } A[m][n])$$



$$ $$

$$\text{ROW MAJOR MAPPING FORMULA USED BY COMPILER}$$

$$\text{Address of A[i][j]} =  L_0 + [i\times n + j]\times w$$

$$ $$


$$\text{COLUMN MAJOR MAPPING FORMULA USED BY COMPILER}$$

$$\text{Address of A[i][j]} =  L_0 + [j\times m + i]\times w$$
 

With row major mapping we go from right to left (j, i).
With row major mapping we go from left to right (i, j);


Both use the same number of operation, so both are efficient. But C++ compiler uses row major mapping.
