## How to create a Node

___

It starts of with the first pointer, also called the HEAD Node. Where First is defined in the STACK, and the rest of the nodes sitting in the HEAP. This is the normal declaration of a HEAP pointer.

This can be defined as follows:

```
class Node{
public:
    int data;
    Node *next;
}
```

From the above we see, that class ```Node``` has a member (pointer) of type ```Node``` as well. Such a structure is called a self referential structure. And such a structure is used in LL.


And in order to create a Node, we need to use a pointer, as Nodes can ONLY be created in HEAP. We can define a Class ```LinkedList``` and define a pointer to that Node inside it.


```
class LinkedList{
private:
  Node *first;
```  

Simple example to create a Node:

## How to address NULLS or check for it.

___

How to make or set a Node pointet to NULL

```
Node *p=NULL

```

OR

```
Node *p=0

```

And to check for NULL? This is important. Since if the point is NOT null then it will be pointing to a Node. This is important to know, as we dont really have a handle to a node. Since the only way yo get to a node, is to traverse the list till the end. And we need to know when we have reached the end. So.


We can us the following approaches, they all mean the same thing, to check it the node is null or 0:


```
if (p==NULL)

```


```
if (p==0)

```

```
if (!p)

```

Thus is p HAS value, that is pointing to address, so any inverse of above


```
if (p)

```

Above is best approach.


The above works for p itself, to see if is is NULL.


Another important check is to see if p's next is NULL or pointing to some other node:

```
if (p->next==NULL)
```

That it is not pointing to anything, and we have reached the end.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    LinkedList(int A[], int n);
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    int Delete(int index);
    int Length();
    
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    first->next=NULL;
    last = first;
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}   
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==1){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
            p=q;
            p=p->next;
            
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
