## Drawbacks of BST

___


We have the following keys to create BST:

$$Keys:40,20,30,60,50,10,70$$

$$Keys:10,20,30,40,50,60,70$$

Same keys but ORDER is different.


Lets create and see.

First set keys : height = $$O(log_2{n})$$ . This is minimim

```

                      Node=7             Height
                        +-+            + <----+ 0
                        +40            |
                        /  \           |
                       /    \          |
                      /      \         |
                     /        \        |
                   +-+        +-+      |<-----+ 1
                   +20        +60      +
                   / \         / \     |
                  /   \       /   \    |
                 /     \     /     \   |
               +-+    +-+   +-+   +-+  v <----+ 2   Min
               +10    +30   +50   +70


                        h=2

                        Min

```

Second set Keys: height = $$O(n)$$ This is maximim

```

          Node=7
           +-+                                Height
           +10                              + <----+ 0
              \                             |
               \                            |
                \                           |
                 \                          |
                 +-+                        |
                 +20                        |<-----+ 1
                    \                       +
                     \                      |
                      \                     |
                     +-+                    |
                     +30                    v <----+ 2   Min
                         \                  +
                          \                 |
                           \                |
                          +-+               v <----+ 3
                          +40               +
                              \             |
                               \            |
                                \           |
                               +-+          v <----+ 4
                               +50          +
                                  \         |
                                   \        |
                                    \       |
                                   +-+      v <----+ 5
                                   +60      +
                                      \     |
                                       \    |
                                        \   |
                                       +-+  v <----+ 6  Max
                                       +50

                           h=6

                           Max
                                
```

So from the above we cannot control the height of BST. So it depends on how we insert it. And can we control the order of insertion? No we cannot. This is how e get the information from the user.

So we cannot control the order of insertion, or the height of BST.

Is there a method to control the height of BST?

AVL trees are height balanced BST's

How does AVL trees control their height.
