## Merging two LL

___

Merging is the process of combining two sorted lists into a single sorted list.

Here we dealing with sorted lists.


first (p)

200-> 210 -> 220 -> 230
2->8->10->15


second (q)

300-> 310 -> 320
4->7>12->14


With arrays we needed an additional array when we performed merging. With LL, We don't need an additional list.

Procedure:

1. We need a third pointer (t), and last (l) pointer that will point on the last node of the LL. and we use p for first list and q for the second list.
2. Set p equal to first (p=first). Move t and l(ast) equal to p (l = p; t = p).
3. Compare p (2) and q(4) which is smaller.(p->data < q->data)
4. In our case p is smaller. Move p to next node(p=p->next). And make l's next null(l->next=NULL)
5. Compare p(8) and q(4) which is smaller.
6. In our q is smaller.Move l's next to q (l->next = q). Move l to q(l=q). Move q to next node(q=q->next). And make l's next null(l->next=NULL)



  

```c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
```

```c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL,*s=NULL,*result_t=NULL;
```

```c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display();
    void Display_s();
    void Display_t();
    void Insert(int index, int x);
    void Insert_s(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void ReverseLL_Elements();
    void ReverseLL_Links();
    void ReverseLL_LinksRecurs(Node *q, Node *p);
    void ConcatLL(Node *q, Node *p);
    void MergingLL(Node *q, Node *p);
}
```

```c++
void LinkedList::MergingLL(Node *q, Node *p){
    Node *res = NULL;
    Node *l=NULL;
    
     if(p->data < q->data){
            res=l=p;
            p=p->next;
            res->next=NULL;
        }else
        {
            res=l=q;
            q=q->next;
            res->next=NULL;
        }
    

    while (p && q){
        if (p->data < q->data){ 
            l->next = p;
            l=p;
            p=p->next;
            l->next=NULL;
        }else
        {
            l->next = q;
            l=q; 
            q=q->next;
            l->next=NULL;
        }
    }
    
    if(p) l->next=p;
    if(q) l->next=q;
    
    result_t=res;
}
```

```c++
void LinkedList::ConcatLL(Node *q, Node *p){
    while(p->next !=NULL){
        p=p->next;
    }
    p->next = q;
}
```

```c++
void LinkedList::ReverseLL_LinksRecurs(Node *q, Node *p){
    if (p){
        ReverseLL_LinksRecurs(p, p->next);
        p->next = q;
    }
    else
        first = q;
}
```

```c++
void LinkedList::ReverseLL_Links(){
    Node *r,*q, *p=first;
    r=NULL;
    q=NULL;
    while (p){
        r = q;
        q = p;
        p=p->next;
        q->next=r;
    }
    first=q;   
}
```

```c++
void LinkedList::ReverseLL_Elements(){
    Node *p=first;
    int *A = new int[Length()]; 
    int i = 0;
    while (p){
        A[i]=p->data;
        p=p->next;
        i++;
        
    }
    p=first;
    i--;
    
    for(; i>=0; i--)
    {
       p->data =  A[i];
       p=p->next; 
    }
}
```

```c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
    
    Node *q=second;
    while(second)
    {
      second=second->next;
      delete q;
      q=first;
    }
}
```

```c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
```

```c++
void LinkedList::Display_s(){
    Node *p=second;
    s = second;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}

```

```c++
void LinkedList::Display_t(){
    Node *p=result_t;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
```

```c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
```

```c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    f = first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
   
}    
```

```c++
void LinkedList::Insert_s(int index, int x){
    Node *t, *p=second;
    s = second;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = second;
        second = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}  
```

```c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
```
