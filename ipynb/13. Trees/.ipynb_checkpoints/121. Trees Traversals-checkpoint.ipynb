 ##  Traversals
 
___


That is visiting all the nodes/vetices of the tree. We have the following possible traversals:

1. Preorder: visit (node), Preorder (left subtree), Preorder (right subtree), or root, left, right
2. Inorder : Inorder (left), visit (node), Inorder (right), or left, root, right
3. Postorder : Postorder (left), Postorder (right), visit (node), or left, right, root 
4. Level order : Level by Level, top to bottom, left to right. 

 
 

## Traverse in Action 1

___

```
                             +------------+
                                   A
                             +------------+
                             /             \
                            /               \
                           /                 \
                          /                   \
                        +-+                   +-+
                         B                     C
                        +-+                   +-+
```

Preorder: $A, B, C$

Inorder: $B, A, C$

Postorder: $B, C, A$

Level order: $A, B, C$

## ## Traverse in Action 3

## ## Traverse Larger Tree

___

Based on the definitions we had (we'll see other ways as well)

```
                         +------------+
                               A
                         +------------+
                         /             \
                        /               \
                       /                 \
            +-----------------++  +-----------------+
            |       +-+        |  |       +-+       |
            |        B         |  |        C        |
            |       +-+        |  |       +-+       |
            |       / \        |  |       / \       |
            |      /   \       |  |      /   \      |
            |     /     \      |  |     /     \     |
            |   +-+    +-+     |  |    +-+   +-+    |
            |    D      E      |  |     F     G     |
            |   +-+    +-+     |  |    +-+   +-+    |
            +-----------------++  +-----------------+
 
                           
                       
```
We break tree into smaller trees. Left child is not just one node, but has its own children.So we group its children into subtree. Same applies to right child.


Preorder: $A, (),()$, 

Now do subtree.

Preorder: $A, (B,D,E),(C,F,G)$

Preorder: $A, B,D,E,C,F,G$

$$$$

We do the same for other, will only give final answer.

Inorder: $D,B,E, A, C,F,C,G$

Postorder: $D,E,B,F,G,C,A$

Level order: $A,B,C,D,E,F,G$
