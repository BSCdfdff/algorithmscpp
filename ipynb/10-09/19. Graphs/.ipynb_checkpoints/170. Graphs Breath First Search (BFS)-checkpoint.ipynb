## Breath First Search (BFS)

____

Remember graph are a non linear data structure. There are two suggested traversals methods:

1. Breath First Search (BFS)
2. Depth First Search (DFS)

they are similar to tree traversal methods. How are the similar?

BFS is similar to level order of BT.


But here we do not have a binary tree, we only have a tree.

We have the following graph, which can be converted into a tree:


```

                                 +---+        +---+
                                 | 1 | +----+ | 2 |
                                 +---+ +-+    +---+
                                   +     |      +
                                   |     |      |
                                   |     |      |
                                   +     +---+  +
                                 +---+       ++---+
                                 | 4 | +----+ | 3 |
                                 +---+        +---+
                                     \       /
                                      \     /
                                       +---+
                                       | 5 |
                                       +---+
                                      /     \
                                     /       \
                                 +---+        +---+
                                 | 6 | +----+ | 7 |
                                 +---+        +---+

```

But how can we convert it to a tree? If a graph that does not have any cycles or closed edges is a tree. Lets prepare a tree (not BT)for this graph:


 
$$BFS: 4,5,1,3,7,6,2$$


What is BFS?


1. We can start traversals from any starting vertex we want. 
2. When you visit that vertex, you must explore that vertex completely, that is, connect all the adjacent edges.
3. While exploring, you can visit adjacent edges in any order.


This is what BFS is.

We use the following textbook terms

1. Visiting (going up a particular vertex)
2. Exploring (visiting add the adjacent vertices of 4)

Visiting, going on that vertex, and Expltoring, going to all the neighbours.


Lets start from vertex 1, But you can start from any vertex.

So here we write down all the visited vertices:

$$\text{BFS:} 1$$

And we use Queue data structure to drop that 1 in queue:

$$\text{Queue:} 1$$


This is the initial starting point. Next, whatever we do, is repeating steps.

1. Remove(same as Visit) vertex (we only have 1), and explore that vertex. Explore means we go to all its neighbours
2. Adjacent is: 2,3,4.  But we can visit it in any order.

$$
\require{cancel}
\text{Queue:} \cancel{1}
$$

3. We do:
    + Visit 2, insert into Queue, and BFS(list). 
    + Visit 3, insert into Queue, and BFS(list). 
    + Visit 3, insert into Queue, and BFS(list). 
    
Vertex 1 is now completely explored.

And we have the following:


$$\text{BFS:} 1,2,3,4$$

The final tree is called BFS Spanning tree. And the dotted edges are called cross edges.And the cross edges only connect vertices close by, that is on same or next level, but does not connect far away vertices.

$$O(n)$$

## Examples of BFS

___


Using the following graph:

```

                                 +---+        +---+
                                 | 1 | +----+ | 2 |
                                 +---+ +-+    +---+
                                   +     |      +
                                   |     |      |
                                   |     |      |
                                   +     +---+  +
                                 +---+       ++---+
                                 | 4 | +----+ | 3 |
                                 +---+        +---+
                                     \       /
                                      \     /
                                       +---+
                                       | 5 |
                                       +---+
                                      /     \
                                     /       \
                                 +---+        +---+
                                 | 6 | +----+ | 7 |
                                 +---+        +---+


```

The following are all valid.


$$\text{BFS 1}:1,4,3,2,5,7,6$$

$$\text{BFS 2}:5,7,3,6,4,2,1$$

$$\text{BFS 3}:2,3,1,5,4,7,6$$
