## Remove Duplicate elements from sorted LL


____

We have the following sorted list:

$$3->5->5->8->8->8$$


(For saving space we only will be taking: 3->5->5->8)

To remove duplicates we will need two pointers, p and q. We need to check in data for current node and next node are the same, if they are the same, then delete any one, but we will delete second one.

If data are not matching, move p to q, and moeve q to q's next, if not matching, we slide them again. 

Lets deal with match first

1. Delete second node.
2. That is q.
3. Before delete make p's next on q's next (p->next= q->next)
4. Delete q.
5. Set q equal to p-s next (q=p->next)

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    void InsertLast(int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void InsertSorted(int x);
    int CheckifSortedLL();
    void RemoveDups();
   
}
````

````c++
void LinkedList::RemoveDups(){
    Node *p=first;
    Node *q=first->next;
    
    while(q){
        if (p->data == q->data){
            p->next= q->next;
            delete q;
            q=p->next;    
        }
        else{
            p=q;
            q=q->next;
        }
    }

}
````

````c++
int LinkedList::CheckifSortedLL(){
    Node *p=first;
    int x = -32768;
    while(p){
        if (p->data < x)
            return 0;
        x=p->data;    
        p=p->next;
    }
    return 1;    
}
````

````c++
void LinkedList::InsertSorted(int x){
    Node *t=NULL,*q=NULL,*p=first;
    
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    //if it is first node
    if (p==NULL)
        first=t;
    else{    
        while(p && p->data < x){
                cout<<"p data: "<<p->data<<endl;
                q=p;
                p=p->next;
        }
        t->next=p;
        q->next=t;
    }
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::InsertLast(int x){
    Node *t=NULL, *p=first;
    Node *last=NULL;
    t = new Node;
    t->data = x;
    t->next=NULL;

    //count number of elements
    //loop till last node.
    for(int i =0; i<Length()-1;i++){
      cout<<"p:"<<p->data<<endl;  
      p=p->next;
    }
    
    last = p;
    
    //There are no nodes
    if(first==NULL)
        first=last=t;
    else{
        last->next=t;
        last=t;
    }
}   
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````

````c++
LinkedList linky;
linky.Insert(0,3);
linky.Insert(1,5);
linky.Insert(2,5);
linky.Insert(3,8);
linky.Insert(4,8);
linky.Insert(5,8);
linky.Display();
linky.RemoveDups();
linky.Display();

````
