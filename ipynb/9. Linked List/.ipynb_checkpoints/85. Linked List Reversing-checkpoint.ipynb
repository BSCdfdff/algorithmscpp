## Reversing a LL


There are two methods:

1. Reversing or interchanging elements (data or contents of nodes, addresses stays the same)(e.g. first, last and then move inwards)
2. Reversing links, thats is changing the addresses of nodes.

E.g if we have :

200-> 210 -> 250 -> 300
2->4->6->8

Now we want 300-> 250->201->200. So the data remains where it is in the list, only the addresses are reversed. Thats means after reversing, where 200 was the first node, now 300 will become first node.

With LL, the preferred method is to reverse links, rather than reversing elements. Why?


Reversing Elements method:

1. Take an array of the same size. We have pointer p, set equal to first. And we have counter i as counter for array.
2. Copy elements in array. Using pointer p, and counter i for array. Copy first element 2, then move p and i to next element. 
3. Reverse copy from array to linked list. Bring p to first node after copy. Keep i at last element in array. Copy last element of array (i) into p, first node, decrement i, and move p to next node. And so we go along. 
4. Now 8 will be the first node
5. Remember the links or addresses stay the same

This method needs an extra array equal to the size of the LL.


 
What are sliding pointers? It is similar to tailing pointers, where the pointers slide or move in unison with each other.

Reversing Links method:

1. Here we will use 3 pointers: r, q and p. p will be set to first, and r and q set to NULL. Seqence of movements is p, then q, then r.
2. Movements can be achieved as followss (r = q, q = p, p=p->next).
3. The above are the steps for sliding pointers.
4. Additional step q's next should point on r (q->next=r)

Above needs clarity. Main idea is that the last node, links must point to previous nodes. And last node becomes 1. You read it the largest to smallest values.

1. See written notes

This is the methid we prefer. We dont like to move data around, as per the first method. As we may be copying large amounts of data, whereas we just simply slide pointers which is of a smaller size.




```c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
```

```c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
```

```c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void ReverseLL_Elements();
    void ReverseLL_Links();
}
```

```c++
void LinkedList::ReverseLL_Links(){
    Node *r,*q, *p=first;
    r=NULL;
    q=NULL;
    while (p){
        r = q;
        q = p;
        p=p->next;
        q->next=r;
    }
    first=q;   
}
```

```c++
void LinkedList::ReverseLL_Elements(){
    Node *p=first;
    int *A = new int[Length()];
    int i = 0;
    while (p){
        A[i]=p->data;
        p=p->next;
        i++;
        
    }
    p=first;
    i--;
    cout<<"***NOW WE REVERSE***"<<endl;
    for(; i>=0; i--)
    {
       p->data =  A[i];
       p=p->next; 
    }
}
```

```c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first; 
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
```

```c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
```

```c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
```

```c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
```

```c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
```

```c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
```

```c++
LinkedList linky;
linky.Insert(0,2);
linky.Insert(1,4);
linky.Insert(2,6);
linky.Insert(3,8);
linky.Display();
linky.ReverseLL_Elements();
linky.Display();
linky.ReverseLL_Links();
linky.Display();

```
