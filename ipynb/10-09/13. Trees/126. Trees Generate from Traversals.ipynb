# Generate BT from Traversals

___

We know work our ways backwards now. We have the traversals, and need to generate three.


E.g we have the following preorder: A, B C

But from this preorder we, awill be able to generate the following 5 trees.


```
      +-----------+     +-----------+  +-----------+              +-----------+        +-----------+
            A                 A              A                          A                    A
      +-----------+     +-----------+  +-----------+              +-----------+        +-----------+
      /            \            \                   \             /                    /
     /              \            \                   \           /                    /
    /                \            \                   \         /                    /
   /                  \            \                   \       /                    /
 +-+                  +-+          +-+                 +-+    -+                  +-+
  B                    C            B                   B     B                    B
 +-+                  +-+          +-+                 +-+    -+                  +-+
                                  /                      \     \                  /
                                 /                        \     \                /
                                /                          \     \              /
                               /                            \     \            /
                             +-+                            +-+   +-+        +-+
                              C                              C     C          C
                             +-+                            +-+   +-+        +-+



```

So from the above, we can see if only preorder is given we cannot generate a unique tree.

This is the catalan number of trees.

$$
\text{Catalan Number}\;\;=\;\;
T(n)\;\; =\;\; 
\frac{^{2n}C_{n}}{n+1}
$$
 
The amount of trees that will be generated.

This applies to a given inorder or postorder, by themselves.

What about a combination, and which combination?

Pre with Post, will generate multiple trees. No, does not work

Pre and In Order, Yes that works.
Post and In Order, Yes that works.

## Count the Nodes

___


This is recursive function.


```
int count (Node *p)

    int x, y;

    if (p){
      x= count(p->lchild);
      y= count(p->rchild);

      return x+y+1;

    }
  return 0;
}  

```

From the above we have procdedure:

1. LChild
2. RChild
3. Then Count

But this looks very similar to

1. LChild
2. RChild
3. Then print

And that is post order.

From this, we can also conclude most of operations, like count, we always will be doing it in post order form. Most commonly used on BT.s

Another version in code:

```
int count (Node *p)

    int x, y;

    if (p)
      return count(p->child) + count(p->rchild) + 1;
   return 0;
}  


```


## Finding Height

___


```
int Height(Node *root){
    int x=0, y =0;
    if (root==0)
        return 0;
    x=Height(root->lchild);
    y=Height(root->rchild);
    if(x>y)
        return x+1;
    else
        return y+1;    
}

```

````c++
 class Node{
public:
    Node *lchild;
    int data;
    Node *rchild;
}*f=NULL;
````

````c++
class Queue{
private:
  int size;
  int front;
  int rear;
  Node **Q;  
    
public:
    Queue(){
        size=100;
        front = rear = 0;
        Q=new Node*[100];
       
        
    }
    Queue(int sz){
        size=sz;
        front = rear = 0;
        Q=new Node*[sz];
    }
    ~Queue(){
        delete []Q;
    }
    void Display();
    void EnQueue(Node *x);
    Node* DeQueue();
    int IsFull();
    int IsEmpty();
    
}
````

````c++
void Queue::EnQueue(Node *x){
    if (IsFull())
      cout<<"Queue Full"<<endl;
    else
    {
      rear = (rear+1)%size;
      if (rear ==0)  rear++;
      Q[rear] =x; 
    }
}
````

````c++
Node* Queue::DeQueue (){
    Node *x=NULL;
    if (IsEmpty())
      cout<<"Queue Empty"<<endl;
    else
    {
      front = (front+1)%size;
      if (front ==0)  front++;
      x=Q[front];
    }
    return x;
}
````

````c++
int Queue::IsFull(){
   if ((rear+1)%size==front){
       return 1;
   }
   return 0;
}
````

````c++
int Queue::IsEmpty(){
    if (rear==front)
      return 1;
    return 0;
}
````

````c++
void Queue::Display (){
   int i = front+1;
   int r =0;
   do
   {   cout<<Q[i]<<" - "<<i<< " ";
       i = (i+1)%size;
       if (i ==0)  i++;
       r =  (rear+1)%size;
       if (r ==0) r++;
   }while(i!= r);
    cout<<endl;
}
````

````c++
class Stack{
private:
  int size;
  int top;
  Node **S;  
    
public:
    Stack(){
        size=10;
        S=new Node*[10];
        top = -1;
        
    }
    Stack(int sz){
        size=sz;
        S=new Node*[sz];
        top = -1;
        
    }
    ~Stack(){
        delete []S;
    }
    //void Display();
    int StackTop();
    int IsFullStack();
    int IsEmptyStack();
    void Push(Node *x);
    Node* Pop();
}
````

````c++
int Stack::StackTop (){
    if (!IsEmptyStack())
      return top;
    return -1;   
}
````

````c++
int Stack::IsFullStack (){
    if (top==size-1)
       return 1;
    return 0;
}
````

````c++
int Stack::IsEmptyStack (){
    if (top==-1)
      return 1;
    return 0;
}
````

````c++
void Stack::Push (Node *x){
    if (IsFullStack())
      cout<<"Stack overflow"<<endl;
    else
    {
      top++;
      S[top] =x; 
    }
    
}
````

````c++
Node* Stack::Pop (){
    Node *x=NULL;
    if (IsEmptyStack())
      cout<<"Stack Empty"<<endl;
    else
    {
      x=S[top];
      top--;
    }
    return x;
}
````

````c++
class Tree{
public:
    Node *root;  
    
    Tree(){
        root = NULL;
    }
    ~Tree(); 
    void CreateTree();
    
    void PreOrderIterative(Node *p);
    void InOrderIterative(Node *p);
    //void PostOrderIterative(Node *p);
    void LevelOrder(Node *p);
    int Height(Node *root);
    int Count(Node *root);    
} 
````

````c++
Tree::~Tree(){
    Node *p=root;
    while(root)
    {
      root=root->rchild;
      delete p;
      p=root;
    }
}
````

````c++
void Tree::CreateTree(){
    Node *p, *t;
    int x, y, z;
    Queue q(100);
    cout<<"Enter root value: "<<endl;
    cin>>x;
    root = new Node;
    root->data=x;
    root->lchild=root->rchild=NULL;
    q.EnQueue(root);
    
    while(!q.IsEmpty()){
        p=q.DeQueue();
        cout<<"Enter left child of " <<p->data<<endl;
        cin>>y;
        if (y!=-1)
        {
            t = new Node;
            t->data=y;
            t->lchild = t->rchild=NULL;
            p->lchild=t;
            q.EnQueue(t);
        }
        cout<<"Enter right child of "<<p->data<<endl;
        cin>>z;
        if (z!=-1)
        {
            t = new Node;
            t->data=z;
            t->lchild = t->rchild=NULL;
            p->rchild=t;
            q.EnQueue(t);
        }
    }
}
````

````c++
void Tree::PreOrderIterative(Node *p){
    Stack s(100); 
    while (p || !s.IsEmptyStack()){
     if (p){
         cout<<p->data<<" ";
         s.Push(p);
         p = p->lchild;
     }
     else
     {
       p = s.Pop();
       p = p->rchild;
     }
   }
}
````

````c++
void Tree::InOrderIterative(Node *p){
    Stack s(100); 

    while (p || !s.IsEmptyStack()){
     if (p){
         s.Push(p);
         p = p->lchild;
     }
     else
     {
       p = s.Pop();
       cout<<p->data<<" ";
       p = p->rchild;
     }
    }
} 
````

````c++
void Tree::LevelOrder(Node *p){

    Queue q(100);
    cout<<p->data<<" ";
    q.EnQueue(p);
    
    while (!q.IsEmpty()){
       p=q.DeQueue();
       if (p->lchild){
           cout<<p->lchild->data<<" ";
           q.EnQueue(p->lchild);
       }
       if (p->rchild){
           cout<<p->rchild->data<<" ";
           q.EnQueue(p->rchild);
       }
    }
}
````

````c++
int Tree::Height(Node *p){
    int x=0, y =0;
    if (p==0)
        return 0;
    x=Height(p->lchild);
    y=Height(p->rchild);
    if(x>y)
        return x+1;
    else
        return y+1;    
}
````

````c++
int Tree::Count (Node *p){

    int x, y;

    if (p)
      return Count(p->lchild) + Count(p->rchild) + 1;
   return 0;
}  
````
