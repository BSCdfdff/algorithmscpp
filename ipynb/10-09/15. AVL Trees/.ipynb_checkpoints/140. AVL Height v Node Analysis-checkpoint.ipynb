## Height is given, what is min/max nodes

___


If height is given,

1. What is minimim Nodes possible (for the given height):

    Here we need to look at a few examples to get formula


2. What is maximim Nodes possible (for the given height)(This is for any Binary Tree):


$$Max\;\; Nodes = 2^{h+1}-1$$

$$n = 2^{h+1}-1$$

But since we starting height at 1, we need to update this formula:


$$Max\;\; Nodes = 2^{h}-1$$

$$n = 2^{h}-1$$

Similar we done this for BST Trees:
    1. General BSTs
    2. Strict BSTs
    3. Complete/Strict M-ary trees

Back to minimum nodes (given height):


Just with one node, given that we start height now counting at 1. Then minimum nodes required will be 1. (Maximum will also be 1)


If h= 2, then n =2, minimum nodes required

If h= 3, then n =3, minimum nodes required.

```


                                            +---+
                                            | 20|
                                          / +---+
                                         /
                                        /
                                   +---+
                                   | 10|
                                  /+---+
                                 /
                                /
                             +---+
                             | 5 |
                             +---+



```

```

                                            +---+
                                            | 20|
                                          / +---+ \
                                         /         \
                                        /           \
                                   +---+           +---+
                                   | 10|           |20 |
                                  /+---+           +---+
                                 /
                                /
                             +---+
                             | 5 |
                             +---+


```

Again, If h= 3, then n =4, minimum nodes required, to make it a valid AVL tree.

We can prepare a table to see all possibilities:

$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
h\T& 1\T&2 \T&3 \T&4  \T& 5 \T&6 \T& 7 \\\hline
n(h)\T& 1\T&2 \T&4 \T&7  \T& 12 \T&20 \T& 33 \\\hline
\hline
\end{array}
$$

To get value, we we add the previous two numbers + 1.

Formula for minimum nodes:


$$
n(h) =
\begin{cases}
0  & _\text{where $h = 0$ } \\
\\
1  & _\text{where $h = 1$ } \\
\\ 
n(h-1) + n(h-2)+ 1 & _\text{otherwise}
\end{cases}
$$

(On aside, this formula is a the same as fibonacci series formula. Its also a balanced series. Any two consequtive numbers, the ratio will be 1.6 approximately)

In short we can say:

$$ Min\;\;Nodes = \text{Look in table}$$


## Nodes are given, what is min/max Height

___

With BT we converted min formula to maximum formula. We do the same thing here, but working from the above formula as out base.
    

If n nodes are given,

1. What is minimim Height possible (for the given n nodes):


We know max nodes (given height) is $n = 2^{h}-1$. So by converting we get:

$$Min\;\;Height = log_2{n+1}$$

2. What is maximim Height possible (for the given n nodes):

We know min nodes (given height) is $n =  \text{Look in table}$. So by converting we get (similar):

$$ Max\;\;Height = \text{Look in table}$$

We go further and use the formula to get maximum height, we get approximate value of:

$$ Max\;\;Height = 1.44log_2{n+2}$$

So max is also log, or multiple of log.

The conclusion, AVL tress height is always i.t.o. log.
