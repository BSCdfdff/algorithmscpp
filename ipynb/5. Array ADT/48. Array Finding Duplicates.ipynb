```c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
```

```c++
class Array{
private:
    int *A;
    int size;
    int length;
    void Swap (int *x, int *y);
    int SumNatNos(int y);
    bool FindDiffenceSuccess (int x,int y, int z);
public:
    Array(){
        size=10;
        A=new int[10];
        length = 0;
        
    }
    Array(int sz){
        size=sz;
        A=new int[sz];
        length = 0;
        
    }
    ~Array(){
        delete []A;
    }
    void Display();
    void Insert (int index, int x);
    void FindDplicatesinSorted();
    void FindDplicatesinSortedCounted();
    
}
```

```c++
int Array::SumNatNos (int y){
    return  (y*(1+y))/2;    
}

```

```c++
bool Array::FindDiffenceSuccess (int x, int y, int z){
    int a=0;
    a= x - y;
    if (a==z){
     return true;    
    }
    return false; 
}
```

```c++
void Array::Swap (int *x, int *y){
    int temp;
    temp =*x;
    *x=*y;
    *y = temp;
}
```

```c++
void Array::FindDplicatesinSorted (){
    int lastDuplicate =0;
    for(int i=0; i < length; i++ ){
        if (A[i]==A[i+1] && A[i] !=lastDuplicate){
           cout<<A[i]<<endl;       
        }
        lastDuplicate=A[i];
    }
}
```

```c++
void Array::FindDplicatesinSortedCounted (){
    int j = 0;
    for(int i=0; i < length; i++ ){
        if (A[i]==A[i+1]){
           j=i+1; 
           while(A[j]==A[i])j++;
           cout<<A[i]<<" is appearing "<<(j-i)<<" times."<<endl;  
           i=j-1; 
        }
    }
}
```

```c++
void Array::Insert(int index, int x){
   if  (index>=0 && index<=length){
     for(int i=length-1;i>=index;i--)
         A[i+1]=A[i];
     A[index]=x;
     length++;  
   }
}
```

```c++
void Array::Display(){
    for(int i=0; i<length;i++)
      cout<<A[i]<<" ";      
    cout<<endl;      
}
```

```c++
Array arr(10);
arr.Insert(0,3);
arr.Insert(1,6);
arr.Insert(2,8);
arr.Insert(3,8); 
arr.Insert(4,10);
arr.Insert(5,12);
arr.Insert(6,15);
arr.Insert(7,15);
arr.Insert(8,15);
arr.Insert(9,20);
arr.FindDplicatesinSorted();

Array arrA(10);
arrA.Insert(0,3);
arrA.Insert(1,6);
arrA.Insert(2,8);
arrA.Insert(3,8); 
arrA.Insert(4,10);
arrA.Insert(5,12);
arrA.Insert(6,15);
arrA.Insert(7,15);
arrA.Insert(8,15);
arrA.Insert(9,20);
arrA.FindDplicatesinSortedCounted();

```
