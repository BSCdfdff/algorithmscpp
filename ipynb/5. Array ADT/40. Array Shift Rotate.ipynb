## Difference between a normal Insert At, and Rotate/Shift

___

With a normal insert we insert and stop until we are at the index. We insert the value at this supplied index, BUT then we increase/increment the length.

With Shift or Rotate, we NEVER increase the length, as we not increasing the length.

In next lession, we have situation where we need insert a value into a sorted list. And this is a combination of:

1. Rotation to Right (shift to right, but increase(increment) the length as well)
2. Insert At

___

```c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
```

```c++
class Array{
private:
    int *A;
    int size;
    int length;
    void Swap (int *x, int *y);
public:
    Array(){
        size=10;
        A=new int[10];
        length = 0;
        
    }
    Array(int sz){
        size=sz;
        A=new int[sz];
        length = 0;
        
    }
    ~Array(){
        delete []A;
    }
    void Display();
    void Insert (int index, int x);
    void Revers1();
    void Revers2();
    void LeftRotat();
    void RightRotat();
    void LeftRevers();
    void RightRevers();
}
```

```c++
void Array::LeftRotat(){
    int temp;
    int firstelement = A[0];
    for(int i=0; i<length; i++){
       if(i==length-1) 
           A[i]=firstelement;
       else  
           A[i]=A[i+1];
    }
}
```

```c++
void Array::RightRotat(){
    int temp;
    int lastelement = A[length-1];
    for(int i=length-1; i>=0; i--){
       if(i==0) 
           A[i]=lastelement;
       else  
           A[i]=A[i-1];
    }
}
```

```c++
void Array::LeftRevers(){
    int temp;
    for(int i=0; i<length; i++){
       if(i==length) 
           A[i]=0;
       else  
           A[i]=A[i+1];
    }
}
```

```c++
void Array::RightRevers(){
    int temp;
    for(int i=length; i>=0; i--){
       if(i==0) 
           A[i]=0;
       else  
           A[i]=A[i-1];
    }
}
```

```c++
void Array::Swap (int *x, int *y){
    int temp;
    temp =*x;
    *x=*y;
    *y = temp;
}
```

```c++
void Array::Revers1(){
    int *B= new int[14];
    
    for(int i=length-1, j = 0; i>=0, j<length-1; i--, j++){
       B[j] = A[i]; 
    }
    
    for(int i=0; i<length-1;i++){
       A[i] = B[i]; 
    }
    
    
    for(int i=0; i<14;i++)
      cout<<B[i]<<" ";      
    cout<<endl;  
    
    
}
```

```c++
void Array::Revers2(){
    int temp;
    for(int i=0, j = length-1; i<j; i++, j--){
       cout<<"i:"<<A[i]<<endl;
       cout<<"j:"<<A[j]<<endl;
       Swap(&A[i], &A[j]);
    }
    
}
```

```c++
void Array::Insert(int index, int x){
   if  (index>=0 && index<=length){
     for(int i=length-1;i>=index;i--)
         A[i+1]=A[i];
     A[index]=x;
     length++;  
   }
}
```

```c++
void Array::Display(){
    for(int i=0; i<length;i++)
      cout<<A[i]<<" ";      
    cout<<endl;      
}
```

```c++
Array arr(14);
arr.Insert(0,4);
arr.Insert(1,8);
arr.Insert(2,10);
arr.Insert(3,15);
arr.Insert(4,18);
arr.Insert(5,21);
arr.Insert(6,24);
arr.Insert(7,27);
arr.Insert(8,29);
arr.Insert(9,33);
arr.Insert(10,34);
arr.Insert(11,37);
arr.Insert(12,39);
arr.Insert(13,41);
arr.Insert(14,43);
arr.Display();
cout<<"Left Rotate"<<endl;
arr.LeftRotat();
arr.Display();
cout<<"Right Rotate"<<endl;
arr.RightRotat();
arr.Display();
cout<<"Left Reverse"<<endl;
arr.LeftRevers();
arr.Display();
cout<<"Right Reverse"<<endl;
arr.RightRevers();
arr.Display();
```
