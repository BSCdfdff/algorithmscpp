## Row Major Mapping

___

This is for indexes that starts at 0
 
Here we have Four(4) Dimensions:

 
$$\text {type } A[d1][d2][d3][d4]$$

With row, we know we move from left to right.


$$\text{Address of A}[i_1][i_2][i_3][i_4]$$
$$=  L_0 + [i_1\times{d_2}\times{d_3}\times{d_4}+{i_2}\times{d_3}\times{d_4}+{i_3}\times{d_4}+{i_4}]\times w$$

How to read the above:

$${i_1} \text{ is multiplied with}\rightarrow {d_2}\times{d_3}\times{d_4}$$ 

That is, after ${d_1}$, the rest of dim must be multiplied

$${i_2} \text{ is multiplied with}\rightarrow  {d_3}\times{d_4}$$ 

That is, after ${d_2}$, the rest of dim must be multiplied

$${i_3} \text{ is multiplied with}\rightarrow  {d_4}$$ 

That is, after ${d_3}$, the rest of dim must be multiplied

Then we just have ${i_4}$, as there is no more indexes

Lets rewrite:

$$L_0 + [\underline{ {i_1}\times {d_2}\times{d_3}\times{d_4}} + 
\underline{ {i_2}\times{d_3}\times{d_4} } + 
\underline{ {i_3}\times{d_4} } + 
\underline{ {i_4} }]  \times w
$$

## General formula for n Dimensions for Row Major

___
 
This is for indexes that starts at 0
 
**Picture**
 
This is for indexes that starts at 0
 
Here we have Four(4) Dimensions:

 
$$\text {type } A[d1][d2][d3][d4]$$

With row, we know we move from left to right.


$$\text{Address of A}[i_1][i_2][i_3][i_4]$$
$$=  L_0 + [i_1\times{d_2}\times{d_3}\times{d_4}+{i_2}\times{d_3}\times{d_4}+{i_3}\times{d_4}+{i_4}]\times w$$


Lets rewrite:

$$L_0 + [\underline{ {i_1}\times {d_2}\times{d_3}\times{d_4}} + 
\underline{ {i_2}\times{d_3}\times{d_4} } + 
\underline{ {i_3}\times{d_4} } + 
\underline{ {i_4} }]  \times w
$$ 

$$ $$ 

Lets firstly deal with the $i\text{'s}$. (indexes).


All underlined terms are added. So there are 4 terms. For addition we use $\sum$ (sum) symbol and ${i_1},{i_2},{i_3},{i_4} \text { of }i$ is used with $\sum$. And for the subscript ${1,2,3,4}$ we will use $p$. So $p$ will start at $1$ and end at $4$. But instead, we end it, at a more generic $n$. Now we can write:

$$\sum_{p=1}^n {i_p} $$ 


Secondly lets deal with $d\text{'s}$. (dimensions).

Every index is multiplied with dimensions, that is product of dimenssions, and we use the $\prod$ symbol. And that will be $\prod$ of $d$. And similar to the sum logic above: These dimensions are taking subscripts from ${2,3,4}$, and we will use $q$ to represent it. But subscripts starts at always as values after $p$. That is $p+1$. So we make $q$ starting point $p+1$, and end point 4, or a more generic $n$. Now we can write:

$$$$

$$\prod_{q=p+1}^n {d_q} $$ 

$$$$

Lastly the whole equation is multiplied by $w$




$$$$

$$\sum_{p=1}^n[{i_p}\times \prod_{q=p+1}^n {d_q}\ ]\times w $$ 

$$ $$ 

## General formula for n Dimensions for Column Major

___
 
This is for indexes that starts at 0
 
**Picture**
 
Here we have Four(4) Dimensions:

 
$$\text {type } A[d1][d2][d3][d4]$$

With column, we know we move from right to left.


$$\text{Address of A}[i_1][i_2][i_3][i_4]$$
$$=  L_0 + [i_4\times{d_1}\times{d_2}\times{d_3}+{i_3}\times{d_1}\times{d_2}+{i_2}\times{d_1}+{i_1}]\times w$$



Lets rewrite:

$$L_0 + [\underline{ {i_4}\times {d_1}\times{d_2}\times{d_3}} + 
\underline{ {i_3}\times{d_1}\times{d_2} } + 
\underline{ {i_2}\times{d_1} } + 
\underline{ {i_1} }]  \times w
$$ 

$$ $$ 

Lets firstly deal with the $i\text{'s}$. (indexes).


All underlined terms are added. So there are 4 terms. For addition we use $\sum$ (sum) symbol and ${i_1},{i_2},{i_3},{i_4} \text { of }i$ is used with $\sum$. And for the subscript ${1,2,3,4}$ we will use $p$. So $p$ which represents the subscripts of $i$ will start at 4, but a more generic $n$. Ans $p$ will end at $1$. Now we can write:

$$\sum_{p=n}^{1} {i_p} $$ 


Secondly lets deal with $d\text{'s}$. (dimensions).

Every index is multiplied with dimensions, that is product of dimenssions, and we use the $\prod$ symbol. And that will be $\prod$ of $d$. Now the subscripts of $d$, which is represented by $q$, starts at $1$, and ends at $p-1, and we represent this with $n$. Now we can write:

$$$$

$$\prod_{q=1}^{(p-1)} {d_q} $$ 

$$$$

Lastly the whole equation is multiplied by $w$




$$$$

$$\sum_{p=n}^n[{i_p}\times \prod_{q=1}^{(p-1)} {d_q}  ]\times w $$ 

$$ $$
