## Heapify

___

Heapify is concerned with creating/deletion a HEAP, but in a more efficient way. Lets explore..

Previously we saw, that in order to insert  a value into HEAP (more information see previous sessions)

1. We firstlsy find bottom, and insert from bottom
2. Then work our way to top, and get the correct position of the value, in 
3. That is by creating a Leaf Node, and sending it up the tree, possibly towards root. (worst case)

Previously we saw, that in order to delete a value into HEAP (more information see previous sessions)


1. We delete root
2. Get the lowest value from leaf node, and replace root value with this value.32. And rearrange the items 
3. And rearrange the tree, to maintain it max/min characteristic.

Direction for Insert: We move from bottom to top

Direction for Delete: We move from top to bottom.

How does Heapify improve this operation?

1. We have the following array which we can create a BT from:


$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|}
\hline 
5\T&10\T&30\T&20\T&35\T&40\T&15  \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7 \\\hline 
_\T&_\T&_\T&_\T&_\T&_\T&_{} \\\hline 
\end{array}
$$

2. And we can create a BT:

```
                                           +---+
                                           | 5 |
                                        /  +---+  \
                                       /           \
                                      /             \
                                 +---+             +---+
                                 | 10|             | 30|
                                /+---+\           /+---+\
                               /       \         /       \
                              /         \       /         \
                           +---+       +---+ +---+       +---+
                           | 20|       | 35 || 40|       | 15|
                           +---+       +- -+ +---+       +- -+
 

```

3. But the above is not a MAx HEAP. But we created A HEAP, by taking 5, as the initial value, and insert  values into HEAP by scanninig next values in array. (we know this process). And we ended up with following tree:


```
                                           +---+
                                           | 40|
                                        /  +---+  \
                                       /           \
                                      /             \
                                 +---+             +---+
                                 | 30|             | 35|
                                /+---+\           /+---+\
                               /       \         /       \
                              /         \       /         \
                           +---+       +---+ +---+       +---+
                           | 5 |       | 20| | 10|       | 15|
                           +---+       +- -+ +---+       +- -+
```

But time taken is $$O(nlog_2{n})$$


With Heapify we do the following:

1. We have ou array:


$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|}
\hline 
5\T&10\T&30\T&20\T&35\T&40\T&15  \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7 \\\hline 
_\T&_\T&_\T&_\T&_\T&_\T&_{} \\\hline 
\end{array}
$$

2. From which we have created a BT:

```

                                           +---+
                                           | 5 |
                                        /  +---+  \
                                       /           \
                                      /             \
                                 +---+             +---+
                                 | 10|             | 30|
                                /+---+\           /+---+\
                               /       \         /       \
                              /         \       /         \
                           +---+       +---+ +---+       +---+
                           | 20|       | 35| | 40|       | 15|
                           +---+       +---+ +---+       +---+

```

3. With the previous approach, we starting creating/inserting into HEAP, by scanning the array from the first element, that is were index = 1; And we moved from left to right.
4. With Heapify we start at the last item, that is 15, index = 7. Here we wil move from right to left
5. But also we wont work towards the parent, but work downwards (from element, e.g 15). That is previously eements were compared with parent, now elements will ebe compared with children.Seem odd, but lets explore...
6. So we compare 15 with children
7. Does 15 have any children? No
8. So only 15 is there, single element. So its a HEAP.
9. Move the next element 40
10. So we compare 40 with children
11. Does 40 have any children? No
12. So only 40 is there, single element. So its a HEAP.
13. Move the next element 35, same no children, its HEAP
14. Move the next element 20, same no children, its HEAP
15. From the above we have not done any modifications.
16. Next element is 30.
17. Does 30 have any children? Yes
18. Here we do the Insert comparison (as before), but perform it downwards, towards it children.
19. And we swap 40 with 30. And thats a HEAP now, after the swap:

```
                                           +---+
                                           | 5 |
                                        /  +---+  \
                                       /           \
                                      /             \
                                 +---+             +---+
                                 | 10|             | 40|
                                /+---+\           /+---+\
                               /       \         /       \
                              /         \       /         \
                           +---+       +---+ +---+       +---+
                           | 20|       | 35| | 30|       | 15|
                           +---+       +---+ +---+       +---+

```

20. Next element will be 10.
21. Does 10 have any children? Yes
22. Here we do the Insert comparison (as before), but perform it downwards, towards it children.
23. And we swap 35 with 10. And thats a HEAP now, after the swap:

```
                                           +---+
                                           | 5 |
                                        /  +---+  \
                                       /           \
                                      /             \
                                 +---+             +---+
                                 | 35|             | 40|
                                /+---+\           /+---+\
                               /       \         /       \
                              /         \       /         \
                           +---+       +---+ +---+       +---+
                           | 20|       | 10| | 30|       | 15|
                           +---+       +---+ +---+       +---+

 
```
24. Next element will be 5.
25. Does 5 have any children? Yes
26. Here we do the Insert comparison (as before), but perform it downwards, towards it children.
27. And we swap 5 with 40. And thats a HEAP now, after the swap:


```

                                           +---+
                                           | 40|
                                        /  +---+  \
                                       /           \
                                      /             \
                                 +---+             +---+
                                 | 35|             | 5 |
                                /+---+\           /+---+\
                               /       \         /       \
                              /         \       /         \
                           +---+       +---+ +---+       +---+
                           | 20|       | 10| | 30|       | 15|
                           +---+       +---+ +---+       +---+

```
28. And this is a MAX HEAP, and we have done very few adjustments.
29. But is it similar to the MAX Heap we created, using the normal way? No, it is not.
30. But does it matter? No.
31. Its still a Max Heap.

Heapify is the fastest method for creating a HEAP and takes

$$O(n)$$
