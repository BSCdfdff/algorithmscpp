## Symmetric Matrix

___


Lets call this $M$:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  \\ 
{_2}  \\
{_3}  \\
{_4}  \\
{_5}  \\
\end{array}
\begin{bmatrix}
2 & 2 & 2 & 2 & 2\\
2 & 3 & 3 & 3 & 3\\
2 & 3 & 4 & 4 & 4\\
2 & 3 & 4 & 5 & 5\\
2 & 3 & 4 & 5 & 6\\
\end{bmatrix}
$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
& {_1} &  & {_2} &  & {_3} &  & {_4} &  & {_5}  
\end{array}
$$
$$$$


if we swap $i$ and $j$, the we get the same combination of values. That is ```M[3,4]```, is the same as ```M[4,3]```.

 
$$
i = \text { index of row} \\
j = \text { index of column}
$$


Here we have:

$$
\require{cancel} 
\text{if  } M[i,j],  \text{    then  } M[j,i] 
$$


That means it I know i, j, I will know j, i. So that means we do not have to store eveything.

So we can either store the upper triangular matrix or the lower triangular matrix. they both produce the same result.
