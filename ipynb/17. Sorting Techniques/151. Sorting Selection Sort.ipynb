Summary. We look for a position and find out which value should fill this position

So the comparisons ans swaps where as follows:

First pass
1. Five comparisons
2. One Swap


Second pass
1. Four comparisons
2. One Swap


Third pass
1. Three comparisons
2. One Swap

Forth pass
1. Two comparisons
2. One Swap

Fifth pass
1. Once comparisons
2. One Swap

$$\text{No of Comparisons}=O({n^2}) $$
$$\text{No of Swaps}=O({n}) $$

This is the only algorithm with minimum number of swaps. 

That is to avoid swaps, we have taken j and k pointers. That is once we find the correct position, then we interchange. We do not interchange all the time.

Also when I perform k number of passes, we get k smallest elements. That is intermediate results are useful/
