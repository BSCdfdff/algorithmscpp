## Analysis of Row Major formula

___

How many multiplications are performed for getting the result.

Here we have Four(4) Dimensions:

 
$$\text {type } A[d1][d2][d3][d4]$$

With row, we know we move from left to right.


$$\text{Address of A}[i_1][i_2][i_3][i_4]$$
$$=  L_0 + [i_1\times{d_2}\times{d_3}\times{d_4}+{i_2}\times{d_3}\times{d_4}+{i_3}\times{d_4}+{i_4}]\times w$$


Lets rewrite:

$$L_0 + [\underline{ {i_1}\times {d_2}\times{d_3}\times{d_4}} + 
\underline{ {i_2}\times{d_3}\times{d_4} } + 
\underline{ {i_3}\times{d_4} } + 
{i_4} ]  \times w
$$ 

Number of multiplicat

$$\;\;\;\;3\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; 2\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;1\;\;\;\;\;\;\;\; $$ 


so for:
$$4\text{ Dim } =  3+2+1$$
$$5\text{ Dim } =  4+3+2+1$$


$$ $$

$$ $$

$$\text{O}({n^2})$$

To reduce the number of multiplications

## Horners Rule

Lets take only the portion that was in the brackets

$$[\underline{ {i_1}\times {d_2}\times{d_3}\times{d_4}} + 
\underline{ {i_2}\times{d_3}\times{d_4} } + 
\underline{ {i_3}\times{d_4} } + 
{i_4} ] 
$$ 

And rewrite the formula from right to left:


$$[ {i_4}+{i_3}\times{d_4}+{i_2}\times{d_3} + {i_1}\times{d_2}\times{d_3}]   
$$ 

Now start removing common terms.

${i_4}$ is fine, 

Next common term is ${d_4}$

$$ {i_4}+ {d_4} \left[{i_3}+{i_2}\times{d_3} + {i_1}\times{d_2}\times{d_3}\right]   
$$ 

Next common term is ${d_3}$

$$ {i_4}+ {d_4} \times\left[{i_3}+{d_3} \times\left[ {i_2} + {i_1}\times{d_2}\right]\right]    
$$ 

Now there is nothing more common

But now we have only : $ 3 \text{ multiplications }  $ 


So for:
$$4\text{ Dim } =  3$$
$$5\text{ Dim } =  4$$
 

$$ $$  

$$n\text{ Dim } =  (n-1)$$
  
$$ $$  

$$\text{O}({n})$$


By taking commons we have reduced the number of multiplications. And this rule is called Horners Rule.
