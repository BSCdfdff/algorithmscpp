## Deleting Node from LL

We have the following scenarios:

1. Delete first Node
2. Delete at given position

Deleting the first Node:

1. Get pointer ```p```, and set it equal to the first node (```p = first```)
2. Move first to the next node (```first=first->next```)
3. ```delete p```
4. Optionally , get the value of ```p```, before deleting it, and assign it to x (```x = p->data```)


Delete at given position:

1. Get point ```p``` and set it equal to first. (```p=first```)
2. Get also a $\text{ trailing }$ point ```q```, so allow us to get access to previous node. (```q``` will be one step behind ```p```)(```q=p; p = p->next```)
3. Remember we loop $n-1$ less (```n``` = index to be deleted)
4. Once we get to index:
5. Set q's next equal to p's next (```q->next = p->next```)
4. Delete p
5. Optionally , get the value of ```p```, before deleting it, and assign it to x (```x = p->data```)




````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    void InsertLast(int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void InsertSorted(int x);
   
}
````

````c++
void LinkedList::InsertSorted(int x){
    Node *t=NULL,*q=NULL,*p=first;
    
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    //if it is first node
    if (p==NULL)
        first=t;
    else{    
        while(p && p->data < x){
                cout<<"p data: "<<p->data<<endl;
                q=p;
                p=p->next;
        }
        t->next=p;
        q->next=t;
    }
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::InsertLast(int x){
    Node *t=NULL, *p=first;
    Node *last=NULL;
    
    
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    //Check if first position
    //count number of elements
    for(int i =0; i<Length()-1;i++){
      cout<<"p:"<<p->data<<endl;  
      p=p->next;
     
        
    }
    
    last = p;
    cout<<"p:"<<p<<endl;  
    cout<<"p->data:"<<p->data<<endl;  
    cout<<"last: "<<last<<endl;
     
    
    
    //There are no nodes
    if(first==NULL)
        first=last=t;
    else{
        
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        
        
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        last->next=t;
        last=t;
    }
    
   
}   
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
