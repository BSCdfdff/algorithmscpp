## Max Number of Nodes (Height is Given)

___


From the above we have establised that the Ma Number of nodes formula is the same for General and Struct Binary Tree.

That is:

$$Max\;\; Nodes = 2^{h+1}-1$$




## Min Number of Nodes (Height is Given)

___

With min number of nodes we must satisfy the rule of 0 ot 2 children. That means we went from bottom 2, then up on 2, then up one 2, till we get root of 1.

So formula is:

$$Min\;\; Nodes = 2h+1$$

Test height = 2

$$Min\;\; Nodes = 2\times2+1=5$$


Once we have the above Min and Max to get the number of nodes (given height). Now we can get the height (given nodes) formula, by using the nodes formula.



## Min Height (Given Nodes)

___



We know from General Tree, the min Height we can get from max nodes formula:, which is 

$$Max\;\; Nodes = 2^{h+1}-1$$

After converting it (was done in previous session), we get:

 

$$Min\;\;Height=log_2(n+1)-1$$

## Max Height (Given Nodes)

___

This we can get max height from the min node formula:


$$Min\;\; Nodes = 2h+1$$

$$n = 2h+1$$

$$2h+1 = n$$

$$2h= n-1$$
$$h=\frac{n-1}{2}$$

Test: 

$$h=\frac{n-1}{2}$$

$$n=3$$
$$h=\frac{3-1}{2}=\frac{2}{2}=1$$

```

                                         Node=3

                                          +-+          Height
                                          +-+        + <----+ 0
                                          / \        |
                                         /   \       |
                                        /     \      |
                                       /       \     |
                                     +-+       +-+   |
                                     +-+       +-+   |<-----+ 1


```


$$n=7$$
$$h=\frac{7-1}{2}=\frac{6}{2}=3$$
