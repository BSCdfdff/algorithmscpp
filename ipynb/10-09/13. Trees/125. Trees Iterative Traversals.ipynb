## Preorder Iterative Traversals

___

Remember with Recursion, an internal or system stack is used.

So to convert recursive function to iteratibve function, that is a function with loops, then we NEED to use a STACK as well.

We know the procedure is :

1. print Node (root)
2. Left Child
3. Right Child

Using the iterative approach, with out own STACK:

1. Add pointer p at Root Node, address 200 (8). (check if p is not NULL, only then print)
2. Print data 8 at this node, adress 200. 
3. Add 200 to stack.
4. Move pointer to left child, address 210 (3)
5. Print data 3 at this node, address 210.
6. Add 210 to stack.
7. Move pointer to left child, address 270 (4)
8. Print data 4 at this node, address 270.
9. Add 270 to stack.
10. Move pointer to left child, address NULL
11. In stack we have: 270, 210, 200. (Top to bottom)
12. We need to empty the stack whenever we get NULL address
13. Pop off 270 (4) to p pointer.
14. Move pointer to right child, address NULL
15. Pop off 210 (3) to p pointer.
16. Move pointer to right child, address 300 (9)
17. Print data 9 at this node, address 300.
18. Add 300 to stack.
19. Move pointer to left child, address NULL.
20. In stack we have: 300, 200. (Top to bottom)
21. We need to empty the stack whenever we get NULL address
22. Pop off 300 (9) to p pointer.
23. Move pointer to right child, address NULL
24. Pop off 200 (8) to p pointer.
25. Move pointer to right child, address 230 (5)
26. Print data 5 at this node, address 230.
27. Add 230 to stack.

Repeat steps 4 to 22 for address 230(5)

19. Move pointer to left child, address 230 (5)
20. Print data 3 at this node, address 230.
21. Add 230 to stack.
22. Move pointer to left child, address 310 (7)
23. Print data 4 at this node, address 310.
24. Add 310 to stack.
25. Move pointer to left child, address NULL
26. In stack we have: 310, 230. (Top to bottom)
27. We need to empty the stack whenever we get NULL address
28. Pop off 310 (7) to p pointer.
29. Move pointer to right child, address NULL
30. Pop off 230 (5) to p pointer.
31. Move pointer to right child, address 320 (2)
32. Print data 2 at this node, address 320.
33. Add 320 to stack.
34. Move pointer to left child, address NULL.
35. In stack we have: 320. (Top to bottom)
36. We need to empty the stack whenever we get NULL address
37. Pop off 320 (5) to p pointer.
38. Move pointer to right child, address NULL 
39. Stack is empty, and p is null, thats when we stop

## Inorder Iterative Traversals

___

With small change we can convert to Inorder Traversal


We know the procedure is :

1. Left Child
2. print Node (root)
3. Right Child

So instead of printing, first we get the left child, and only print when p becomes null, That is we pop address from stacj to get new address, print, and then move to right child.


$$$$

$$In\;\;order = 4,3,9,8,7,5,2$$

$$$$


```
void InOrderIterative(Node p){
    int x = -1

    while (p!=NULL || IsEmpty()){
     if (p!=NULL){
         x = p->data
         s.Push(x);
         p = p->lchild;
     }
     else
     {
       p = s.Pop();
       x = p->data
       cout<<x<<" ";
       p = p->rchild;
     }
}


```

## Postorder Iterative Traversals

___

Here we need to do a bit more work.


We know the procedure is :

1. Left Child
2. Right Child
3. print Node (root)

The main point here:

1. We need to push the address two times
  + One for going and right side (+ve)
  + Second time, so we can print it (-ve)
2. That is before going to right side, we push the address back into stack, to allow us to print it later.  Then go right child. 
3. But how do we know which address is for right side, and other is for left side???
4. We use negative addresses?? But address are of node type, they cannot be negative. But we cast it to negative and cast it back to Node type.
5. First time we pushing positivbe address. Means go to right child
6. Second time we pushing negative address. Means print the data.
 
$$$$

```

   +-----+-----+-----+  +-----+-----+-----+      +-----+-----+-----+  +-----+-----+-----+
           270                  300                      310                  320
```


We will be traversing as follows Top to bottom, Left to right

$$$$

$$Level\;\;order = 8,3,5,4,9,7,2 $$

$$$$

Here we will need a Queue of NodeType.

We know the procedure is :

1. Top to Bottom
2. Left to Right


1. Start with pointer p at root, address 200
2. Print 8 ast this address
3. Add address 200 to Queue,
3. Queue contents: 200
4. This was initial step, the rest will be repeating steps.
5. Dequeue address 200, and assign to p.
6. Move pointer to left child, address 210 (3)
7. Print 3 ast this address
8. Add (Enqueue) address 210 to Queue.
9. Move pointer to right child, address 230 (5)
10. Print 5 ast this address
11. Add (Enqueue) address 230 to Queue.
12. Now repeating steps done (left and right child done and in Queue). 
12. Queue contents: 210, 230

Now we repeat steps 5 to 11.

13. Dequeue address 210(3), and assign to p. (FIFO)
14. Move pointer to left child, address 270 (4)
15. Print 4 ast this address
16. Add (Enqueue) address 270 to Queue.
17. Move pointer to right child, address 300 (9)
18. Print 9 ast this address
19. Add (Enqueue) address 300 to Queue.
20. Queue contents: 230, 270,300

Repeat

21. Dequeue address 230(5), and assign to p. (FIFO)
22. Move pointer to left child, address 310 (7)
23. Print 7 ast this address
24. Add (Enqueue) address 310 to Queue.
25. Move pointer to right child, address 320 (2)
26. Print 2 ast this address
27. Add (Enqueue) address 320 to Queue.
28. Queue contents: 270,300, 310, 320


Repeat
21. Dequeue address 270(4), and assign to p. (FIFO)
22. Move pointer to left child, address NULL
22. Move pointer to right child, address NULL
21. Dequeue address 300(9), and assign to p. (FIFO)
22. Move pointer to left child, address NULL
22. Move pointer to right child, address NULL
21. Dequeue address 310(7), and assign to p. (FIFO)
22. Move pointer to left child, address NULL
22. Move pointer to right child, address NULL
21. Dequeue address 320(2), and assign to p. (FIFO)
22. Move pointer to left child, address NULL
22. Move pointer to right child, address NULL
23. And we stop.

````c++
class Node{
public:
    Node *lchild;
    int data;
    Node *rchild;
}*f=NULL;
````

````c++
class Queue{
private:
  int size;
  int front;
  int rear;
  Node **Q;  
    
public:
    Queue(){
        size=100;
        front = rear = 0;
        Q=new Node*[100];
       
        
    }
    Queue(int sz){
        size=sz;
        front = rear = 0;
        Q=new Node*[sz];
    }
    ~Queue(){
        delete []Q;
    }
    void Display();
    void EnQueue(Node *x);
    Node* DeQueue();
    int IsFull();
    int IsEmpty();
    
}
````

````c++
void Queue::EnQueue(Node *x){
    if (IsFull())
      cout<<"Queue Full"<<endl;
    else
    {
      rear = (rear+1)%size;
      if (rear ==0)  rear++;
      Q[rear] =x; 
    }
}
````

````c++
Node* Queue::DeQueue (){
    Node *x=NULL;
    if (IsEmpty())
      cout<<"Queue Empty"<<endl;
    else
    {
      front = (front+1)%size;
      if (front ==0)  front++;
      x=Q[front];
    }
    return x;
}
````

````c++
int Queue::IsFull(){
   if ((rear+1)%size==front){
       return 1;
   }
   return 0;
}
````

````c++
int Queue::IsEmpty(){
    if (rear==front)
      return 1;
    return 0;
}
````

````c++
void Queue::Display (){
   int i = front+1;
   int r =0;
   do
   {   cout<<Q[i]<<" - "<<i<< " ";
       i = (i+1)%size;
       if (i ==0)  i++;
       r =  (rear+1)%size;
       if (r ==0) r++;
   }while(i!= r);
    cout<<endl;
}
````

````c++
class Stack{
private:
  int size;
  int top;
  Node **S;  
    
public:
    Stack(){
        size=10;
        S=new Node*[10];
        top = -1;
        
    }
    Stack(int sz){
        size=sz;
        S=new Node*[sz];
        top = -1;
        
    }
    ~Stack(){
        delete []S;
    }
    //void Display();
    int StackTop();
    int IsFullStack();
    int IsEmptyStack();
    void Push(Node *x);
    Node* Pop();
}
````

````c++
int Stack::StackTop (){
    if (!IsEmptyStack())
      return top;
    return -1;   
}
````

````c++
int Stack::IsFullStack (){
    if (top==size-1)
       return 1;
    return 0;
}
````

````c++
int Stack::IsEmptyStack (){
    if (top==-1)
      return 1;
    return 0;
}
````

````c++
void Stack::Push (Node *x){
    if (IsFullStack())
      cout<<"Stack overflow"<<endl;
    else
    {
      top++;
      S[top] =x; 
    }
    
}
````

````c++
Node* Stack::Pop (){
    Node *x=NULL;
    if (IsEmptyStack())
      cout<<"Stack Empty"<<endl;
    else
    {
      x=S[top];
      top--;
    }
    return x;
}
````

````c++
class Tree{
public:
    Node *root;  
    
    Tree(){
        root = NULL;
    }
    ~Tree(); 
    void CreateTree();
    
    void PreOrderIterative(Node *p);
    void InOrderIterative(Node *p);
    //void PostOrderIterative(Node *p);
    void LevelOrder(Node *p);
    //int Height(Node *root);    
} 
````

````c++
Tree::~Tree(){
    Node *p=root;
    while(root)
    {
      root=root->rchild;
      delete p;
      p=root;
    }
}
````

````c++
void Tree::CreateTree(){
    Node *p, *t;
    int x, y, z;
    Queue q(100);
    cout<<"Enter root value: "<<endl;
    cin>>x;
    root = new Node;
    root->data=x;
    root->lchild=root->rchild=NULL;
    q.EnQueue(root);
    
    while(!q.IsEmpty()){
        p=q.DeQueue();
        cout<<"Enter left child of " <<p->data<<endl;
        cin>>y;
        if (y!=-1)
        {
            t = new Node;
            t->data=y;
            t->lchild = t->rchild=NULL;
            p->lchild=t;
            q.EnQueue(t);
        }
        cout<<"Enter right child of "<<p->data<<endl;
        cin>>z;
        if (z!=-1)
        {
            t = new Node;
            t->data=z;
            t->lchild = t->rchild=NULL;
            p->rchild=t;
            q.EnQueue(t);
        }
    }
}
````

````c++
void Tree::PreOrderIterative(Node *p){
    Stack s(100); 
    while (p || !s.IsEmptyStack()){
     if (p){
         cout<<p->data<<" ";
         s.Push(p);
         p = p->lchild;
     }
     else
     {
       p = s.Pop();
       p = p->rchild;
     }
   }
}
````

````c++
void Tree::InOrderIterative(Node *p){
    Stack s(100); 

    while (p || !s.IsEmptyStack()){
     if (p){
         s.Push(p);
         p = p->lchild;
     }
     else
     {
       p = s.Pop();
       cout<<p->data<<" ";
       p = p->rchild;
     }
    }
} 
````

````c++
void Tree::LevelOrder(Node *p){

    Queue q(100);
    cout<<p->data<<" ";
    q.EnQueue(p);
    
    while (!q.IsEmpty()){
       p=q.DeQueue();
       if (p->lchild){
           cout<<p->lchild->data<<" ";
           q.EnQueue(p->lchild);
       }
       if (p->rchild){
           cout<<p->rchild->data<<" ";
           q.EnQueue(p->rchild);
       }
    }
}
````
