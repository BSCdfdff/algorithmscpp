## Bin/Bucket Sort

___


Similar to count sort, where we need extra memory space.


This is index based sorting. This is one:

1. Easiest to understand
2. And Fastest.
3. But consumes lots of memory

We have a list of element in array. And we need to sort it. We then have another array, called Count, and size of array is equal to the largest element of the array we want to sort. The largest element is 9, so we have array of size 10 (0-9). And we need to initialise array Count with some empty values. 

But instead of initializing the Count array with 0, we inilialize it to null. Which means those null values are pointers. But pointers to what?


So we need to scan array A, and drop these elements into Bins of count array.


Procedure:

1. Scan array A.
2. At index 0, get value 8.
3. Scan array Bins, and insert element 8, at index 8.

8. Next we need to empty all the bins from left to right, and copy it back to array A, but sorted.

## Lets Code

____

Code is very similar to Count sort.

We use double pointers, as Bins is pointer to pointer.

And we also use an insert function, to insert elements at the end of LL

We have one while loop to scan through Bins.

Then we have another while loop, to copy the elements to array A, and Delete it elements from Bin, and whichever element is deleted, it is inserted into array A.

That is, it is similar to Count sort, instead here we are emptying bins.

```
void BinSort(int A[], int n){

  int max, i, j;
  Node **Bins
  
  max = findMax(A,n);
  Bins = new Node*[max+1];
  
  for(i=0; i <max+1;i++)
    Bins[i]=NULL;
   
  for(i=0;i<n;i++)
    Insert(Bins[A[i]], A[i]);
  
  i=0, j=0;
  while(i<max+1){
    while(Bins[i]!=NULL)
      A[j++]=Delete(Bins[i]);
    i++;  
  }
}

```
