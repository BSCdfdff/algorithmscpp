## Hashing Functions

___



One of the primary requirements of Hashing Function is that the values in hash table should be uniformed distributed, and should not be concentrated.



And the selection of hash function depends on the programmer.
 
With Chaining the hash table size can be anything

$$ h1(x) = x\%{size} + 1$$

## Mid Square Method.

____

Here we do a square of key, and take the middle digit. Lets explore.

Key = 11


1. Perform square of 11. 

$$11^2=121$$

2. Then take the middle digit 2.

So 11 will be stored at index 2.

When we search for the key, we apply square, take middle digit, and go to that location.

We can also take multiple middle digits, where we have (after squaring), lots of digits.







 

## Folding Method

____


Lets say the key= 123347. Here we have a 6 digit key. Then we start "folding" the digits, that is we take 12, then 33, then 47. And addintg them up.

$$12+33+47=92$$

So we can store the number in index 92. And if that is too big, we can perform mor add 9+2 to give 11.

Remember when searching we need to perform the same procedure.

So the above are numeric keys. What if we have string and keys. Key ="ABC". How do we convert this into index in array.

We can take ASCII codes:


A=65

B=66

C=67

Amd make it a single number 656667. Peform folding

$$65+66+67=198$$ 

And then if this is out of range, we can drop 1.

So from the above is clear, we can DESIGN our own hash function, but with emphasis on uniformly distributing the values.
