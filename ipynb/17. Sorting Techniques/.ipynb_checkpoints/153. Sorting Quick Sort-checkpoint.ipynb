## Quick Sort

___

Mostly discussed and most famous technique. Sometimes its seen as diffifult, but it is very interesting. 

It revolves around the following basic idea. Lets explore:

If we have the following list, what can we see in single glance:

$$\fbox{10}, 30, 20,70, 40, 90, 80$$

We can see that 10 is already sorted. Just by looking at the list. That is, it is the smallest element.

From the following list we can see in single glance, that 90 is already sorted.

$$30, 40, 20,70, 50, 10, \fbox{90}$$


In single glance, which element is already sorted.already sorted? it will be 50.

$$40, 30, 20,\fbox{50}, 90, 70, 80$$


But on what bases is 50 sorted? The 10 and 90 was easy to see? But how did we determine that 50 was sorted.

We see:

+ All elements before 50, is smaller than 50
+ All elements after 50, is greater than 50

From the above, we can find a definition of quick sort: It works on the idea that a element is sorted if, all the elements  before that element is less and all elements after that element is more. Then that element is in sorted position.

## Lets Code
___

Lets explore, by taking 3 elements:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}

_0\T&_1\T&_2\T&_3  \\\hline 
_i\T&\T&\T&_{j} \\\hline 
\end{array}
$$



1. The vlaue 20 is pivot. We include infinity as starting point for j, and i will start at 20.
2. Increment i until pivot(20), while pivot is less than i, and only stop, when pivot becomes greater than i, in our case we stop at 30 

```
do {
      i++;
}while(A[i]<=pivot);
```

3. Decrement j until pivot(20), while pivot is greater than j, and only stop, when pivot becomes greater than i, in our case we stop at 30 

4. But i becomes greater than j, so we swap j and pivot



```
int Partition(int A[], int l, int h){
  int pivot=A[l];
  int i=l, j = h;
  
  do{
  
    do {
      i++;
    }while(A[i]<=pivot);
    
    do {
      j--;
    }while(A[j]>pivot;
    
    if(i < j)
        Swap(A[i], A[j]);
        
  
  }while(i<j);
  
  Swap(A[l], A[j]);
  
  return j;

}
```
