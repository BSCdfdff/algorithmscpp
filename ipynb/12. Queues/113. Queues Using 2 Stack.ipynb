We can construct the following rule:

1. Push to S1
2. But first check S2, if Empty, transfer (i.e. Pop from S1, Push to S2) all elements to S2
3. But first check S2, if Full, Pop from S2.
4. If both S1 and S2 are empty, Queue is empty.
