Inserting at any given position

2. Create new node with temporary pointer t,and set the value.
3. Loop p into n - 1, then stop (this is position 1 less that where you want to insert)
4. Set t's next equal to p's next (```t->next = p->next```)
5. Set t's prev equal to p. As above explanation this is important. (```t->prev = p```)
6. Need to check if there is a node after p-next? 
7. If p's next is not NULL, that means there is a node, but how can we get p-next, i.e the node after p-next, how to we get its prev.
8. Just the way we say it: ```p-next-prev=t```, is the pointer coming back from 
9. And p's next wil be equal to t (```p->next=t```

## Deleting HEAD Node

___



1. Take pointer p on this node, that is first (p=first)
2. Move first to next node.(first=first->next)
3. Remove value, save it, and delete p
4. First's prev should now  be null.
5. But should also check if first is only node (extra step)


Very simple.


## Deleting node at given position

___

Deleting a given node.

1. Here, instead of using three pointers, we will do it using only one pointer.
2. We use p pointer to get to node we want to delete, n-1

We need to modify the following nodes:

1. P-'s prev's next should point on p's next. (```p->prev->next = p->next```). That is the forword link from 3 to 2. Remember we busy deleting 7.
2. P's next's prev should point to p's prev (```p->prev->next = p->next```) . That is link from 2 to 3.

3. But we need check if there is any node after p, If there is no node, then it is an error.
4. Thus we must check if p's next is null, if not then ONLY do we perform 2 above. Otherwise will get error here

A bit more thinking involved, as we only using one pointer.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{ 
public:
    Node *prev;
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display();
    LinkedList(int A[], int n);
    int Length();
    void Insert(int index, int x);
    int Delete(int index);
 
}
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    
    
    int x = -1;
    
    if(index <0 || index > Length())
        return -1;  
    
    
    if(index==0){
        p=first;
        if (first->next!=NULL){
            first = first->next;
            first->prev=NULL;
            x=p->data; 
            delete p;
        }
        else{
            x=p->data; 
            delete p;
        }
    }
    else{
        p=first;
        for(int i =0; i<index-1;i++)
        {
           p=p->next;
        }
        
        p->prev->next = p->next;
        
        if (p->next)
            p->next->prev = p->prev;
         
        x=p->data;
        delete p;
        
    }  
    return x;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p;
    
    if (first==NULL){
        first= new Node;
        first->prev=first->next=NULL;
    }        
    
    p=first;
    
    if(index <0 || index > Length())
        return;
    
    if(index==0){
        t = new Node;
        t->data = x;
        t->prev=NULL;
        t->next = first;
        first->prev=t;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        
        t = new Node;
        t->data = x;
        
        t->next= p->next;
        t->prev = p;
        
        if (p->next)
          p->next->prev = t;      
        
        p->next = t;
      
    }
}   
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->prev=first->next=NULL;
    last = first; 
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i]; 
        t->next=NULL;
        //dont agree
        //last next is null???
        //does no harm
        //t->next=last->next;
        
        t->prev = last;
        last->next=t;
        last=t; 
    }     
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    if (first->next==NULL){
         cout<<first->data<<" ";
    }
    else{
        while(p){
            cout<<p->data<<" ";
            p=p->next;
        }
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    
    if (first->next==NULL)
         len++;
    else{
        while(p){
            len++;
            p=p->next;
        }
    }
    return len;
}
````
