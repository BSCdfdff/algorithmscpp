# Tower of Hanoi

## Tower of Hanoi
___





Problem: We have 5 disks on Tower A, and need to move it over Tower C,

Rules:

1. One Disk at time
2. No larger disk should be over a smaller disk
3. Can use Tower B, as axillary tower


A = from (this is always the source) (1st parameter)

B = using (this is the intermediary) (2nd parameter)

C = to (this is always the destination) (3rd parameter)

So A is the source (move from), and C is the destination (move to). Those are the parameter positions. And B is just intermediary position.S


This problem can use solved using recursion.





Lets break up problem for:

1. One Disk (set)
2. Two Disk (set)
3. Three Disk (set)

**One Disk (set)**

TOH (1, A, B, C)

Move Disk from A to C using B (this move is the destination)

Whenever we have the above statement, its final move, the rest is taken as recursive function


A = Source
C = Destination
B = using (not really, but use it)

**Two Disk (set)**

TOH (2, A, B, C)

1. Move Disk from A to B, not to C

But this move is NOT the destination. So moving there is temporary or we can say that moving there is recursively.

So,

1. TOH (2, A, C, B)


How we we read above? A is source went to destination, which must be in place where C (3rd parameter). So that leaves C, to take the place of the 2nd parameter 

2. Move the larger disk from A to C

Is this the final destination? Yes, its same step, where we have a single disk, so we write:

Move Disk from A to C, using B 

But this is not 

Next Move disk form B to C 

Is this the final destination? No. This is also a recursive step

3. TOH (1, B, A, C)

How do we read this? B is source and needs to go to destination, so We put dowb C, and A is the intermediary.


**Three Disk (set)**

This is a combination of moving 1 disk and 2 disks, ie. the functions above gets reused.

TOH (3, A, B, C)

1. TOH (2, A, C, B)
2. Move Disk from A to C, using B
3. TOH (2, B, A, C)



Step 1, we perform what we did when we had 2 disks.

It seem that we moving 2 disks, but we simple reusing the recursive function of it.

Step 2, this is single move

Step 3. We reuse step 2 function again.

From the above we can generate move for n number of disks


TOH (n, A, B, C)

1. TOH (n-1, A, C, B)
2. Move Disk from A to C, using B
3. TOH (n-1, B, A, C)

## Using recursion


Lets trace it .






In order transversal..

How much time

n= 3, making 15 calls
n=2, making 7 calls


Which is:

$$1+2+ {2^2} + {2^3} = {2^4} - 1 = 15 $$

Time taken is
$$1+2+ {2^2} + {2^3}+...+ {2^n}= {2^{n+1}} - 1 $$


$$\text O(2^n) $$


$$\text{Exponential} $$


Very time consuming..
