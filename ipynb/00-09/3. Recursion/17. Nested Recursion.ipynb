# Nested Recursion

## Nested Recursion




Where inside the recursive function when it calls itself, the parameter/argument used is the recursive function again:

```
void fun(int n){              
    if (n > 0){                
        cout<<n<<endl;         
        fun(fun(n-1));              
    }
}
```

So the nested fun(n-1) must first give a result, before outer fun() can be executed.



````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
int fun(int n);
````

````c++
int fun(int n){              
    if (n > 100)                
       return n -10;
    else    
       return fun(fun(n+11));              
    
}
````

````c++
int y, x=98;
y =fun(x);
````
