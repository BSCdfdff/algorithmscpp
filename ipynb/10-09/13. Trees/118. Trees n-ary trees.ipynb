## Min Nodes (Given Height)

___

For min nodes, givem the height, we observe the following:



Where h =2, we count 3 nodes from bottom, then another three above it, then 1

Where h =3, we count 3 nodes from bottom, then another three above it, then another three above it, then 1

So its height times m-ary (its our case its 3, which is the degree) + 1

$$Min = 3 (degree) \times {h} +1$$

$$Min \;\;Nodes= m \times {h} +1$$

$$n= m{h} +1$$

## Max Nodes (Given Height)

___

For max nodes, givem the height, we observe the following:



Where h =2, we have levels: 1, 3, 9

Where h =3, we have levels: 1, 3, 9, 27

Which looks like our old GP series, that is 

$$ a + ar + ar^2 + ar^3 +...+ ar^k = \frac{a(r^{k+1} - 1)}{r -1}$$$$

So in our case:

1. $a = 1$
2. $r = m$
3. $k = h$


$$Max\;\; Nodes = \frac{(m^{h+1} - 1)}{m -1}$$

$$n=\frac{(m^{h+1} - 1)}{m -1}$$
