## Analysis of all Operations

___


### Push

1. Must first check of stack is not full before inserting


```
if st->top != st->size -1

```

2. Insert value into stack if not full

```
st=>top++
st->S[st->top] =x

That is insert the value after you have incremented the index in the array

```

### Pop

1. Must first check of stack is stack is not empty

```
int x = -1;
if st->top != -1
```

2. Store the value and decrement the top pointer

```
x = st->S[st->top]
st->top--; 
```
3. If nothing is deleted, -1 wil be returned.

From the above the following is becoming evident. When inserting a value we assign first move the index to the next empty space in array, then we assign a value to this index position. When we delete an element, we get the value at this position, store it in some variable, and decrement that index of the array, where that value was stored. 

So when we push we increase the index of array by 1, and assign a value, and when we delete we remove(store for later use) a value from stack, by decrementing an the index of array.

Time Taken by Push and Pop operations is constant. $O(1)$

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Stack{
private:
  int size;
  int top;
  int *S;  
    
public:
    Stack(){
        size=10;
        S=new int[10];
        top = -1;
        
    }
    Stack(int sz){
        size=sz;
        S=new int[sz];
        top = -1;
        
    }
    ~Stack(){
        delete []S;
    }
    void Display();
    int StackTop();
    int IsFull();
    int IsEmpty();
    void Push(int x);
    int Pop();
    int Peek(int pos);
}
````

````c++
int Stack::StackTop (){
    if (!IsEmpty())
      return top;
    return -1;   
}
````

````c++
int Stack::IsFull (){
    if (top==size-1)
       return 1;
    return 0;
}
````

````c++
int Stack::IsEmpty (){
    if (top==-1)
      return 1;
    return 0;
}
````

````c++
void Stack::Push (int x){
    if (IsFull())
      cout<<"Stack overflow"<<endl;
    else
    {
      top++;
      S[top] =x; 
    }
    
}
````

````c++
int Stack::Pop (){
    int x=-1;
    if (IsEmpty())
      cout<<"Stack Empty"<<endl;
    else
    {
      x=S[top];
      top--;
    }
    return x;
}
````

````c++
int Stack::Peek (int pos){
    int x=-1;
    int mapPos = top - pos +1;    
    if (mapPos < 0)
      cout<<"Invalid Pos"<<endl;
    else
    {
      x=S[mapPos];
    }
    return x;
}
````

````c++
void Stack::Display (){
   for (int i = top; i>=0; i--)
   {
     cout<<S[i]<<" ";
   }
   cout<<endl;
}
````

````c++
Stack s(5);
s.Push(10);
s.Push(20);
s.Push(30);
s.Push(40);
s.Push(50);
s.Display();
````
