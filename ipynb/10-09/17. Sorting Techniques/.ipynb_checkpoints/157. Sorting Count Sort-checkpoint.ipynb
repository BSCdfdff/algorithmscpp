## Analysis

___

Time Taken:

1. We scan through array A, which has n elements
2. We scan through array Count, which has n elements

$$O(n)$$

So the time taken is Linear


Space Taken:

Its need a big size of array, as the size is determined by the largest(maximum) element of the source array.

So if the maximum size of largest element is big, then we cannot use count array.

6. Based on these counts in array C, we need copy the elements "back" array A. Lets use a while loop, using i and j counters. Lets use j for C array. We need to move j, while index is 0. if not 0, then copy index (value) to A, at index i of A, and increment j. and decrement value if C. But do not move i, only move i if index is zero (do this in if)

```
while(i < max +1){
    if(C[i] > 0){
      A[j++]=i;
      C[i]--
    }
    else
      i++;
}

```

```
void CountSort(int A[], int n){
    int max, i,j;
    int *C;
    
    max = FindMax(A,n);
    C= new int[max+1]; //we start at index 0
    
    for (int i, i < max+1, i++)
      C[i]=0;
      
    for (int i, i < n, i++)
      C[A[i]]++;
    
    i = 0, j =0;
     
    while(i < max +1){
        if(C[i] > 0){
          A[j++]=i;
          C[i]--
        
        }
        else
          i++;
    }

}

```


````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<< 
#define EXTRACTION_OPERATOR operator>> 
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
int FindMax(int A[],int n){
  int max = numeric_limits<int>::min(); 
  for (int i=0; i < n; i++){
     if (A[i] > max)
         max=A[i];
  }
  return max;  
} 
````

````c++
void CountSort(int A[], int n){
    int max, i,j;
    int *C; 
    
    max = FindMax(A,n);
    C= new int[max+1]; //we start at index 0
    
    for (i=0;i < max+1; i++)
      C[i]=0;
      
    for (i=0; i < n; i++)
      C[A[i]]++;
    
    i = 0, j =0;
     
    while(j < max +1){
        if(C[j] > 0){
          A[i++]=j;
          C[j]--;
        
        }
        else
          j++;
    }

}
````

````c++
//Sorted:  2 3 5 6 7 8 10 12 
int T[] = {2,5,8,12,3,6,7,10};
int n=8;

CountSort(T,n);

for (int i = 0; i<n;i++){
     cout<<T[i]<<" ";
 }
 cout<<endl;
````
