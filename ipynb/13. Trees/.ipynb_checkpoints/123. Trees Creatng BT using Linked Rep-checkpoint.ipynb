## Creating BT using Linked Rep

___

```

                         +------------+
                               5
                         +------------+
                         /             \
                        /               \
                       /                 \
                      /                   \
                    +-+                   +-+
                     8                     6
                    +-+                   +-+
                      \                   /
                       \                 /
                        \               /
                       +-+             +-+
                        9               3
                       +-+             +-+
                       / \
                      /   \
                     /     \
                   +-+    +-+
                    4      2
                   +-+    +-+

```


We start generatin the tree from root node. Then we create left child, then right child. But we first need to check if child is there, before we can attacha another.


Ad we will create the tree level by level.

We will use a Queue, to store the tree, together witk Linked representation, that is using Nodes.

Remember we need to know how to use :

1. Queue (circular)
2. Doubly linked List

Precedurally, 

1. Create first node
2. And set left anf right child to NULL
3. And set root equal to this node.
4. Set its value
5. And the address is $200$

```
                                           root
                                   +-----+-----+-----+
                                   |     |  5  |     |
                                   +-----+-----+-----+
                                           200
```
5. And insert the address in Queue


$$
Q=

This is the initial work that we wil do, the rest wil be a repetition, once we get the steps right.

(Repetition)

6. Remove $200$ from $Q$
7. Set pointer $p$ equal to that address $(200)$
8. First, check if is has left child (address $200$, node $p$)
    + From our tree, yes it has left child (value $8$)
    + Create node $t$
    + Insert value $8$
    + And set left and right child to NULL
    + Set $p$'s left child equal to $t$ (lets say address is $210$)
    + Insert $210$ into $Q$
    
    
```
                                          root
                                   +-----+-----+-----+
                                   | 210 |  5  |  \  |
                                   +-----+-----+-----+
                                           200
                                 /             
                                /               
                               /                 
                              /    
                   
                   +-----+-----+-----+
                   |  \  |  8  |  \  |
                   +-----+-----+-----+
                           210 
```

9. Now do the same for right child (address $200$, node $p$)
    + From our tree, yes it has left child (value $6$)
    + Create node $t$
    + Insert value $6$
    + And set left and right child to NULL
    + Set $p$'s right child equal to $t$ (lets say address is $230$)
    + Insert $230$ into $Q$
10. The above from step 6 to 9 is complete batch, that will be repeated. That is 

    + Get new address from $Q$
    + Make pointer $p$ pointer point to address
    + Find left and right child (node $t$)
    + Create new node $t$
    + Point $p$ to new node $t$
    + And insert $t$ into $Q$
    
```

                                   +-----+-----+-----+
                                   | 210 |  5  | 230 |
                                   +-----+-----+-----+
                                           200
                                 /                    \
                                /                      \
                               /                        \
                              /                          \

                   +-----+-----+-----+               +-----+-----+-----+
                   |  \  |  8  |  \  |               |  \  |  6  |  \  |
                   +-----+-----+-----+               +-----+-----+-----+
                           210                               230
```
