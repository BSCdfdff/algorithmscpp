## Tridiagonal Matrix

___




Lets call this $M$:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  \\ 
{_2}  \\
{_3}  \\
{_4}  \\  
{_5}  \\
\end{array}
\begin{bmatrix}
a_{11} & a_{12}   & 0      & 0      & 0      \\
a_{21} & a_{22}   & a_{23} & 0      & 0      \\
0      & a_{32}   & a_{33} & a_{34} & 0      \\
0      & 0        & a_{43} & a_{44} & a_{45} \\
0      & 0        & 0      & a_{54} & a_{55} \\
\end{bmatrix}
$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
& {_1} &  & {_2} &  & {_3} &  & {_4} &  & {_5}  
\end{array}
$$
$$$$

Here we only have three "diagonals" of values. One in main diagonla, then upper diagonal, then lower diagonal.


$$
i = \text { index of row} \\
j = \text { index of column}
$$

Here we can make the following observations: The elements, that is i and j, have the same value in the main diagonal. In the lower diagional, if you subtract i and j, it is equal to 1. In the upper diagonal, if you subtract i and j, it is equal to -1.



We can thus formulate:

Main Diagonal:
$$ i - j = 0 $$

Lower Diagonal:
$$ i - j = 1 $$

Upper Diagonal
$$ i - j = -1 $$

So we can combine these three conditions into one formula:

$$ |i - j| \le {1} $$


That is if you subtract i and j, and you get a value less than or equal to 1, that is an absolute value, those values must be non zero.


We can thus say that:

Here we have:

$$$$
$$$$

$$
\require{cancel} 
M[i,j] = 0,  \text{    if } \lvert{i}-{j}\rvert \gt {1} 
$$

$$$$

$$
\require{cancel} 
M[i,j] \bcancel{=} 0,  \text{    if } \lvert{i}-{j}\rvert \le {1} 
$$


Some examples of the above:

**For ALL Zero values:**

Lest take ```M[4,1]```
$$
\lvert{i}-{j}\rvert \gt {1} 
$$
$$
 \lvert{i}-{j}\rvert=\lvert{4}-{1}\rvert =3
$$

$$
\lvert{i}-{j}\rvert \gt {1}

$$

$$

$$

$$

$$
\lvert{i}-{j}\rvert \gt {1} 
$$
$$
 \lvert{i}-{j}\rvert=\lvert{3}-{5}\rvert =2
$$


**For ALL NON-Zero values:**

Lest take ```M[2,2]```
$$
\lvert{i}-{j}\rvert \le {1} 
$$
$$
 \lvert{i}-{j}\rvert=\lvert{2}-{2}\rvert =0

## Non Zero Elements ($\require{cancel} \bcancel{=} 0$)

___

How non zero elements do we have?


Upper diagonal has 4 non zero elements, Main diagonal has 5  non zero elements,  Lower diagonal has 4  non zero elements

Non Zero elements:
$$5+4+4 = 13 \text{ elements}$$

So for any $n$ number of elements:

$$n+n-1+n-1$$


$$3n-2$$

Which has the following formula:

$$\text{Non-Zero Elements } = 3n-2$$

In our case:

$$3(5)-2= 15 - 2 = 13  $$





## Zero Elements

___


And since we have a square matrix: $n\times{n} = n^2$.

Thats is, to get the remaining rows and colums:

$$\text{Zero Elements } = \text{Total } - (3n-2)$$
$$$$
$$n^2 - (3n-2)$$
$$$$


In our case:

$$ 5^2 - 3(5) + 2 = 25 -15 + 2 = 12 $$

## How to access elements from single dim array

___

Let fill the elements row by row:


$$

{_2}  \\

$$


$$

$$
$$$$



From the above we can see that the rows or even cols dont have uniform number of elements. Rather we can represent us diagonal by diagonal.

Which diagonal do we take first? Upper, Main or Lower.


Let fill the array, starting with first row, then second row, third, fourth and fifth row.


Lets do it in the following order:
1. Lower
2. Main
3. Upper



Lets call this $A$:


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|r|r|r|r|}
\hline 
a_{21}\T&a_{32}\T&a_{43}\T&a_{54}\T&a_{11}\T&a_{22}\T&a_{33}\T&a_{44}\T&a_{55}\T&a_{12}\T&a_{23}\T&a_{34}\T&a_{45}   \\\hline 
_0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8\T&_9\T&_{10}\T&_{11}\T&_{12}  \\\hline
_{Lower Diag}\T&\T&\T&\T&_{Main Diag}\T&\T&\T&\T&\T&_{Upper Diag}\T&\T&\T&  \\\hline
\hline 
\end{array}
$$


From the above we can see that the rows or even cols dont have uniform number of elements. Rather we can represent us diagonal by diagonal.

Which diagonal do we take first? Upper, Main or Lower.
 

Lets build up a formula for accessing a specific element:

Lets look any any n case:

To get $\text{index}$ for **Lower Diagonal**:

$$$$

$$\text{if } i-j = 1$$
Then, 

$$A[\text {index }] = i-1$$

$$$$


To get $\text{index}$ for **Main Diagonal**:

$$$$

$$\text{if } i-j = 0$$
We need to skip the lower diagonal elements, which is $n -1$, and to get to the index, it will be $i - 1$, 

Then, 

$$A[\text {index }] = n-1+i-1$$

$$$$


To get $\text{index}$ for **Upper Diagonal**:

$$$$

$$\text{if } i-j = -1$$
We need to skip the lower and main diagonal elements, which is $2n -1$, and to get to the index, it will be $i - 1$, 

Then, 

$$A[\text {index }] = 2n-1+i-1$$

$$$$
