## Priority Queues

___

Two methods to implement priority queues:

1. Limited set of priorities
2. Element Priority


## Limited set of priorities

____

This method is useful in operating systems. Some Os used priority based scheduling. So higher priority queues will executed first.

So we have the following (limited) priorities (1 to 3), where the following letter has the following prioriries:

## Element Priority

___


This is not FIFO.


Here the element has its own priority. E.g. element 8 has priority 8.


$$
Elements \; \; \;
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
6&8&2&10&15&2&9&17&5 \\\hline
\hline
\end{array}
$$
 
 
 
$$
Q \; \; \;
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|}
\hline 
&&&&&&&& \\\hline
\hline
\end{array}
$$

We have rules:

0. This is not FIFO.
1. Smaller number, highest priority


Method 1:

1. Insert in same order as you get it. $O(1)$
2. Delete maximum priority by searching for it, and shift. $O(n)$


Method 2:

1. Insert in increasing order of priority,search for it, and shift to get priority right $O(n)$
2. Delete last element of array $O(1)$
