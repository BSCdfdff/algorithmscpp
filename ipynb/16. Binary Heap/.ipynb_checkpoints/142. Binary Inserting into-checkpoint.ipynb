## Inserting into a Heap
___



We will be doing it for MAX Heap. The same will apply for MIN Heap.

We have the following MAx Heap (Completed) BT:


```


                           height                                          height
                           + 0-->                  +---+                   <--0 +
                           |                       | 30|                        |
                           |                     / +---+ \                      |
                           |                    /         \                     |
                           |                   /           \                    |
                           |             +---+               +---+              |
                           | 1-->        | 20|               |15 |         <--1 |  1-1=0
                           |            /+---+\             /+---+\             |
                           |           /       \           /       \            |
                           |          /         \         /         \           |
                           |       +---+       +---+   +---+       +---+        |
```

We start array by 1, as since we take the formula:

Node at index i:

$$ \text{left child} = 2\times i ($$

$$ \text{right child of i} = 2\times{i} + 1$$

So we keep the array index starting at 1.



We want to insert key=40:

Procedure:

1. For max Heap, Root must always have the largest element.
2. Key 40, must be in root.
3. Insert key 40, into next free space in array, that is index 8, why? We want ot maintain a complete BT.

$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline

_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8 \\\hline 
_\T&_\T&_\T&_\T&_\T&_\T&_\T&_{i} \\\hline 
\end{array}
$$

4. Where is 40 in the BT?  Key 40 is at index 8. Index 8 is the lchild of Node 5. 

$$ \text{left child} = 2\times i = 2\times4 = 8$$


5. It will be the Lchild of 5.


```

                                                   +---+
                                                   | 30|
                                                 / +---+ \ 
                                                /         \
                                               /           \

                             | 40|
                             +---+

```

6. Another way is to find the parent of key 40.:


$$ \text{parent index} = \frac{index}{2} = \frac{8}{2} = 4$$

7. At index 4, we have node 5.
8. By inserting into the next blank space in the array, we also maintaining the complete BT, as we inserting the key at the correct position in the tree.
9. Key 40 is inserted, but is it a max heap? No. Key 40 must be the root.
10. We need to rearrange BT, for key 40 to become the root.
11. This is the repeating step coming now.
12. Compare 40 with parent, that is node 5 (we know how get parent index, formula above).
13. Is Node 40 greater, or less than parent 5? 
14. Is is greater, Copy 40 to temp, copy node 5 to Node 40 place.
15. Set i to parent node index (5), that is 4.

## Time complexity and others

___

We always insert new element as a leaf node. Then we arrange the elements. And we will be moving up the tree. So it s dependent on height of Tree. Anad what is height of tree, is always $log_2{n}$. That is insertion time is always $log_2{n}$.


$$Insertion = O(log_2{n})$$
 



## How to create a HEAP from scratch

___

We have the following array:

$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
10\T&20\T&20\T&25\T&5\T&40\T&35  \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7 \\\hline 
_{i}\T&\T&_\T&_\T&_\T&_\T&_{} \\\hline 
\end{array}
$$

1. Insert first element of the array in Heap

```
                                                   +---+
                                                   | 40|
                                                   +---+
                                                   
                                                   
```

2. Next element (index 2) in array is 20.
3. Insert Key 20 into BT. But his is not max HEAP so we follow the INSERT procedure above to move Node 20 to root:

```


                                          +---+
                                          | 10|
                                        / +---+
                                       /
                                      /
                                +---+
                                | 20|
                                +---+

```
4. After following the INSERT procedure, that is rearranging the BT, we have the following MAX HEAP:

```


                                          +---+
                                          | 20|
                                        / +---+
                                       /
                                      /
                                +---+
                                | 10|
                                +---+
```
5. But our source array also go updated:

$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|}
\hline 
20\T&10\T&20\T&25\T&5\T&40\T&35  \\\hline 
_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7 \\\hline 
_\T&\T&_{i}\T&_\T&_\T&_\T&_{} \\\hline 
\end{array}
$$

6. That is we have a HEAP defined (elements up to 20), inside a source array. 
7. This is called IN PLACE. 

8. Next element (index 3) in array is 30.
9. Insert Key 30 into BT. But his is not max HEAP so we follow the INSERT procedure above to move Node 30 to root:

```

```
