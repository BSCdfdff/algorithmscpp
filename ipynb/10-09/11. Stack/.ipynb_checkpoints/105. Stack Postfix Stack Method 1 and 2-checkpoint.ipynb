```c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
```

```c++
class Stack{
private:
  int size;
  int top;
  char *S;  
    
public:
    Stack(){
        size=10;
        S=new char[10];
        top = -1;
        
    }
    Stack(int sz){
        size=sz;
        S=new char[sz];
        top = -1;
        
    }
    ~Stack(){
        delete []S;
    }
    void Display();
    int StackTop();
    int IsFull();
    int IsEmpty();
    void Push(char x);
    char Pop();
    int Peek(int pos);
    char* ConverttoPostfix(char *exp);
    int IsOperand(char x);
    int Pre(char x);
}
```

```c++
char* Stack::ConverttoPostfix (char *exp){
    top=-1;
    char *postfix = new char[strlen(exp)+1];
    int i = 0, j = 0;

    while(exp[i]!='\0'){
        if (IsOperand(exp[i]))
            postfix[j++] = exp[i++];
        else
        {
           if (Pre(exp[i]) > Pre(S[top])){
             Push(exp[i++]);
           }
           else
             postfix[j++] = Pop();      
        }           
    }
    

    while(!IsEmpty()){
       postfix[j++] = Pop(); 
    }
    postfix[j]='\0';
          
   return postfix;
}
```

```c++
int Stack::IsOperand(char x){
   if (x=='+' || x=='-'  || x=='/'  || x=='*')
      return 0;
  else
      return 1;     
}
```

```c++
int Stack::Pre(char x){
  if (x=='+' || x=='-')
    return 1;
  else if (x=='*' || x=='/') 
      return 2;
  return 0;  
}
```

```c++
int Stack::StackTop (){
    if (!IsEmpty())
      return top;
    return -1;   
}
```

```c++
int Stack::IsFull (){
    if (top==size-1)
       return 1;
    return 0;
}
```

```c++
int Stack::IsEmpty (){
    if (top==-1)
      return 1;
    return 0;
}
```

```c++
void Stack::Push (char x){
    if (IsFull())
      cout<<"Stack overflow"<<endl;
    else
    {
      top++;
      S[top] =x; 
    }
    
}
```

```c++
char Stack::Pop (){
    char x=-1;
    if (IsEmpty())
      cout<<"Stack Empty"<<endl;
    else
    {
      x=S[top];
      top--;
    }
    return x;
}
```

```c++
int Stack::Peek (int pos){
    int x=-1;
    int mapPos = top - pos +1;    
    if (mapPos < 0)
      cout<<"Invalid Pos"<<endl;
    else
    {
      x=S[mapPos];
    }
    return x;
}
```

```c++
void Stack::Display (){
   for (int i = top; i>=0; i--)
   {
     cout<<S[i]<<" ";
   }
   cout<<endl;

}
```

```c++
char expr[] ="a+b*c-d/e";
int sz = strlen(expr);
cout<<"Len Array: "<<sz<<endl;
Stack st(sz);
//st.Push('#');
char *postfixy=st.ConverttoPostfix(expr);
cout<<postfixy<<endl;
```
