````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    Node *lchild;
    int data;
    Node *rchild;
}*f=NULL;
````

````c++
class Queue{
private:
  int size;
  int front;
  int rear;
  Node **Q;  
    
public:
    Queue(){
        size=100;
        front = rear = 0;
        Q=new Node*[100];
       
        
    }
    Queue(int sz){
        size=sz;
        front = rear = 0;
        Q=new Node*[sz];
    }
    ~Queue(){
        delete []Q;
    }
    void Display();
    void EnQueue(Node *x);
    Node* DeQueue();
    int IsFull();
    int IsEmpty();
    
}
````

````c++
void Queue::EnQueue(Node *x){
    if (IsFull())
      cout<<"Queue Full"<<endl;
    else
    {
      rear = (rear+1)%size;
      if (rear ==0)  rear++;
      Q[rear] =x; 
    }
}
````

````c++
Node* Queue::DeQueue (){
    Node *x=NULL;
    if (IsEmpty())
      cout<<"Queue Empty"<<endl;
    else
    {
      front = (front+1)%size;
      if (front ==0)  front++;
      x=Q[front];
    }
    return x;
}
````

````c++
int Queue::IsFull(){
   if ((rear+1)%size==front){
       return 1;
   }
   return 0;
}
````

````c++
int Queue::IsEmpty(){
    if (rear==front)
      return 1;
    return 0;
}
````

````c++
void Queue::Display (){
   int i = front+1;
   int r =0;
   do
   {   cout<<Q[i]<<" - "<<i<< " ";
       i = (i+1)%size;
       if (i ==0)  i++;
       r =  (rear+1)%size;
       if (r ==0) r++;
   }while(i!= r);
    cout<<endl;
}
````

````c++
class Tree{
public:
    Node *root;  
    
    Tree(){
        root = NULL;
    }
    ~Tree(); 
    void CreateTree();
    
    void Preorder(Node *p);
    void Inorder(Node *p);
    void Postorder(Node *p);
    void Levelorder(Node *p);
    int Height(Node *root);    
} 
````

````c++
Tree::~Tree(){
    Node *p=root;
    while(root)
    {
      root=root->rchild;
      delete p;
      p=root;
    }
}
````

````c++
void Tree::CreateTree(){
    Node *p, *t;
    int x, y, z;
    Queue q(100);
    cout<<"Enter root value: "<<endl;
    cin>>x;
    root = new Node;
    root->data=x;
    root->lchild=root->rchild=NULL;
    q.EnQueue(root);
    
    while(!q.IsEmpty()){
        p=q.DeQueue();
        cout<<"Enter left child of " <<p->data<<endl;
        cin>>y;
        if (y!=-1)
        {
            t = new Node;
            t->data=y;
            t->lchild = t->rchild=NULL;
            p->lchild=t;
            q.EnQueue(t);
        }
        cout<<"Enter right child of "<<p->data<<endl;
        cin>>z;
        if (z!=-1)
        {
            t = new Node;
            t->data=z;
            t->lchild = t->rchild=NULL;
            p->rchild=t;
            q.EnQueue(t);
        }
    }
}
````

````c++
void Tree::Preorder(Node *p){
    if(p)
    {
        cout<<p->data<< " ";
        Preorder(p->lchild);
        Preorder(p->rchild);
    }
}
````

````c++
void Tree::Inorder(Node *p){
   if(p)
   {
        Inorder(p->lchild);
        cout<<p->data<< " ";
        Inorder(p->rchild);
   } 
}
````

````c++
void Tree::Postorder(Node *p){
  if(p)
  {
      Postorder(p->lchild);
      Postorder(p->rchild);
      cout<<p->data<< " ";
  }    
}
````

````c++
void Tree::Levelorder(Node *root){
    Queue q(100);
    
    cout<<root->data<<" ";
    q.EnQueue(root);
    
    while(!q.IsEmpty()){ 
        root=q.DeQueue();
        if (root->lchild){
            cout<<root->lchild->data<<" ";
            q.EnQueue(root->lchild);
        }
        if (root->rchild){
            cout<<root->rchild->data<<" ";
            q.EnQueue(root->rchild);
        }
    }
    
}
````

````c++
int Height(Node *root){
    int x=0, y =0;
    if (root==0)
        return 0;
    x=Height(root->lchild);
    y=Height(root->rchild);
    if(x>y)
        return x+1;
    else
        return y+1;    
}
````

````c++
Tree t;
t.CreateTree();
cout<<"Preorder ";
t.Preorder(t.root);
cout<<endl;
cout<<"Inorder ";
t.Inorder(t.root);
cout<<endl;
cout<<"Postorder ";
t.Postorder(t.root);
cout<<endl;
cout<<"Levelorder ";
t.Levelorder(t.root);
cout<<endl;
````
