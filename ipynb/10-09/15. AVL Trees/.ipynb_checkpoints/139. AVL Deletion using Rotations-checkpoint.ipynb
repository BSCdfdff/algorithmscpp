+ Deletion from AVL Tree

## Deletion from AVL Tree

___

It is the same as deleting from BST. Lets refresh how deletion in BST wil work:

1. Search for key
2. If found delete it
3. Who will take its place:
    + Inorder Predecessor, or 
    + Inorder Successor


But remember AVLs are height balanced BST trees, so as we did when we inserted a new node. We need to follow the same procedure for deletion



For Deletion we have the following rotations:

1. L1 - Rotation
2. L-1 - Rotation
3. L0 - Rotation
4. R1 - Rotation
5. R-1 - Rotation
6. R0 - Rotation


These are the same rotation we did for insertiion, but kust with different names.

Lets conver the first 3 Rotations:




## L1 Rotation

___


Also known as LL Rotation.


We have the following Tree:

```
                                           2-1=1
                    + 0-->                 +---+            <--0 +
                    |                      | 30|                 |
                    |                    / +---+  \              |
                    |                   /          \             |
                    |                  /            \            |
                    |             +---+              +---+       |
              1-0=1 | 1-->        | 20|              |40 |  <--1 | 0
                    |            /+---+              +---+       |
                    |           /                                |
                    |          /                                 |
                    |       +---+                                |
                    |       | 10|                                |
                0   v 2-->  +---+                           <--2 v


```

**Delete Node 40**

**After Deletion**

```
                                           2-0=2
                    + 0-->                 +---+            <--0 +
                    |                      | 30|                 |
                    |                    / +---+                 |
                    |                   /    ^                   |
                    |                  /     |                   |
                    |             +---+      |                   |
              1-0=1 | 1-->        | 20|      |              <--1 |
                    |            /+---+      +                   |
                    |           /        imbalanced              |
                    |          /                                 |
                    |       +---+                                |
                    |       | 10|                                |
                0   v 2-->  +---+                           <--2 v



```

1. Node 30 becomes imbalanced
2. What is reason why node 30 became imbalanced?
    + We deleted of node was done from RHS
    
NOTE for all L1, L-1, L0 Rotations, DELETION will occur from RHS.

3. So if we delete a node from RHS, it will become heavy on LHS
4. That is it is imbalanced on LHS.
5. If we look at node 30 lchild, the bf = 1

$$L1 = \text{lchild bf} = 1$$

6. Thats why its called L1 Imbalance, or L1 Rotation.
7. And the Rotation we will be LL Rotation


**After L1 Rotation**

```
                                           +---+
                                           | 20|
                                         / +---+ \
                                        /         \
                                       /           \
                                  +---+           +---+
                                  | 10|           |30 |
                                  +---+           +---+


```

So purpose of this term, is to identify which rotation to perform. That is the lchild bf must be equal to 1.








## L-1 Rotation

___


Also known as LR Rotation.


We have the following Tree:

```
                                           2-1=1
                                                           2-1=1
                         + 0-->            +---+           <--0 +
                         |                 | 30|                |
                         |               / +---+ \              |
                         |              /         \             |
                         |             /           \            |
                         |        +---+           +---+         |
                   0-1=-1| 1-->   | 10|           |40 |    <--1 | 0
                         |        +---+\          +---+         |
                         |              \                       |
                         |               \                      |
                         |                +---+                 |
                         |                |20 |                 |
                     0   v 2-->           +---+            <--2 v




```

**Delete Node 40**
**After Deletion**

```
                                           2-0=2
                         + 0-->            +---+           <--0 +
                         |                 | 30|                |
                         |               / +---+                |
                         |              /    ^                  |
                         |             /     |                  |
                         |        +---+      |                  |
                   0-1=-1| 1-->   | 10|      |             <--1 |
                         |        +---+\     +                  |
                         |              \  imbalanced           |
                         |               \                      |
                         |                +---+                 |
                         |                |20 |                 |
                     0   v 2-->           +---+            <--2 v


```
1. Node 30 becomes imbalanced
2. What is reason why node 30 became imbalanced?
    + We deleted of node was done from RHS
    
NOTE for all L1, L-1, L0 Rotations, DELETION will occur from RHS.

3. So if we delete a node from RHS, it will become heavy on Left Right
4. That is it is imbalanced on LHS.
5. If we look at node 30 lchild, the bf = -1

$$L-1 = \text{lchild bf} = -1$$

6. Thats why its called L-1 Imbalance, or L-1 Rotation.
7. And the Rotation we will be LR Rotation


**After L-1 Rotation**

```
                                           +---+
                                           | 20|
                                         / +---+ \
                                        /         \
                                       /           \
                                  +---+           +---+
                                  | 10|           |30 |
                                  +---+           +---+


```

## L0 Rotation

___


Also known as  Rotation.


We have the following Tree:

```
                                           2-1=1
                  + 0-->                   +---+           <--0 +
                  |                        | 30|                |
                  |                      / +---+ \              |
                  |                     /         \             |
                  |                    /           \            |
                  |               +---+           +---+         |
            1-1=0 | 1-->          | 10|           |40 |    <--1 |
                  |              /+---+\          +---+         |
                  |             /       \                       |
                  |            /         \                      |
                  |         +---+       +---+                   |
                  |         | 5 |       |20 |                   |
              0   v 2-->    +---+       +---+              <--2 v





```
**Delete Node 40**
**After Deletion**

```
                                           2-0=2
                  + 0-->                   +---+           <--0 +
                  |                        | 30|                |
                  |                      / +---+                |
                  |                     /    ^                  |
                  |                    /     |                  |
                  |               +---+      |                  |
            1-1=0 | 1-->          | 10|      |             <--1 |
                  |              /+---+\     +                  |
                  |             /       \  imbalanced           |
                  |            /         \                      |
                  |         +---+       +---+                   |
                  |         | 5 |       |20 |                   |
              0   v 2-->    +---+       +---+              <--2 v


                         


```
1. Node 30 becomes imbalanced
2. What is reason why node 30 became imbalanced?
    + We deleted of node was done from RHS
    
NOTE for all L1, L-1, L0 Rotations, DELETION will occur from RHS.

3. So if we delete a node from RHS, it will become heavy on Left Right
4. That is it is imbalanced on LHS.
5. If we look at node 30 lchild, the bf = 0

$$L0 = \text{lchild bf} = 0$$

6. Thats why its called L0 Imbalance, or L0 Rotation.
7. And the Rotation we will be  Rotation

Note: Here we can perform any of the two L Rotations:

  L1 Rotation
  L-1 Rotation

**After L1 Rotation**

```

                                            1-2=-1
                          + 0-->            +---+           <--0 +
                          |                 | 10|                |
                          |                /+---+ \              |
                          |               /        \             |
                          |              /          \            |
                          |         +---+          +---+         |
                          | 1-->    | 5 |          |30 |    <--1 | 1-0=1
                          |         +---+         /+---+         |
                          |                      /               |
                          |                     /                |
                          |                  +---+               |
                          |                  | 20|               |
                      0   v 2-->             +---+          <--2 v



```


**After L-1 Rotation**

```
                                            2-1=1
                   + 0-->                   +---+           <--0 +
                   |                        | 20|                |
                   |                      / +---+ \              |
                   |                     /         \             |
                   |                    /           \            |
                   |               +---+           +---+         |
             1-0=1 | 1-->          | 10|           |30 |    <--1 | 0
                   |              /+---+           +---+         |
                   |             /                               |
                   |            /                                |
                   |         +---+                               |
                   |         | 5 |                               |
               0   v 2-->    +---+                          <--2 v

```

These will be rotations we perform on LHS, where we delete on RHS
 

## R1, R-1, R0 Rotations

___

Same as above, but we will perform rotations on RHS, where we delete a node on LHS.
