## Check for Loop in LL

___
Here we need to check if a LL

1. Have a Loop
2. Is linear

8-> 5-> 4-> 7-> 3-> 8-> 9


What is a loop? Where the last node is pointing to some node in the link list. So 9 will point to 4. So that is forming a  loop. And the other side, where it is not coming back to some node inside the LL, it is Linear.



Check for Linear:

1. Loop until we get the last node
2. And the last node is null

Check for Loop:

1. Loop the list, using p
2. We use two pointer q and p. (start off at same first node)
3. Where p use 1 step and 1 uses two steps. 
4. If theres a loop, then both will meet, since one is faster than other.
5. They wont meet ever if it is linear.


Lets see if we can form a loop:

```
Node *t1, *t2;

LinkedList linky;
linky.Insert(0,10);
linky.Insert(1,20);
linky.Insert(2,30);
linky.Insert(3,40); 
linky.Insert(4,50);

t1=f->next->next; 
t2=f->next->next->next->next;
t2->next = t1;
```

t1 will point to 30. How is this done? First t1 is set to Head which is first node and that is 10,  then it moves 2 positions to get to 30.

t2 will point to 50. First it is set to HEAD, which is 10. Moves 4 positions to get to it.

So t2 should be pointing to null as its not pointing to anything

But to create a loop, we can make t2 point back to t1, which is 30.

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL,*s=NULL,*result_t=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
  Node *second;

    
public:
    LinkedList(){
        first = NULL;
        second =NULL;

    }
    ~LinkedList(); 
    void Display();
    void Display_s();
    void Display_t();
    void Insert(int index, int x);
    void Insert_s(int index, int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    int CheckLoopLL();
}
````

````c++
int LinkedList::CheckLoopLL(){
    Node *p, *q;
    p=q=first;
    do{
        p=p->next;
        q=q->next;
        //if q is not null, then assign q's next
        q=q?q->next:q;
       
    }while (p && q && p!=q);
    
    if (p==q) 
        return 1;   
    else 
        return 0;
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first; 
    
    cout<<"Start Address of first: "<<first<<endl;
    cout<<"Data of first: "<<first->data<<endl;
    cout<<"Next of first: "<<first->next<<endl;
    cout<<"-----END FIRST---"<<endl;
    
    cout<<"Start Address of last: "<<last<<endl;
    cout<<"Data of last: "<<last->data<<endl;
    cout<<"Next of last: "<<last->next<<endl;
    cout<<"-----END LAST---"<<endl;
    
    
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        
        cout<<"Index i: "<<i<<endl;
        cout<<"Array Value: "<<A[i]<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        last->next=t;
        cout<<"last->next=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
         cout<<"-----END LAST---"<<endl;
        last=t; 
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of last: "<<last<<endl;
        cout<<"Data of last: "<<last->data<<endl;
        cout<<"Next of last: "<<last->next<<endl;
        cout<<"-----END [last=t]---"<<endl;
        
        cout<<"last=t"<<endl;
        cout<<"Start Address of t: "<<t<<endl;
        cout<<"Data of t: "<<t->data<<endl;
        cout<<"Next of t: "<<t->next<<endl;
        cout<<"-----END T---"<<endl;
        
        
    }
    
    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
    
    Node *q=second;
    while(second)
    {
      second=second->next;
      delete q;
      q=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    f = first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
void LinkedList::Display_s(){
    Node *p=second;
    s = second;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}

````

````c++
void LinkedList::Display_t(){
    Node *p=result_t;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    f = first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
   
}    
````

````c++
void LinkedList::Insert_s(int index, int x){
    Node *t, *p=second;
    s = second;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = second;
        second = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}  
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````
