## Finding Duplicates using Bits

___

Lets call this $A$:


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
f\T&i\T&n\T&d\T&i\T&n\T&g\T&\text {\\}0  \\\hline 
_0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\\\hline 
\end{array}
$$

Here we will be using bits to find duplicates in a string. This method is not only applicable for strings/characters but can be used for integers as well. But it is best suited for the former.

In order to use bits for comparison, as first need to familiarise ourselves with bitwise operations:

1. Left Shift <<
2. Bits ORing (also known as Merging)
3. Bits ANDing (also known as Masking)

## Shift Operation

___
Lets call this $H$:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&0\T&0\T&0\T&0\T&1  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

Lets set $H = 1$

What does the following mean:

$$H=H<<1$$

That means all bits are shifted by $1$ place, moving from right to left. And the blank bits on right is filled with zero's

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&0\T&0\T&0\T&1\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

And this new number is $2$.

What happens if we left shift by $2$ places.

$$H=H<<2$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&0\T&0\T&0\T&0\T&1  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$



$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&0\T&0\T&1\T&0\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

We get the result of $4$.

What observation can we make? By shifting to left we increase the number by multiples of $2$.

Just from this fact we can see the result of  $$H=H<<5$$ that will be $32$.

## ANDing Operation

___
$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&0\T&0\T&0\T&0\T&1  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

For explaining $\text{AND}$, we take $2$ digits:

$$a=10 \rightarrow 1010$$
$$a=06 \rightarrow 0010$$

If we now do bitwise $\text{AND}$ of variables $a$ and $b$:

$$a \text {&} b$$

Lets first reference the $\text{AND}$ Table:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline 
0\T&0\T&=0 \\\hline 
1\T&0\T&=0 \\\hline 
0\T&1\T&=0 \\\hline 
1\T&1\T&=1 \\\hline 
\end{array}
$$


And we $\text{AND}$ the two variables:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline 
1\T&0\T&1\T&0 \\\hline 
0\T&1\T&1\T&0 \\\hline 
0\T&0\T&1\T&0 \\\hline 
\end{array}
$$

So what is this he binary form of:

$$ 0010  = 2$$

So $10 \text{ AND } 6$ is equal to 2.

___
For explaining $\text{OR}$, we take $2$ digits:

$$a=10 \rightarrow 1010$$
$$a=06 \rightarrow 0010$$

If we now do bitwise $\text{OR}$  of variables a and b:

$$a \text {|} b$$

Lets first reference the OR Table:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline 
0\T&0\T&=0 \\\hline 
1\T&0\T&=1 \\\hline 
0\T&1\T&=1 \\\hline 
1\T&1\T&=1 \\\hline 
\end{array}
$$


And we $\text{OR}$ the two variables:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline 
1\T&0\T&1\T&0 \\\hline 
0\T&1\T&1\T&0 \\\hline 
1\T&1\T&1\T&0 \\\hline 
\end{array}
$$

So what is this he binary form of:

$$ 1110  = 14$$


So $10 \text{ OR } 6$ is equal to 14.


## Masking

___
Here we have $H$ (with value $16$)

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&1\T&0\T&0\T&0\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

And another variable a, where eveything is zero.

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
0\T&0\T&0\T&0\T&0\T&0\T&0\T&0  \\\hline 
\end{array}
$$

Masking is where we check if bit (in source table $H$) is on or off.

We can see if this is true/false with the help of a.


First we set:
$$ a =1$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
0\T&0\T&0\T&0\T&0\T&0\T&0\T&1  \\\hline
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$


And in order to see if bit $4$ in table $H$ is on we use:

1. Left shift <<
2. AND 

And why do we need to perform $\text{AND}$:

1. If we you get the result of two operations as $1$, then bit it on
2. If we you get the result of two operations as $0$, then bit it off

Then we left shift a four($4$) positions to the left:

$$ a << 4$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
0\T&0\T&0\T&1\T&0\T&0\T&0\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

And we perform AND operation on H and a:

$$ a \text{&} H$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline 
1\T&0\T&0\T&0\T&0 \\\hline 
1\T&0\T&0\T&0\T&0 \\\hline 
1\T&0\T&0\T&0\T&0 \\\hline 
\end{array}
$$

We got the result of $1$, so the bit ($16$ in table $H$) is ON.


So Masking is whether (we checking) if a bit is ON/OFF in memory.

Remember here we use the secondary table to check if value in $H$ is on or off. The result is not applied to $H$.


In Summary:

$$ a << 4$$
$$ a << 4$$
$$ a \text{&} H$$
$$ \text{Bit is ON or OFF}$$


## Merging

___

Is where merging is setting a bit ON (if it is OFF) in memory

Lets say as before we have $16$ (position $3$) ON, but we also want to set $4$ (position $2$) on:


Here we have $H$ (with value $16$). And want want position $2$  (equal to $4$)on as well.

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&1\T&0\T&0\T&0\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

This can be achieved with:

1. Help a another table
2. Left shift
3. OR operation

And why do we need to perform OR (for our case to put bit OFF):

1. In order to put a bit on, that is off. We need to perform an | (OR) operation, as $0 | 1 = 1$


First we set:
$$ a =1$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
0\T&0\T&0\T&0\T&0\T&0\T&0\T&1  \\\hline
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

Then we left shift a two(2) positions to the left:

$$ a << 2$$


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
0\T&0\T&0\T&0\T&0\T&1\T&0\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$

And we perform $\text{OR}$  operation on $H$ and $a$:


$$H= a | H$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline 
1\T&0\T&0\T&0\T&0 \\\hline 
0\T&0\T&1\T&0\T&0 \\\hline 
1\T&0\T&1\T&0\T&0 \\\hline 
\end{array}
$$

We got the result of $1$ in position $2$, and we maintained the existing positions (i.e. $16$)


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|}
\hline 
_7\T&_6\T&_5\T&_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
0\T&0\T&0\T&1\T&0\T&1\T&0\T&0  \\\hline 
_{128}\T&_{64}\T&_{32}\T&_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}
$$


In Summary:

$$ a << 4$$
$$ a << 2$$
$$ H=a | H$$
$$$$
$$ \text{Changes is APPLIED to H}$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|}
\hline
_4\T&_3\T&_2\T&_1\T&_0 \\\hline 
1\T&0\T&1\T&0\T&0 \\\hline 
_{16}\T&_{8}\T&_{4}\T&_{2}\T&_{1} \\\hline 
\end{array}

## Finding Duplicates in a string

___
Using:

1. Masking
2. Merging



Lets call this $A$:


$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}

$$

Similar to hashing we need space, an additional table and we used an additional table of size $26$.

But now we dealing with bits, so we need $26$ bits. But we dealing with bytes and $1 \text{ byte}$ is equal to $8 \text{ bits}$. We need to deal with multiples of $8$.  So the best way to accommodate is $32 \text{bits}$, and $32$. How many bites are that? To get the number of bits we perform $Log 32$.

$$log_2 {32}=5$$

So $32\text{ bits}$ is equal to $\text{4 bytes}$. 


So which data type have these properties or data size. That will be an int. An int is of size $\text{4 bytes}$. 

What do we do with the $4\text{ bytes}$ This will work like a hash table

From the above table, we only need up to $25 \text{ bits}$. And we need to set all positions to 0. And this is done with the following, ```int  H =0;```.

The above table will now be used for finding duplicates.

As with using hashing table, we will find the letter, in $A$, get the ASCII code for it, E.g. $f = 102$. But since we using a table for size $26 \text{ bits}$ (or $32 \text{ bits}$) we need to map it to that size. So $102 - 97 = 5$. This will be the position in table we need to set to $1$. 


Here is the list:

$$  f = 102 - 97 = 5$$
$$  i = 105 - 97 = 8$$
$$  n = 110 - 97 = 13$$
$$  d = 100 - 97 = 3$$
$$  g = 103 - 97 = 6$$



But we need now an extra variable, to be used to store the $1\text{ bit}$, that will be used to shift left on the $H$ table.

Since $f$ is at position $5$, we will perform the following:

Remember we first need to map the ASCII codes to our table.

```
A[i] - 97 = 5
```
so,

```
x = 1
x=x<<A[i] - 97
```


Aside:

So there will be the following operations in this sequence:

**ANDing operation**: (Masking)

This is to check if the bit is set in $H$. This will be used to see if there is a duplicate.


**ORing operation**: (Merging)

This will be used to set the bit in $H$ table to $1$, if it is $0$.



The above will shift to position $5$. then from there we can, we can perform the AND operation, to check if bit exist:

```
if (x&H>0)
  cout<<"
```
So if it exist, then we know that the bit was set previously. And we can print it is a duplicate.

But we did not set it yet. So in the else, we set H, using OR.

```
else
 
 H=x|H;
```

````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
char  A[]="finding";
int  H = 0;
int x=0;
int k=0;

for(int i = 0; A[i]!='\0';i++){
    k = A[i]-97;
    x=1;
    x=x<<k;
    if((x & H)>0)
        cout<<A[i]<<" is a Duplicate"<<endl;
    else
       H=x|H;
}
````
