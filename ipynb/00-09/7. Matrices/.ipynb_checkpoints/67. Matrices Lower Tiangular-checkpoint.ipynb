## Lower Triangular Matrix

___
Lets call this $M$:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
{_1}  \\ 
{_2}  \\
{_3}  \\
{_4}  \\
{_5}  \\
\end{array}
\begin{bmatrix}
a_{11} & 0 & 0 & 0 & 0\\
a_{21} & a_{22} & 0 & 0 & 0\\
a_{31} & a_{32} & a_{33} & 0 & 0\\
a_{41} & a_{42} & a_{43} & a_{44} & 0\\
a_{51} & a_{52} & a_{53} & a_{54} & a_{55}\\
\end{bmatrix}
$$

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{r}
& {_1} &  & {_2} &  & {_3} &  & {_4} &  & {_5}  
\end{array}
$$
$$$$

All the elements/values in the lower triangular part has values > $0$. The rest is $0$.

$$
i = \text { index of row} \\
j = \text { index of column}
$$


Here we have:

$$
\require{cancel} 
M[i,j] = 0,  \text{    if } i<j 
$$


$$
\require{cancel} 
M[i,j] \bcancel{=} 0,  \text{    if } i>j 
$$


That is, all the elements of matrix is zero, where index $i$ is less than index $j$. Where  index $i$ are greater than $j$, all the the values are greater than $0$



## Non Zero Elements ($\require{cancel} \bcancel{=} 0$)

___

First row has 1 non zero elements, Second row has 2  non zero elements,  Third row has 3  non zero elements,etc

Non Zero elements:

$$1+2+3+4+5 = 15 \text{elements}$$

So for any $n$ number of elements:

$$1+2+3...+n$$

Which has the following formula:

$$\text{Non-Zero Elements } = \frac {n(n+1)}{2}$$

In our case:

$$\frac {5(5+1)}{2} = \frac {5(6)}{2} = \frac {30}{2} = 15  $$





## Zero Elements

___


And since we have a square matrix: $n\times{n} = n^2$.

Thats is, to get the remaining rows and colums:

$$\text{Zero Elements } = \text{Total } - \frac {n(n+1)}{2}$$
$$$$
$$n^2 - \frac {n(n+1)}{2} $$
$$$$
$$= \frac{n(n-1)}{2}$$
$$$$

In our case:

$$\frac {5(5-1)}{2} = \frac {5(4)}{2} = \frac {20}{2} = 10  $$

Lets build up a formula for accessing a specific element:



To access ```M[4,3]```:

That will be in 4th row and 3rd column. So in our array that is index 8. How do we get to index 8

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|r|r|r|r|}
\hline 
a_{11}\T&a_{21}\T&a_{22}\T&a_{31}\T&a_{32}\T&a_{33}\T&a_{41}\T&a_{42}\T&a_{43}\T&a_{44}\T&a_{51}\T&a_{52}\T&a_{53}\T&a_{54}\T&a_{55}   \\\hline

\T&\T&\T&\T&\T&\T&\T&\T&*\T&\T&\T&\T&\T&\T&  \\\hline 
\end{array}

So in order to get to index $8$,  need to skip $3$ rows (row $1$, row $2$, row $3$). So row $1$ has $1$ element, row $2$ has $2$ elements, row $3$ has $3$ elements:

$$1+2+3$$


Now after skipping 3 rows, we at the beginning of the 4th row. No how many element do we need to move ahead to get to ```A[4,3]``` ?. We need to move 2 elements ahead:

$$[1+2+3]+2=8 \text{ (index)}$$


Lets take another example

To access ```M[5,4]```, that is index $13$:



$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|r|r|r|r|}

_0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8\T&_9\T&_{10}\T&_{11}\T&_{12}\T&_{13}\T&_{14}  \\\hline 
_{row 1}\T&_{row 2}\T&\T&_{row 3}\T&\T&\T&_{row 4}\T&\T&\T&\T&_{row 5}\T&\T&\T&\T&  \\\hline
\T&\T&\T&\T&\T&\T&\T&\T&\T&\T&\T&\T&\T&*\T&  \\\hline 
\end{array}
$$


So in order to get to index $13$,  need to skip $4$ rows (row $1$, row $2$, row $3$, row $4$). So row $1$ has $1$ element, row $2$ has $2$ elements, row $3$ has $3$ elements, row $4$ has $4$ elements::

$$1+2+3+4$$

Now after skipping 4 rows, we at the beginning of the 5th row. No how many element do we need to move ahead to get to ```A[5,4]``` ?. We need to move 3 elements ahead:

$$[1+2+3+4]+3=13 \text{ (index)}$$


We can now formulize:

For any index i, j:

$$\frac {i(i-1)}{2} + j -1 $$

In our examples:

```M[4,3]```:
$$$$
$$\frac {4(4-1)}{2} + 3 -1 = \frac {4(3)}{2} + 2 $$
$$$$
$$\frac {12}{2} + 2 = 6 + 2 = 8 $$

```M[5,4]```:
$$$$
$$\frac {5(5-1)}{2} + 4 -1 = \frac {5(4)}{2} + 3 $$
$$$$
$$\frac {20}{2} + 3 = 10 + 3 = 13 $$

## Set size using the row major formula

___

$$
\text{Non-Zero Elements } = \frac {n(n+1)}{2}
$$

````c++
class LowerTri{
private:
    int *A; 
    int size;
    int length;

public:
    LowerTri(){
        size=2;
        A=new int[2*(2+1)/2];
        length = 0;
        
    }
    LowerTri(int sz){
        size=sz;
        A=new int[sz*(sz+1)/2];
        length = 0;
        
    }
    ~LowerTri(){
        delete []A;
    }
    void DisplayRowM();
    void DisplayColM();
    void SetRowM(int i, int j, int x );
    void SetColM(int i, int j, int x );
    int GetRowM(int i, int j);
    int GetColM (int i, int j);
    
    
}
````

## Row Major - Get/Set Non zero values

Will only be present when if ```(i>=j)``` and the row major formula for getting the position in our single dim array:

$$
\frac {i(i-1)}{2} + j -1
$$

````c++
int LowerTri::GetRowM(int i, int j){
    if (i>=j)
      return A[i*(i-1)/2 + j -1];
    return 0;
}
````

````c++
void LowerTri::SetRowM(int i, int j, int x ){
    if (i>=j)
      A[i*(i-1)/2 + j -1]=x;
}
````

## Column Major - Get/Set Non zero values

Will only be present when if ```(i>=j)``` and the row major formula for getting the position in our single dim array:

$$n(j-1) - \frac {(j-2)(j-1)}{2} + (i - j) $$

````c++
void LowerTri::SetColM(int i, int j, int x ){
    if (i>=j)
      A[j*(j-1) - (j-2)*(j-1)/2 + (i - j)]=x;
}
````

````c++
int LowerTri::GetColM(int i, int j){
    if (i>=j)
      return A[j*(j-1) - (j-2)*(j-1)/2 + (i - j)];
    return 0;
}
````

## Display Non zero values

Will only be present when if ```(i>=j)``` and the row major formula for getting the position in our single dim array:

$$
\frac {i(i-1)}{2} + j -1
$$

````c++
void LowerTri::DisplayRowM(){
     for(int i=1; i<= size; i++){
        for(int j=1; j<=size; j++){
            if (i>=j)
               cout<<A[i*(i-1)/2 + j -1]<<" ";
            else
               cout<<"0 "; 
        }
        cout<<endl;
    }
}
````

````c++
void LowerTri::DisplayColM(){
    for(int i=1; i<= size; i++){
        for(int j=1; j<=size; j++){
            if (i>=j)
               cout<<A[j*(j-1) - (j-2)*(j-1)/2 + (i - j)]<<" ";
            else
               cout<<"0 "; 
        }
        cout<<endl;
    }
}
````

## Row Major

````c++
int d;

cout<<"Enter the Dimensions or Size of array:";
cin>>d;

LowerTri ML(d);

int x;
cout<<"Enter All Elements";
for (int i = 1;i<=d; i++){
    for (int j = 1;j<=d; j++){
       cin>>x;
       ML.SetRowM(i,j,x); 
        
    }
}    
cout<<"****Display****"<<endl;
ML.DisplayRowM();
````

````c++
int dd;

cout<<"Enter the Dimensions or Size of array:";
cin>>dd;

LowerTri MLL(dd);

int xx;
cout<<"Enter All Elements";
for (int ii = 1;ii<=dd; ii++){
    for (int jj = 1;jj<=dd; jj++){
       cin>>xx;
       MLL.SetColM(ii,jj,xx); 
        
    }
}    
cout<<"****Display****"<<endl;
MLL.DisplayColM();
````
