## Delete at given position

___

1. Get point ```p``` and set it equal to first. (```p=first```)
2. Get also a $\text{ trailing }$ point ```q```, so allow us to get access to previous node. (```q``` will be one step behind ```p```)(```q=p; p = p->next```)
3. Remember we loop $n-1$ less (```n``` = index to be deleted)
4. Once we get to index:
5. Set ```q's next``` equal to ```p's next``` (```q->next = p->next```)
4. Delete ```p```
5. Optionally , get the value of ```p```, before deleting it, and assign it to x (```x = p->data```)

````c++
#include <iostream>
#include <climits>
#include <math.h>
#define INSERTION_OPERATOR operator<<
#define EXTRACTION_OPERATOR operator>>
#define ADDITION_OPERATOR operator+
using namespace std;
````

````c++
class Node{
public:
    int data;
    Node *next;
}*f=NULL;
````

````c++
class LinkedList{
private:
  Node *first;
    
public:
    LinkedList(){
        first = NULL;
    }
    ~LinkedList(); 
    void Display();
    void Insert(int index, int x);
    void InsertLast(int x);
    LinkedList(int A[], int n);
    int Delete(int index);
    int Length();
    void InsertSorted(int x);
   
}
````

````c++
int LinkedList::Delete(int index){
    Node *p, *q=NULL;
    int x = -1;
    if(index <0 || index > Length())
        return -1;  
    if(index==0){
        p=first;
        first = first->next;
        x=p->data; 
        delete p;
    }
    else{
        p=first;
        for(int i =0; i<index;i++)
        {
            q=p;
            p=p->next;
        }
        q->next=p->next;
        x=p->data;
        delete p;
    }  
    return x;
}
````

````c++
void LinkedList::InsertSorted(int x){
    Node *t=NULL,*q=NULL,*p=first;
    
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    //if it is first node
    if (p==NULL)
        first=t;
    else{    
        while(p && p->data < x){
                cout<<"p data: "<<p->data<<endl;
                q=p;
                p=p->next;
        }
        t->next=p;
        q->next=t;
    }
}
````

````c++
LinkedList::LinkedList(int A[], int n){
    Node *last, *t;
    int i = 0;
    
    first= new Node;
    first->data=A[0];
    f = first;
    first->next=NULL;
    last = first;
    
    for(i=1; i<n;i++)
    {
        t = new Node;
        t->data=A[i];
        t->next=NULL;
        last->next=t;
        last=t; 
    }    
}
````

````c++
LinkedList::~LinkedList(){
    Node *p=first;
    while(first)
    {
      first=first->next;
      delete p;
      p=first;
    }
}
````

````c++
void LinkedList::Display(){
    Node *p=first;
    while(p){
        cout<<p->data<<" ";
        p=p->next;
    }
    cout<<endl;
}
````

````c++
int LinkedList::Length(){
    Node *p=first;
    int len=0;
    while(p){
        len++;
        p=p->next;
    }
    return len;
}
````

````c++
void LinkedList::InsertLast(int x){
    Node *t=NULL, *p=first;
    Node *last=NULL;
    t = new Node;
    t->data = x;
    t->next=NULL;

    //count number of elements
    //loop till last node.
    for(int i =0; i<Length()-1;i++){
      cout<<"p:"<<p->data<<endl;  
      p=p->next;
    }
    
    last = p;
    
    //There are no nodes
    if(first==NULL)
        first=last=t;
    else{
        last->next=t;
        last=t;
    }
} 
````

````c++
void LinkedList::Insert(int index, int x){
    Node *t, *p=first;
    
    if(index <0 || index > Length())
        return;
    t = new Node;
    t->data = x;
    t->next=NULL;
    
    if(index==0){
        t->next = first;
        first = t;
    }
    else{
        for(int i =0; i<index-1;i++)
            p=p->next;
        t->next=p->next;
        p->next=t;
    }
}    
````

## Delete First

````c++
LinkedList linky;
linky.Insert(0,3);
linky.Insert(1,7);
linky.Insert(2,9);
linky.Insert(3,15);
linky.Insert(4,20);
linky.Display();
cout<<"Deleted: "<<linky.Delete(0)<<endl;
linky.Display();
````
