## Rotation using Insertion

___

We can use Rotations athe the time of creating AVL tree, that is during insertion.

What type of rotations do we get:

1. LL - Rotation
2. RR - Rotation
3. LR - Rotation
4. RL - Rotation

Remember the idea behind AVL trees, to balance the tree. So when we create or insert into a tree, the tree will become imbalanced. So to bring the tree back to balanced state, we rotate the nodes of a tree in such a way, so bring the tree back to balance.

Here we go...



## Left Left (LL) - Rotation

___



**1. Initial**

```
                                                1-0=1
                         + 0-->                 +---+         <--0 +
                         |                      |30 |              |
                         |                    / +---+              |
                         |                   /                     |
                         |                  /                      |
                         |            +---+                        |
                     0   | 1-->       |20 |                        |  0
                         v            +---+                        v

```


**2. After insert**
```



                                                2-0=2
                         + 0-->                 +---+         <--0 +
                         |                      |30 |              |
                         |                    / +---+              |
                         |                   /    ^                |
                         |                  /     |                |
                         |            +---+       |                |
                  1-0=1  | 1-->       |20 |       |                |
                         |           /+---+       |                |
                         |          /             |                |
                         |         /              |                |
                         |      +---+             +                |
                         |      |10 |         imbalanced           |
                     0   v 2--> +---+                              v



```

**3. Perform Rotation**








From the pic we maek the following observations:

1. We place finger under 30
2. And pull it a down, like a string


**4. After Rotation**

```
                                                1-1=0
                         + 0-->                 +---+         <--0 +
                         |                      |20 |              |
                         |                    / +---+ \            |
                         |                   /         \           |
                         |                  /           \          |
                         |            +---+              +---+     |
                    0    v 1-->       |10 |              |30 |<--1 v  0
                                      +---+              +---+



```

Observations:

1. We rotated around 30.
2. If LL imbalanced, we rotate clockwise, that is to RHS
    + Last insert was on LHS (LL Imbalance)
3. We ONLY ever count 3 nodes (see this rule later, when we have more nodes), when we do rotation
4. After rotation, the bf on that node (30) on which we perfromed the rotation became 0.


## Right Right (RR) - Rotation

___




**1. Initial**

```
                                              
                      
                                                0-1=-1
                         + 0-->                 +---+         <--0 +
                         |                      |10 |              |
                         |                      +---+ \            |
                         |                             \           |
                         |                              \          |
                         |                               +---+     |
                    0    v 1-->                          |20 |<--1 v  0
                                                         +---+        


```


**2. After insert**
```



      
                                       0-2=-2
                         + 0-->        +---+                   <--0 +
                         |             |10 |                        |
                         |             +---+ \                      |
                         |               ^    \                     |
                         |               |     \                    |
                         |               |      +---+               |
                         | 1-->          |      |20 |          <--1 | 0-1=-1
                         |               |      +---+\              |
                         |               |            \             |
                         |               +             \            |
                         |          imbalanced        +---+         |
                         |                            |30 |         |
                         v 2-->                       +---+    <--2 v 0



```

**3. Perform Rotation**






From the pic we maek the following observations:

1. We place finger under 10
2. And pull it a down, like a string


**4. After Rotation**

```
                                                1-1=0
                         + 0-->                 +---+         <--0 +
                         |                      |20 |              |
                         |                    / +---+ \            |
                         |                   /         \           |
                         |                  /           \          |
                         |            +---+              +---+     |
                    0    v 1-->       |10 |              |30 |<--1 v  0
                                      +---+              +---+



```

Observations:

1. We rotated around 10.
2. If RR imbalanced, we rotate counter clockwise, that is to LHS

4. Thats it, with all rotations (4 covering here), we get the SAME result for 3 nodes.



## Left Right (LR) - Rotation

___


**1. Initial**

```
                                                1-0=1
                         + 0-->                 +---+         <--0 +
                         |                      |30 |              |
                         |                    / +---+              |
                         |                   /                     |
                         |                  /                      |
                         |            +---+                        |
                     0   | 1-->       |10 |                        |  0
                         v            +---+                        v


```


**2. After insert**
```

                                                2-0=2
                          + 0-->                +---+     <--0 +
                          |                     |30 |          |
                          |                   / +---+          |
                          |                  /    ^            |
                          |                 /     |            |
                          |           +---+       |            |
                   2-2=0  | 1-->      |10 |       +       <--1 |
                          |           +---+\ imlalanced        |
                          |                 \                  |
                          |                  \                 |
                          |                 +---+              |
                          |                 |20 |              |
                   0      v 2-->            +---+         <--2 v


``` 

**3. Perform Rotation**







From the pic we maek the following observations:

1. We place finger under 10
2. And pull it a down, like a string to LHS
3. Then we perform LL rotation for 30.


**4. After Rotation Step 1**

```
                         + 0-->                 +---+         <--0 +
                         |                      |30 |              |
                         |                    / +---+              |
                         |                   /    ^                |
                         |                  /     |                |
                         |            +---+       |                |
                  1-0=1  | 1-->       |20 |       |                |
                         |           /+---+       |                |
                         |          /             |                |
                         |         /              |                |
                         |      +---+             +                |
                         |      |10 |         imbalanced           |
                     0   v 2--> +---+                              v



```

**4. After Rotation Step 2**

 ```
                         + 0-->                 +---+         <--0 +
                         |                      |20 |              |
                         |                    / +---+ \            |
                         |                   /         \           |
                         |                  /           \          |
                         |            +---+              +---+     |
                    0    v 1-->       |10 |              |30 |<--1 v  0
                                      +---+              +---+


```

Observations:

1. Here performed double rotation
    + First around 10
        + to LHS
    + Then around 30
        + to RHS
1. If LR imbalanced, we rotate counter clockwise, then clockwise
    + Route: go LHS
    + Then RHS (LR - Imbalance)
3. We end up with same result as LR rotation.
4. Thats it, with all rotations (4 covering here), we get the SAME result for 3 nodes.
