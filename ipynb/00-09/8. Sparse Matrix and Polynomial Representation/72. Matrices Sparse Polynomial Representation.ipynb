____

2. Compressed Spare Row

## Coordinate List/Three column Representation

___
For every non zero element we need three things:

1. Row Number
2. Column Number
3. Element/Value

So if we scan the table able, we have value $3$ in coordicate ```[3,8]```, and we have value $8$ in ```[3,8]```, and so on and it can represented as follows



$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
_\T&_0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8\T&_9\T&_{10}  \\\hline 
\newcommand\T{\Rule{0pt}{1em}{.3em}}
_{row}\T&8\T&1\T&2\T&2   \T&4\T&6\T&7\T&8\T&8\T&\T&  \\\hline 
_{col}\T&9\T&8\T&3\T&6   \T&1\T&3\T&4\T&2\T&5\T&\T&  \\\hline 
_{val}\T&8\T&3\T&8\T&{10}\T&4\T&2\T&6\T&9\T&5\T&\T&  \\\hline 
\end{array}
$$

The first column contains totals:

1. Number of rows
2. Number of cols
3. Total non zero values


Commonly used method

## Compressed Sparse Rows

___
Here a Sparse Matrix is represented using three (3) arrays:

1. Array for List of Non zero elements ($A$)
2. Array for (Cumulative) Rows ($IA$)
3. Array for Columns ($JA$)


Array for list of non zero elements, scanning each row, and listing them in the same order they are appearing:

Lets call this $A$:

$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
3\T&8\T&10\T&4\T&2\T&6\T&9\T&6  \\\hline 
\end{array}
$$

$$$$


(ele refers to all nown zero elements)

Then we need an array for rows. With the first value being $0$, Now we start counting the number of elements per row, and keep track of the previous count. That is in $1st$ row, we have $1$ ele. In the $2nd$ row we have $2$ ele, but plus the $1$ ele in $1st$ row that is $3$ ele. In the $3rd$ we have $0$ ele, so the column count remains at $3$. And so we continue.

Lets call this $IA$:
$$
\newcommand\T{\Rule{0pt}{.1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
0\T&_1\T&_2\T&_3\T&_4\T&_5\T&_6\T&_7\T&_8  \\\hline 
\newcommand\T{\Rule{0pt}{1em}{.3em}}
_{Cum Rows}\T&1\T&3\T&3\T&4\T&4\T&5\T&6\T&8  \\\hline 
\end{array}
$$

Finally we need an array for where the values appear in the columns. The value in this array ($JA$), will match up where it appears in the $A$ array.


Again we start scanning each row, and pick the non zero value, e.g. $3$ and see in which col it appears, which is in col $8$. Next value $8$ appears in col $3$. And value $10$, appears in col $6$, and so on.


Lets call this $JA$:

$$
\newcommand\T{\Rule{0pt}{1em}{.3em}}
\begin{array}{|r|r|r|r|r|r|r|r|r|r|}
\hline 
8\T&3\T&6\T&1\T&3\T&4\T&2\T&5  \\\hline 
\end{array}
$$
